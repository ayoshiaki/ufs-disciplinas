## 1. Aquecimento: Tudo em Python é objeto

Antes de *definirmos* nossas próprias classes, vamos observar que você já vem usando objetos desde o primeiro dia de Python. Strings, listas, números — todos são objetos, e cada um pertence a uma classe (um *tipo*).

Execute a célula abaixo:

````python
# Tudo em Python é um objeto, e cada objeto tem um tipo (classe)
nome = "Podrão"
precos = [18.50, 12.00, 15.00]
quantidade = 3

print(type(nome))        # <class 'str'>
print(type(precos))      # <class 'list'>
print(type(quantidade))  # <class 'int'>

# Cada objeto "sabe" fazer coisas — esses são seus MÉTODOS
print(nome.upper())         # método de str
print(precos.append(20.00)) # método de list (retorna None mas modifica)
print(precos)
````

Repare que:

- `type(x)` revela a classe de um objeto.
- `nome.upper()` é um **método** — uma função ligada ao objeto `nome`.
- `precos.append(20.00)` é outro método, que modifica a lista internamente.

A **Programação Orientada a Objetos** é simplesmente a prática de **criar seus próprios tipos** (classes), com seus próprios atributos e métodos, do mesmo jeito que `str` e `list` têm os deles.

## 2. O Problema: A "Sopa de Variáveis"

Imagine que você foi contratado para programar o sistema de um food truck que vende lanches na rua. Seu primeiro instinto pode ser criar variáveis separadas para cada lanche:

```
lanche1_nome, lanche1_preco, lanche1_ingredientes ...
lanche2_nome, lanche2_preco, lanche2_ingredientes ...
```

Isso funciona com 2 lanches. Mas com **30 itens no cardápio**? E se você precisar adicionar um novo atributo como `tempo_preparo`? Você teria que criar 30 novas variáveis manualmente.

Execute o código abaixo e sinta o problema:

````python
# Abordagem sem classes — o problema fica claro já com 3 lanches
lanche1_nome = "X-Tudo"
lanche1_preco = 18.50
lanche1_ingredientes = "pão, hambúrguer, ovo, presunto, queijo, bacon, alface, tomate"

lanche2_nome = "Dogão"
lanche2_preco = 12.00
lanche2_ingredientes = "pão de hot-dog, salsicha, vinagrete, batata palha"

lanche3_nome = "Açaí 500ml"
lanche3_preco = 15.00
lanche3_ingredientes = "açaí, granola, banana, leite condensado"

print(f"O lanche '{lanche1_nome}' custa R${lanche1_preco:.2f}.")

# PROBLEMA: como calcular o total de um pedido com os 3 itens?
# Teríamos que somar variáveis avulsas manualmente.
total = lanche1_preco + lanche2_preco + lanche3_preco
print(f"Total do pedido: R${total:.2f}")
````

**Por que isso é ruim?**

- Os dados estão **soltos** — não há conexão estrutural entre `lanche1_nome` e `lanche1_preco` além do prefixo.
- A **lógica** (como calcular o total) não está vinculada aos dados.
- Erros de digitação (`lanche1_prco`) não são detectados automaticamente.
- Escalar para 30 itens é **inviável**.
- Adicionar um novo campo exige mudar o código em vários lugares.

A **Programação Orientada a Objetos** resolve exatamente isso: agrupa **dados** e **comportamentos** relacionados dentro de uma única estrutura — a **classe**.

## 3. A Solução: Classes e Objetos

A ideia central da POO é unir **dados** (o que o objeto *sabe*) e **comportamentos** (o que o objeto *faz*) em uma única estrutura chamada **classe**.

### Analogia: Receita × Lanche Pronto

Uma **classe** é como uma receita: descreve o que um lanche *tem* (ingredientes, preço) e o que ele *pode fazer* (ser descrito, ser preparado). Mas a receita em si não é comida — ela é só o molde.

Um **objeto** (ou instância) é o lanche pronto, ocupando espaço real (na bancada, ou na memória do computador). Você pode fazer vários lanches a partir da mesma receita, e cada um é independente.

```
  CLASSE                          INSTÂNCIAS (OBJETOS)
┌─────────────────────┐           ┌──────────────────────────┐
│      Lanche         │           │  xtudo                   │
│  (projeto/molde)    │  ──────►  │  nome  = "X-Tudo"        │
│                     │           │  preco = 18.50           │
│  - nome: str        │           └──────────────────────────┘
│  - preco: float     │           ┌──────────────────────────┐
│  - ingredientes: str│  ──────►  │  dogao                   │
│  - descrever()      │           │  nome  = "Dogão"         │
└─────────────────────┘           │  preco = 12.00           │
   Define o layout                └──────────────────────────┘
                                       Cada um ocupa memória
```

### 3.1 Definindo a primeira classe

````python
class Lanche:
    """
    Representa um lanche genérico do food truck.

    Atributos:
        nome (str):         Nome do lanche.
        preco (float):      Preço em reais.
        ingredientes (str): Lista de ingredientes como texto.
    """

    def __init__(self, nome, preco, ingredientes):
        # __init__ é o CONSTRUTOR: executado automaticamente ao criar um objeto.
        # 'self' é a referência ao próprio objeto sendo criado.
        self.nome = nome
        self.preco = preco
        self.ingredientes = ingredientes

    def descrever(self):
        """Exibe as informações do lanche formatadas."""
        print(f"  🍔 {self.nome} — R${self.preco:.2f}")
        print(f"     Ingredientes: {self.ingredientes}")


# ─── Criando objetos (instâncias) a partir da classe ───
xtudo = Lanche("X-Tudo", 18.50,
               "pão, hambúrguer, ovo, presunto, queijo, bacon, alface, tomate")
dogao = Lanche("Dogão", 12.00,
               "pão de hot-dog, salsicha, vinagrete, batata palha")

# Cada objeto tem seu próprio estado na memória
print("=== Cardápio do Podrão ===")
xtudo.descrever()
print()
dogao.descrever()

# E podemos confirmar que são do tipo Lanche
print()
print(f"xtudo é do tipo: {type(xtudo).__name__}")
print(f"xtudo e dogao são o mesmo objeto? {xtudo is dogao}")
````

### Conceitos apresentados

- **Classe** (`Lanche`): estrutura que define um novo *tipo*. Não ocupa memória para os atributos — apenas descreve o layout.
- **Atributos de instância** (`self.nome`, `self.preco`, `self.ingredientes`): variáveis que retêm o estado individual de cada objeto. Cada instância tem os seus.
- **Métodos** (`descrever`): funções ligadas à classe que operam sobre o estado via `self`.
- **Objeto / Instância** (`xtudo`, `dogao`): alocação concreta em memória. Criada com `Lanche(...)`.
- **`__init__`**: método especial chamado de *construtor*, executado automaticamente na criação do objeto.
- **`self`**: referência ao próprio objeto. É como o método descobre "de quem" são os atributos que está acessando.

> 💡 **Nota sobre `self`:** em Python, `self` é apenas uma convenção — poderia se chamar qualquer coisa. Mas **sempre** use `self`, porque todo programador Python do mundo espera esse nome.

## 4. Atributos de Classe vs. Atributos de Instância

Até agora, todos os atributos que criamos vivem dentro de `__init__` e começam com `self.` — são **atributos de instância**. Cada objeto tem os seus, independentes.

Mas existe outra categoria: **atributos de classe**, que são *compartilhados* por todas as instâncias. Eles são definidos **no corpo da classe**, fora de qualquer método.

### Quando usar cada um?

| Tipo | Definido onde | Pertence a | Exemplo típico |
|---|---|---|---|
| **Instância** | dentro de `__init__`, com `self.` | cada objeto tem o seu | `self.nome`, `self.preco` |
| **Classe** | no corpo da classe, sem `self.` | compartilhado por todos | constantes, contadores, configurações globais |

### Exemplo: taxa de serviço do food truck

````python
class LancheComTaxa:
    """Lanche do Podrão com taxa de serviço compartilhada entre todos."""

    # ↓ Atributo de CLASSE — compartilhado por TODAS as instâncias
    taxa_servico = 0.10   # 10% em todos os lanches
    nome_estabelecimento = "Podrão Food Truck"

    def __init__(self, nome, preco):
        # ↓ Atributos de INSTÂNCIA — cada objeto tem os seus
        self.nome = nome
        self.preco = preco

    def preco_final(self):
        """Calcula o preço incluindo a taxa de serviço compartilhada."""
        return self.preco * (1 + LancheComTaxa.taxa_servico)

    def descrever(self):
        print(f"  🍔 {self.nome} ({LancheComTaxa.nome_estabelecimento})")
        print(f"     Preço base: R${self.preco:.2f}")
        print(f"     Com taxa ({LancheComTaxa.taxa_servico*100:.0f}%): R${self.preco_final():.2f}")


a = LancheComTaxa("X-Tudo", 18.50)
b = LancheComTaxa("Dogão", 12.00)

a.descrever()
print()
b.descrever()

print("\n--- Agora o dono decide subir a taxa para 15% ---")
LancheComTaxa.taxa_servico = 0.15   # muda para TODAS as instâncias de uma vez!

a.descrever()
print()
b.descrever()
````

Observe o efeito importante: **mudando `LancheComTaxa.taxa_servico` uma única vez, o preço final de *todos* os lanches muda automaticamente** — porque o atributo pertence à classe, não a cada instância.

### ⚠️ A armadilha clássica: atributos de classe mutáveis

Existe um erro muito comum em Python que acontece quando se usa uma **lista** (ou outro mutável) como atributo de classe sem querer. Veja:

````python
class LancheBugado:
    ingredientes = []   # ⚠️ atributo de CLASSE — lista compartilhada!

    def __init__(self, nome):
        self.nome = nome

    def adicionar_ingrediente(self, ing):
        self.ingredientes.append(ing)   # modifica a lista compartilhada!


x = LancheBugado("X-Tudo")
d = LancheBugado("Dogão")

x.adicionar_ingrediente("bacon")
d.adicionar_ingrediente("salsicha")

print(f"Ingredientes do X-Tudo: {x.ingredientes}")
print(f"Ingredientes do Dogão:  {d.ingredientes}")
print("😱 Os dois lanches compartilham a MESMA lista!")
````

### A forma correta

````python
class LancheCorreto:
    def __init__(self, nome):
        self.nome = nome
        self.ingredientes = []   # ✔ atributo de INSTÂNCIA — cada objeto tem a sua lista

    def adicionar_ingrediente(self, ing):
        self.ingredientes.append(ing)


x = LancheCorreto("X-Tudo")
d = LancheCorreto("Dogão")

x.adicionar_ingrediente("bacon")
d.adicionar_ingrediente("salsicha")

print(f"Ingredientes do X-Tudo: {x.ingredientes}")
print(f"Ingredientes do Dogão:  {d.ingredientes}")
print("✔ Cada lanche tem sua própria lista.")
````

**Regra prática:** se o atributo é um valor *mutável* (lista, dicionário, conjunto), **quase sempre** ele deve ser de instância, criado dentro de `__init__`. Use atributos de classe para **constantes** ou **configurações compartilhadas** imutáveis.

## 5. Prática guiada: classe `Cardapio`

Vamos juntar tudo o que vimos. Nesta prática guiada, construiremos uma classe `Cardapio` que gerencia uma lista de `Lanche`. Isso aproxima o código de um sistema real.

Execute e estude o código abaixo antes de partir para os exercícios finais.

````python
class Lanche:
    """Representa um lanche genérico do food truck."""

    def __init__(self, nome, preco, ingredientes):
        self.nome = nome
        self.preco = preco
        self.ingredientes = ingredientes

    def descrever(self):
        print(f"  🍔 {self.nome:.<25} R${self.preco:.2f}")
        print(f"     {self.ingredientes}")


class Cardapio:
    """Gerencia a coleção de lanches disponíveis no food truck."""

    # Atributo de classe: nome do estabelecimento
    nome_truck = "Podrão"

    def __init__(self):
        self.lanches = []   # instância: cada cardápio tem sua própria lista

    def adicionar(self, lanche):
        self.lanches.append(lanche)

    def mostrar(self):
        print(f"=== Cardápio do {Cardapio.nome_truck} ===")
        for lanche in self.lanches:
            lanche.descrever()
        print(f"({len(self.lanches)} itens disponíveis)")

    def preco_medio(self):
        if not self.lanches:
            return 0.0
        total = sum(l.preco for l in self.lanches)
        return total / len(self.lanches)

    def mais_caro(self):
        if not self.lanches:
            return None
        return max(self.lanches, key=lambda l: l.preco)


# ─── Uso ──────────────────────────────────────────────────
cardapio = Cardapio()
cardapio.adicionar(Lanche("X-Tudo", 18.50, "pão, hambúrguer, ovo, queijo, bacon"))
cardapio.adicionar(Lanche("Dogão", 12.00, "pão, salsicha, vinagrete, batata palha"))
cardapio.adicionar(Lanche("Açaí 500ml", 15.00, "açaí, granola, banana"))
cardapio.adicionar(Lanche("Misto Quente", 8.00, "pão, presunto, queijo"))

cardapio.mostrar()
print()
print(f"Preço médio: R${cardapio.preco_medio():.2f}")
caro = cardapio.mais_caro()
print(f"Mais caro: {caro.nome} (R${caro.preco:.2f})")
````

Observe que `Cardapio` **contém** uma lista de objetos `Lanche`. Isso é uma primeira amostra de **composição** — um conceito que vamos aprofundar na Aula 3. Por enquanto, repare apenas que um objeto pode perfeitamente guardar outros objetos como atributos.

## 7. Resumo da aula

| Conceito | Definição | Exemplo |
|---|---|---|
| **Classe** | Molde que define um novo tipo (atributos + métodos). Não ocupa memória sozinha. | `class Lanche:` |
| **Objeto / Instância** | Alocação concreta na memória, criada a partir da classe. | `xtudo = Lanche(...)` |
| **`__init__`** | Método especial (construtor) chamado automaticamente ao criar o objeto. | `def __init__(self, ...):` |
| **`self`** | Referência ao próprio objeto dentro dos métodos. | `self.nome = nome` |
| **Atributo de instância** | Variável única de cada objeto; criada com `self.` dentro de `__init__`. | `self.preco` |
| **Atributo de classe** | Variável compartilhada por todas as instâncias; definida no corpo da classe. | `taxa_servico = 0.10` |
| **Método** | Função ligada à classe que opera no objeto via `self`. | `def descrever(self):` |

### Armadilhas a lembrar

1. **Nunca** use listas, dicionários ou outros mutáveis como atributo de classe quando você espera que cada instância tenha o seu. Crie-os em `__init__`.
2. `self` é obrigatório como primeiro parâmetro de todo método de instância. Esquecer disso causa `TypeError` na chamada.
3. `Lanche` (a classe) é diferente de `xtudo` (a instância). A classe é o molde, a instância é o produto.
