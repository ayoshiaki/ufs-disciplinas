## 📝 Como estudar de forma eficaz

Algumas dicas de estudo que funcionam para a maioria dos alunos:

### 1. Leia duas vezes
Na primeira passagem, leia de forma corrida para entender a ideia geral. Na segunda passagem (depois da aula), volte aos trechos que agora fazem mais sentido.

### 2. Faça anotações ativas
Não só sublinhe — escreva na margem (ou em um caderno) **com suas próprias palavras** o que cada seção está dizendo. Se você não consegue reformular, é sinal de que ainda não entendeu.

### 3. Execute os exemplos
Todos os exemplos dos livros são curtos. Abra uma célula de Colab, digite o código e rode. Modifique alguma coisa e veja o que acontece. **Aprender POO só lendo não funciona** — você precisa tocar no código.

### 4. Anote suas dúvidas
Se algo não fez sentido, anote a pergunta e traga para a aula. Perguntas durante a aula valem ouro — significam que você leu e se engajou.

### 5. Releia o capítulo depois da aula
Capítulos que pareceram difíceis antes da aula frequentemente parecem simples depois. A releitura consolida o aprendizado.

### 6. Estude com colegas
Explicar um conceito para outra pessoa é uma das melhores formas de descobrir se você realmente entendeu. Se você travar ao explicar, identifica exatamente onde está a lacuna.
