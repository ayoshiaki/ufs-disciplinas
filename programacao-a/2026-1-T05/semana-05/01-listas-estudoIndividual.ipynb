# Listas

O tipo das listas, predefinido em Python, é talvez um dos tipos mais úteis. Da mesma forma que strings e tuplas, uma lista é um iterável. Portanto, uma lista também pode ser considerada uma sequência de valores. Mas certamente listas têm algumas características próprias que as tornam diferentes de outros iteráveis.

+ Strings são sequências cujos elementos são exclusivamente caracteres. Já os elementos de uma lista podem ser de qualquer tipo.
+ Diferentemente de ranges, listas armazenam os seus elementos na memória.
+ Strings e tuplas são **imutáveis**, ou seja, os seus elementos não podem ser alterados. Já listas são **mutáveis**.
  * Elementos individuais da lista podem ser alterados.
  * Listas podem aumentar ou diminuir de tamanho. Pode-se adicionar elementos extras a uma lista como também pode se eliminar alguns elementos de uma lista.

# Indexação

Da mesma forma que com strings e tuplas, e em geral com qualquer iterável, podemos acessar a um elemento individual da uma lista através da sua posição ou índice. Lembrando que os índices começam sempre em zero. Para isto usa-se a notação de indexação `iteravel[indice]`.

Execute o programa a seguir para conferir os valores de alguns elementos individuais da lista `numeros`

````python
print(numeros)
print(numeros[0])
print(numeros[3])
print(numeros[5])
````

Se uma lista tem `n` elementos, como os índices começam em zero, o índice do último elemento será `n-1`.
Por exemplo, nossa lista `numeros` tem 6 elementos, isso significa que os índices válidos para acesso são 0, 1, 2, 3, 4 e 5. Qualquer índice com valor maior que 5 provocará um erro. Verifique



````python
numeros[6]
````

---
**Atenção**: Quando pretender acessar um elemento de uma lista por um índice tenha muito cuidado de que o índice esteja dentro dos limites permitidos. Tentativas de acesso fora dos límites é uma causa frequente de *bugs*.

---


Em geral, o índice do último elemento de uma lista `lst` sempre será `len(lst) - 1`. Assim, por exemplo, uma forma de pegar o último elemento de `numeros` é

````python
numeros[len(numeros)-1]
````

Uma forma mais *pythônica* de acessar ao último elemento é usando um índice negativo. Índices negativos representam as posições de forma invertida, do final para o início. Confira que valores são retornados

# Slices

Python provê um operador que permite selecionar segmentos de qualquer iterável. Tais segmentos são chamados de fatias ou *slices*. A notação é similar ao indexamento, usando colchetes.

Em geral, o operador `[n:m]` retorna uma nova lista formada pela porção da lista original que começa na posição `n` até a posição `m-1`. Em outras palavras, retorna a porçao que está entre a posição `n` e `m`, incluinda a posição `n` mas excluindo a posição `m`. Aqui um par de exemplos

````python
numeros = [1, 5, 10, 20, 7, 3]

numeros[0:3]
````

````python
numeros[3:6]
````

É possível omitir o primeiro índice do slice. Neste caso o slice começa no zero.

````python
numeros[:3]
````

Também é possível omitir o segundo índice. Neste caso o slice vai até o último caracter

````python
numeros[2:]
````

Ainda, podem-se omitir ambos os índices. Neste caso, cria-se uma cópia da lista.

````python
numeros[:]
````

Slices permitem um terceiro parâmetro, o incremento, que tem o papel similar ao incremento dos ranges. Assim, por exemplo

````python
numeros[1: 6: 2]
````

cria uma lista com elementos de `numeros`, iniciando no índice 1 (inclusive), terminando no índice 6 (exclusive), mas pulando os índices de dois em dois.

O incremento de um slice pode ser negativo. Neste caso os elementos são pegos na ordem inversa. Assim, por exemplo, se queremos construir a lista invertida de `numeros`, podemos fazer com

````python
numeros[::-1]
````

# Listas são mutáveis

Além de acessar uma lista para "ver" (ler) os seus elementos, também é possível acessar para alterar os elementos. O operador de indexação, denotado com colchetes, pode aparecer no lado esquerdo de uma atribuição. Assim, por exemplo, podemos alterar alguns elementos de `numeros` assim

````python
print(numeros) # original

numeros[1] = 100 # altera o segundo elemento de numeros (indexado por 1)
numeros[5] = 200 # altera o sexto elemento da lista numeros

print(numeros) # a lista alterada
````

# Referências e Aliases

Até o momento, temos assumido que uma variável é uma entidade que tem um nome e que aramazena algum valor.
Esta visão é apropriada se consideramos apenas valores imutáveis.
Com valores mutáveis precisamos refinar esta visão. De fato, variáveis não armazenam diretamente valores, mas sim **referências** aos valores. Uma referência é um endereço de memória onde um valor específico está armazenado.

Considere as seguintes atribuições


A primeira atribuição cria uma lista e atribui a sua referência à `numeros1`.
Similarmente, a segunda atribuição cria outra lista, apenas como coincidencia, com os mesmos valores da primeira lista e atribui a sua referência a `numeros2`.
Já a terceira atribuição, atribui a `numeros3` a referência de `numeros1`. Assim, como resultado, ambas variáveis, `numeros1` e `numeros3`, contêm uma mesma referência ou, em outras palavras, se referem à mesma lista.

---
**Terminologia**: Quando duas variáveis se referem a um mesmo objeto, dizemos que há um ***alias*** (apelidamento) entre ambas variáveis.  

---



Execute a seguinte célula

````python
numeros1 = [1, 5, 10, 20, 7, 3]
numeros2 = [1, 5, 10, 20, 7, 3]
numeros3 = numeros1
````

Observe agora qual é o resultado da seguinte comparação

````python
numeros1 == numeros2
````

O resultado deve ter sido `True`. Isto acontece por que a comparação `==` é feita verificando se ambas listas contém as mesmas sequências de valores e no nosso exemplo isto é verdade.

Certamente a comparação `numeros1 == numeros3` também dará `True` como resultado, pois neste caso temos que ambas variáveis se referem à mesma lista e, obviamente, uma lista é igual com ela mesma.

Para verificar a existência de *aliases*, Python provê o operador `is`. Enquanto `==` compara os conteúdos das listas, o operador `is` compara se as referências são as mesmas. Verifique

````python
numeros1 is numeros2
````

Qual foi o resultado? Por que deu esse resultado?<br>

# Listas como parâmetros de funções

Quando se passa uma lista como argumento de uma função, a função obtém uma referência da lista.
Tecnicamente, a função cria uma alias com o respectivo parâmetro.
Logo, se a função modifica a lista, quem chamou também verá as modificações. Por exemplo

````python
def alteraUltimo(ns, m) :
  ns[-1] = m

nums = [10, 0, -10, 15]
alteraUltimo(nums, 500)
print(nums)
````

Ao chamar `alteraUltimo(nums, 500)` na linha 5 do código, o parâmetro `ns` da função ficará em alias com `nums` e portanto a alteração `ns[-1] = m` também alterará `nums`.

# Iterando pelos elementos uma lista

Para iterar *visitando* todos os elementos de uma lista podemos usar um laço `for`. Assim, se queremos imprimir todos os elementos da lista `numeros`, cada elemento em uma linha, podemos fazer

````python
for num in numeros :
  print(num)
````

Como um exemplo mais interessante assumamos que não exista a função `sum` e demos uma definição para ela. Lembrando que `sum` calcula a soma de todos os elementos da lista.
Para isto, usamos uma iteração sobre os elementos da lista e o padrão Acumulador, conveniente a este problema. Assim

  

````python
def sum(ns) :
  soma = 0
  for n in ns :
    soma += n
  return soma

print("numeros =", numeros)
sum(numeros)
````

Ambos exemplos anteriores usam `for` para iterar pela lista. Este tipo de iteração é adequado apenas quando queremos *ler* um por um os elementos de uma lista.

Em alguns casos, podemos precisar alterar todos ou alguns elementos da lista. Nesta situação, a solução é iterar pelos elementos da lista de forma indireta, através de todos os índices possíveis.

Por exemplo, se desejamos definir uma função que, dada uma lista de números `ns`, incremente em um todos os seus elementos, iteramos pelos índices desde `0` até `len(ns)-1` alterando os elementos correspondenes.

````python
def incrementaLista(ns) :
  for i in range(len(ns)) :
    ns[i] += 1

print("Antes: ", numeros)
incrementaLista(numeros)
print("Depois: ", numeros)
````

## Iterar diretamente ou indiretamente (através dos índices)?

Como decidir entre iterar diretamente ou através dos índices? A escolha normalmente é fácil usando a seguinte regra:

+ Se somente preciso ler os valores dentro da lista, considere usar `for` direto.
+ Em outros casos considere iteração através dos índices.

Mas, como acontece com muitas regras, podem haver exceções ...

Aqui algumas situações em que certamente devemos usar iteração pelos índices:
+ Preciso atualizar/modificar os elementos da lista.
+ Preciso acessar a cada vez vários elementos ao mesmo tempo, por exemplo elementos adjascentes.
+ Preciso retornar um índice como resultado.


## Iterar de trás para frente

Imagine que precisamos iterar pelos elementos de uma lista `lst` seguindo a ordem invertida, do último para o início. Uma opção é iterar pelos índices da lista seguindo a ordem inversa utilizando um `for i in range(len(lst)-1, -1, -1)`.

Porém, se a iteração é apenas para leitura dos elementos é mais conveniente utilizar a função **`reversed(lst)`**. Esta função retorna um iterador que permite percorrer `lst` de trás para frente, sem modificar o objeto original. Ela é eficiente em memória, pois não criar uma cópia da lista.

Por exemplo, para imprimir os elementos da lista `numeros` na ordem inversa, fazemos

````python
for elem in reversed(numeros) :
  print(elem)
````

---
**Atenção**: para iterar na ordem inversa não crie uma nova lista com os valores da original, porém invertidos. Por exemplo, usando o slice `numeros[::-1]`. Isto será muito ineficiente se comparado com a alternativa de usar a função `reversed.`

# Estendendo uma lista com mais elementos

Python oferece alguns métodos de listas para adicionar mais elementos. De todos, o mais eficiente é o `append` que permite adicionar um elemento ao final da lista. Veja o comportamento do código a seguir

````python
print(numeros)
numeros.append(1000) # Adiciona 1000 a numeros como mais um elemento
                     # Adição é feita no final de números.
print(numeros)
````

Como um exemplo de aplicação do `append`, imagine que deseja ler uma sequência de valores e armazernar todos os valores lidos em uma lista. Podemos usar o padrão Acumulador, criando inicialmente uma lista vazia e depois iterar lendo cada valor e anexando-o à lista com `append`. Por simplicidade, suponhamos que sabemos qual é a quantidade de números na sequência. O programa que cria a lista é este

````python
def lerNumeros(n) :
  ns = []
  for _ in range(n):
    m = int(input())
    ns.append(m)
  return ns

print(lerNumeros(5))
````

## Exercício

Defina uma função que leia do teclado uma sequência de nomes, cada um em uma linha. A sequência termina quando for inserida uma linha vazia. A sequência de nomes deve ser carregada numa lista que será retornada pela função.

````python
def lerNomes() :
  lista = []
  nome = input()
  while nome :
    lista.append(nome)
    nome = input()
  return lista

lerNomes()
````

---
**Atenção**: Ao invés de `lst.append(v)` poderiamos fazer
```Python
lst = lst + [v]
```
Se não consideramos questões de aliasing, ambas alternativas resultam equivalentes.
No entanto a alternativa com `+` deve ser evitada, pois a concatenação vai criar uma nova lista. Tal criação requerirá a cópia de todos os elementos da lista original, tornando esta alternativa ineficiente.

---

Além de `append` existem os métodos `extend` e `insert`
* `lst.extends(iteravel)` : adiciona ao final de `lst` todos os elementos de `iteravel`.
* `lst.insert(i, v)`: insere `v` na posição `i`. Para realizar esta inserção, todos os elementos da posição `i` até o fim são deslocados uma posição à direita. Este deslocamento torna este método ineficente se comparado com `append`.



# Listas e strings

Listas e strings são sequências, porém uma lista de caracteres não é o mesmo que um string. Para transformar de string para lista de caracteres podemos usar a função `list`.

````python
list("casa")
````

Conhecemos o método `split()` que transforma um string em palavras.

````python
"Um string com algumas palavras".split()
````

O método `split` aceita um argumento opcional que especifica os caracteres que atuam como separadores. Por exemplo

````python
"1 9.29A.4 3". split('.')
````

Para fazermos o inverso de `split`, dada uma lista de palavras juntar todas em um único string, usamos o método `join`. O método se aplica ao delimitador, passando como argumento a lista: `delimitador.join(lst)`. Por exemplo

````python
' '.join(["Marcos", "Valerio", "Santos"])
````

````python
'-'.join(["Marcos", "Valerio", "Santos"])
````
