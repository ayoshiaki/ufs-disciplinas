# Padrão: armazenar e depois processar

Usamos este padrão quando precisamos processar alguma coleção de dados e, por necessidade ou por conveniência, carregamos a coleção de dados em memória para depois fazer o processamento. Nestes casos, listas se apresentam como uma alternativa comum para o armazenamento.

Algumas situações em que este padrão é pertinente:
 * Queremos melhorar a modularidade do código, pois várias tarefas estão sendo feitas num único laço de iteração complexo.
 * Os dados precisam ser processados várias vezes.
 * Requeremos de algum preprocessamento dos dados para melhorar a eficiência ou facilitar as próximas tarefas. Um exemplo comum é preordenar os dados.
 * Para obter a resposta preciso obrigatoriamente ter lido todos os dados e só depois navegar novamene na lista de dados.
