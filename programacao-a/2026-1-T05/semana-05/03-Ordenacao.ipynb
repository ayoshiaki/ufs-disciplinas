##**2. Método de Ordenação por Seleção**


O primeiro método que vamos aprender é o de **Seleção**. Este método se baseia na ideia de que se **selecionamos** o menor valor da lista e o colocamos na primeira posição da lista, ele já estará na posição definitiva. O processo de seleção pode ser repetido considerando agora a porção da lista sem o primeiro elemento que já está no seu lugar definitivo. Ordenaremos a lista iterando com este processo.

Mais precisamente, neste método dividimos a lista em duas porções:
+ Uma à esquerda já ordenada e com todos os valores no lugar definitivo.
+ E outra à direita, provavelmente ainda não ordenada, da qual iremos pegando repetidamente o seu menor valor e adicionando no fim da porção ordenada.

Inicialmente, a porção ordenada começará vazia enquanto a outra porção será a lista completa. Veja um exemplo

| Passo | Porção esquerda <br> Ordenada | Porção direita | Troca
| :--: | :--- | :--- | :---: |
| 1 | [   |  <font color="red">50</font>, 55, 10, 32, **7** ] | 7 com 50
| 2 | [ 7 ,  | <font color="red">55</font>, **10**, 32, 50 ] | 10 com 55
| 3 | [ 7, 10,  | <font color="red">55</font>, **32**, 50 ] | 32 com 55
| 4 | [ 7, 10, 32,  | <font color="red">55</font>, **50** ] | 50 com 55
| 5 | [ 7, 10, 32, 50,  | 55 ] |


Em cada linha da tabela, o número em negrito é o selecionado para passar à porção da esquerda, ou seja, o menor da porção direita. Já o número em vermelho, sempre o primeiro da porção direita, é aquele que é trocado com o selecionado.

Observe que a porção da lista ordenada vai recebendo **incrementalmente** os valores em ordem crescente e ao final a lista estará ordenada. Já a porção não ordenada vai perdendo os valores e no final ficará vazia.

### **2.1. Implementação do método de ordenação por Seleção em Python**


Para facilitar o trabalho e modularizar o código, foi criada a função `ordenarSelecao` que recebe a lista a ser ordenada como parâmetro.
Ela utiliza a função `indiceDoMenorDesde` que encontra o menor valor da porção da lista a partir de uma dada posição.

Na função `ordenarSelecao` a variável `i` divide as duas porçoes discutidas acima: a não ordenada e a ainda não ordenada. A primeira porção inclui todas as posições menores que `i` enquanto a segunda vai desde `i` até a posição final.
Cada iteração do laço `for` seleciona o menor da porção direita e o coloca na posição `i`.


````python
def ordenarSelecao(lst):
   for i in range(len(lst)-1):
       pos_menor = indiceDoMenorDesde(i, lst)
       lst[i], lst[pos_menor] = lst[pos_menor], lst[i]

def indiceDoMenorDesde(inicio, lst):
   pos_menor = inicio
   for pos in range(pos_menor+1, len(lst)):
     if lst[pos] < lst[pos_menor]:
       pos_menor = pos
   return pos_menor

nums = [50, 10, 55, 32, 7]
#nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenarSelecao(nums)
print('Fim:', nums)
````

##**3. Método de Ordenação por Inserção**


O método de ordenação por Inserção também é bastante simples de entender.
Usamos este métodos quando jogamos baralho e ordenamos as cartas que estão na nossa mão. Se já temos um conjunto de cartas ordenadas na nossa mão, ao receber uma nova carta resta apenas buscar a posição certa para inseri-la.

Este método segue o padrão Cálculo Incremental e também podemos implementá-lo dividindo a lista em duas porções, como na Ordenação por Seleção. Da mesma forma, a porção à esquerda já está ordenada e a outra provavelmente não. Iteramos passando o primeiro elemento da direita para a esquerda inserindo-o de tal forma que a ordem na esquerda seja mantida. No final da iteração a lista completa estará ordenada.

Observe que, diferentemente da Ordenação por Seleção, os elementos na porção ordenada não estarão necessariamente nas posições definitivas. Aqui um exemplo de como funciona o método

| passo | porção ordenada | porção não ordenada |
| :--:  | :---           | :--- |
| 1 | [ 50, |  **10**, 55, 32, 7 ] |
| 2 | [ 10, 50,  |  **55**, 32, 7 ] |
| 3 | [ 10, 50, 55, | **32**, 7 ] |
| 4 | [ 10, 32, 50, 55,   | **7** ]
| 5 | [ 7, 10, 32, 50, 55   |  ] |


A porção esquerda inicial é formada pelo primeiro elemento da lista enquanto que a porção direita é o restante da lista. A cada passo, inserimos o primeiro elemento da direita na esquerda, porém mantendo a ordem na esquerda.

Se usamos uma variávei `i` para delimitar as porções esquerda e direita. como fizemos na ordenação por seleção, o valor inicial de `i` pode ser `1`. Se assumimos que temos uma função `inserir` tal que `inserir(i, lst)` insere o ì-ésimo elemento da lista na porção à esquerda de `i`, mantendo a ordem, a função de ordenação por inserção pode ser definida assim:


````python
def ordenarInsercao(lst):
    for i in range(1, len(lst)):
      inserir(i, lst)
````

A implementação de `inserir` provavelmente exigirá o deslocamento de uma porção da lista uma posição à direita para criar o espaço em que será colocado o elemento `lst[i]` que está sendo inserido. Mais precisamente, se `lst[i]` for igual a $v$, precisamos deslocar uma posição à direita todos os valores à esquerda da posição `i` que são maiores que $v$. No espaço criado por este deslocamento coloca-se $v$.  Assim, temos a seguinte implementação de `inserir`

````python
def inserir(i, lst) :
        valor = lst[i]
        while i >= 0 and lst[i] > valor:
            lst[i + 1] = lst[i]
            i -= 1
        lst[i + 1] = valor


nums = [50, 10, 55, 32, 7]
#nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenarInsercao(nums)
print('Fim:', nums)
````

Na linha 6, salvamos `lst[i]` em `valor`, pois o deslocamento que será feito poderá sobrepor um novo valor na posição `i`.
As linhas 7-9 deslocam uma posição à direita todos os valores maiores que `valor`.
Finalmente, na linha 10, colocamos `valor` no espaço criado pelo deslocamento.

##**4. Método de Ordenação Bolha (Bubble Sort)**


Dada uma lista `lst` e dois índices `i` e `j` tal que `i < j`, dizemos que o par de elementos `lst[i]` e `lst[j]` estão invertidos se `lst[i] > lst[j]`.  Intuitivamente, analisando isoladamento ambos elementos, `lst[i]` e `lst[j]` estão fora da ordem e uma troca entre eles deixará a `lista` um pouco mais próxima da ordenação. De fato, se continuamos reiteradamente trocando pares de elementos invertidos, em algum momento a lista estará ordenada.

Seguindo este pensamento, vamos focar apenas em trocas de pares adjacentes que estejam invertidos. Podemos fazer uma varredura na lista analisando pares adjacentes e trocando-os caso estejam invertidos. Assim, iniciamos analisando `lst[0]` e `lst[1]`, depois `lst[1]` e `lst[2]`, depois `lst[2]` e `lst[3]`, e assim succesivamente. Em Python, seria assim

```
for i in range(0, len(lst)-1) :
  if lst[i] > lst[i+1] :
    lst[i], lst[i+1] = lst[i+1], lst[i]
```

Chamamos este processo de borbulhar.

O borbulhamento puxa o maior elemento para o final da lista. Ou seja, depois de fazer um borbulhamento o maior elemento da lista já estará na posição certa. O que acontecerá se fazemos um segundo borbulhamento? Trabalhe um exemplo se necessário.
<br>...<br>...<br>...<br>...<br><br><br><br><br><br><br>

O segundo borbulhamento deixará o segundo maior elemento da lista na posição certa. Claramente, se fizermos `n-1` borbulhamentos, onde `n` é o tamanho da lista, a lista ficará completamente ordenada.



````python
def ordenaBolha1(lst) :
  for _ in range(len(lst)-1):
    borbulhar(lst)

def borbulhar(lst) :
  for i in range(0, len(lst)-1) :
    if lst[i] > lst[i+1] :
      lst[i], lst[i+1] = lst[i+1], lst[i]


nums = [50, 10, 55, 32, 7]
nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenaBolha1(nums)
print('Fim:', nums)
````

Vejamos como podemos melhorar este algoritmo. Depois do primeiro borbulhamente, o maior da lista fica no final da lista, ou seja no lugar definitivo. Logo, o segundo borbulhamento não precisa ser feito até a última posição da lista, mas somente até a penúltima. Da mesma forma, o terceiro borbulhamento só precisará ser feito até a antepenúltima posição, e assim por diante. Podemos então adequar a função `borbulhar` para aceitar também como argumento a posição até a qual será  feito o borbulhamento, e com isso uma versão mais eficiente da ordenação é


````python
def ordenaBolha2(lst) :
  for i in range(len(lst)-1, 0, -1):
    borbulharAte(lst, i)

def borbulharAte(lst, k) :
  for i in range(0, k) :
    if lst[i] > lst[i+1] :
      lst[i], lst[i+1] = lst[i+1], lst[i]

nums = [50, 10, 55, 32, 7]
nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenaBolha2(nums)
print('Fim:', nums)
````

Se um borbulhamento não fez nenhuma troca, significa que a lista já está ordenada. Logo, quando borbulharmos e não detectarmos troca, podemos parar a ordenação, pois já está feita.
Adaptamos `borbulharAte2` para que retorne um boolean, `True` se houve alguma troca e `False` caso contrário.
Também adaptamos `ordenarBolha2` para parar caso o borbulhamento não tenha feito trocas.

````python
def ordenaBolha3(lst) :
  for i in range(len(lst)-1, 0, -1) :
    if not borbulharAte3(lst, i) :
      return

def borbulharAte3(lst, k) :
  houveTroca = False
  for i in range(0, k) :
    if lst[i] > lst[i+1] :
      lst[i], lst[i+1] = lst[i+1], lst[i]
      houveTroca = True
  return houveTroca

nums = [50, 10, 55, 32, 7]
nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenaBolha3(nums)
print('Fim:', nums)
````

Uma definição alternativa para `ordenaBolha3` que usa `while` ao invés de `for` é

````python
def ordenaBolha3(lst) :
  i = len(lst)-1
  while borbulharAte3(lst, i) :
    i -= 1

nums = [50, 10, 55, 32, 7]
nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenaBolha3(nums)
print('Fim:', nums)
````
