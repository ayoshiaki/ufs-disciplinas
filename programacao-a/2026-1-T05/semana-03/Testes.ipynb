# Testes

Sob a perspectiva da Engenharia de Software, deseja-se que os programas tenham alguns atributos de qualidade, tais como: código legível, modularidade, facilidade de manutenção e bom desempenho. Talvez o mais relevante destes atributos seja a **corretude** do programa. Afinal, quem usa um programa quer que ele entregue as respostas certas.

Uma forma de garantir essa corretude é através de testes cuidadosamente projetados. Os testes objetivam identificar erros e defeitos antes de colocar o sistema para funcionar em uma situação real, o que melhora a qualidade do produto e reduz o risco de falhas.

Existem diversos tipos de testes. Aqui, focaremos em **testes unitários**, que validam partes isoladas do código (como as funções de um programa). O objetivo é garantir que, isoladamente, cada função se comporte como o esperado.

No contexto de um programa pequeno, podemos testar as funções manualmente.

Um caso de teste verifica a saída da função para uma entrada específica. Mas como podemos escolher um bom conjunto de casos de teste, em vez de simplesmente escolher alguns valores aleatoriamente e sem critério?

Queremos que o conjunto de casos de teste seja abrangente. Uma estratégia eficaz é particionar os casos em grupos onde esperamos que a função se comporte de forma similar, definindo então um ou alguns casos de teste para cada grupo.

Particularmente, precisamos prestar muita atenção a casos especiais que ocorrem nas fronteiras (ou limites) destes grupos. Por exemplo, se os grupos se dividem entre números positivos e negativos, precisaremos ter atenção redobrada com o número zero.

---
**Máxima de programação**: O diabo está nos detalhes. É muito comum que *bugs* (erros) ocorram devido a descuidos com casos especiais, nas fronteiras dos grupos.

# Estudo de caso: Maior de três - Testes unitários manuais

Vejamos um exemplo concreto. Considere que queremos testar uma função `maiorDeTres`, que calcula o maior dentre três números. Quais poderiam ser os grupos de teste?

Não há uma única resposta correta. Podemos iniciar pensando no que é relevante e no que é irrelevante. Neste exemplo, é razoável assumir que a grandeza do número ou o seu sinal não são relevantes, pois o que determina o resultado é apenas a relação de ordem entre eles.

Podemos então propor a seguinte divisão:
+ Todos os três valores são diferentes.
+ Todos os três valores são iguais.
+ Dois valores são iguais e o terceiro é diferente:
  + Dois valores iguais ao máximo e o terceiro diferente.
  + Um valor igual ao máximo e os outros dois iguais entre si.

Vamos pegar um caso de cada grupo e testar a definição a seguir. Os casos para `a`, `b` e `c` serão:
```python
6 4 1
6 6 6
2 6 6
2 2 6
```

````python
def maiorDeTres(a, b, c) :
  if a > b and a > c :
    return a
  elif b > a and b > c :
    return b
  else :
    return c
````

````python
maiorDeTres(6, 4, 1)
````

````python
maiorDeTres(6, 6, 6)
````

````python
maiorDeTres(2, 6, 6)
````

````python
maiorDeTres(2, 2, 6)
````

Todos os testes deram certo, ou seja, a função **passou nos testes** com sucesso. Então, podemos concluir que o programa está totalmente correto?

O que acontece se testarmos com os valores de `a`, `b` e `c` iguais a 6, 6 e 2, respectivamente?

````python
maiorDeTres(6, 6, 2)
````

Testes não garantem que o programa está 100% livre de falhas, mas nos dão muito mais confiança sobre a corretude da função.

Como poderíamos ter detectado este erro? Para este problema existe uma maneira melhor de separar os grupos de teste baseando-se nas possíveis permutações (ordens) dos três valores:
+ Todos os valores diferentes: seis ordens possíveis.
+ Todos os valores iguais: uma ordem possível.
+ Dois valores iguais e o terceiro diferente: três ordens possíveis.

Uma das permutações do último grupo incluiria exatamente o caso de teste 6, 6 e 2, que evidenciou a falha.

# Testes Caixa Preta e Caixa Branca

As estratégias que discutimos acima para escolher os casos de teste não se basearam na forma como o código foi escrito, mas sim na **especificação** da função (ou seja, no seu propósito geral, que é calcular o maior entre três números). Testes criados apenas com base nas entradas e saídas esperadas são denominados **testes caixa preta**.

Quando o código-fonte da função é conhecido por quem está testando, além da divisão em grupos de casos que fazemos para testes caixa preta, podemos também usar a estrutura desse código para a escolha dos casos. Nessa situação, dizemos que estamos fazendo **testes caixa branca**.

No exemplo da função `maiorDeTres`, percebemos que existe um condicional sequencial com três ramos (`if`, `elif`, `else`). Para um bom teste caixa branca, faz-se necessário definir entradas que *exercitem* (passem por) cada um desses ramos.

Além disso, observando as condições lógicas `a > b and a > c` e `b > a and b > c`, notamos que o caso de teste (6, 6 e 2) coloca os dois primeiros números exatamente na **fronteira** da condição de maioridade (`a > b` e `b > a`). Isso reforça como a escolha de valores nas fronteiras das regras de negócio é crucial para encontrar falhas.

# Testes baseados em propriedades

Nos testes convencionais, nós definimos os casos de teste um a um. Uma alternativa mais poderosa é usar ferramentas que gerem automaticamente milhares de casos de teste variados.

Nesse paradigma, nós apenas especificamos quais são as **propriedades universais** que a nossa função deve sempre satisfazer, e a ferramenta testa isso gerando valores aleatórios. Chamamos este paradigma de **testes baseados em propriedades**.

No Python, temos a biblioteca Hypothesis, que nos permite aplicar essa técnica facilmente. Para instalar a biblioteca no seu ambiente, execute o comando a seguir:

````python
!pip install hypothesis
````

Continuando com o exemplo da função `maiorDeTres`, quais propriedades poderiam ser essenciais de se testar?

Certamente, uma propriedade inquestionável é que: *a função deve retornar um número que seja maior ou igual a todos os três parâmetros recebidos*.

Usando a biblioteca Hypothesis, podemos automatizar essa verificação da seguinte forma:

````python
from hypothesis import given, strategies as st

@given(st.integers(), st.integers(), st.integers())
def test_maiorDeTres_maior(a, b, c) :
  m = maiorDeTres(a, b, c)
  assert m >= a and m >= b and m >= c
````

Neste código, a propriedade é validada dentro da função `test_maiorDeTres_maior`. Na última linha, utilizamos o comando `assert` do Python. Ele afirma matematicamente que a variável `m`, o resultado retornado por `maiorDeTres`, deve ser maior ou igual a `a`, `b` e `c`. Se a afirmação for falsa, o Python emitirá um erro.

O trecho `@given` logo acima da definição (na linha 3) é um "decorador" do Hypothesis. Ele define as estratégias de geração de dados. Neste caso, instruímos a ferramenta a fornecer três números inteiros (`st.integers()`) sempre que a função for chamada.

Para iniciar a verificação, basta invocar a função de teste sem passar nenhum argumento. Por padrão, o Hypothesis executará o código 100 vezes, gerando 100 conjuntos diferentes de três números inteiros automaticamente.

````python
test_maiorDeTres_maior()
````

Se você ainda não corrigiu a função `maiorDeTres` anteriormente, observará que ocorreu um erro (`AssertionError`) na linha do `assert`. Isso é a ferramenta provando que a sua função falhou em satisfazer a propriedade para algum dos conjuntos de números testados.

Para descobrirmos rapidamente quais foram os valores gerados que "quebraram" a função, podemos adicionar um comando `print` para exibir as entradas fornecidas antes do erro acontecer:

````python
from hypothesis import given, strategies as st

@given(st.integers(), st.integers(), st.integers())
def test_maiorDeTres_maior(a, b, c) :
  m = maiorDeTres(a, b, c)
  print(f"maiorDeTres({a}, {b}, {c}) = {m}")  # Adicionado para apresentar os casos
  assert m >= a and m >= b and m >= c

test_maiorDeTres_maior()
````

A última saída impressa pela ferramenta evidenciará claramente onde ocorreu a falha lógica (por exemplo: `maiorDeTres(0, 0, -1) = -1`, mostrando que a função ignorou que 1 era maior que 0).
