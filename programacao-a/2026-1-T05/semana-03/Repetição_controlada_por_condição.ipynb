#Estruturas de repetição (loops)
* Repetições são algumas das estruturas de controle mais usadas em programação;
* Essas estruturas possibilitam que uma ação seja seja reproduzida repetidamente;
* Nas linguagens de programação em geral há estruturas de repetição controladas por condição e controladas por contador;
* As duas principais estruturas de repetição do Python são:
  * Repetição controlada por condição - **while**
  * Repetição controlada por contador - **for**

#Estrutura de repetição **while**
* É a estrutura de repetição mais geral;
* Usada para executar repetidamente um bloco de instruções;
* Nessa estrutura a execução do bloco de comandos é controlada pelo valor lógico da condição de teste;
* Se a condição é verdadeira, o bloco de instruções é executado, caso ela seja falsa o bloco não é executado;
* Antes de cada execução do bloco a condição é testada para decidir se o bloco será executado, dessa forma após cada execução do bloco a condição é novamente testada, em caso afirmativo, o bloco é executado novamente;
* O fluxograma a seguir representa o fluxo de execução dessa estrutura.

###Exemplo de uso de while em Python
Para exemplificar o uso dessa estrutura vejamos um programa que leia um número inteiro positivo e exiba todos os números de 1 até ele.

```python
num = int(input('Digite um número inteiro positivo: '))
cont = 1
while cont <= num:
  print(cont)
  cont += 1
print('Fim do programa.')
```

Agora vejamos como seria o código do seguinte programa. Leia vários números, exibindo a metade de cada um deles. O programa deve ser encerrado ao se digitar um número negativo.

```python
num = float(input('Digite um número <para encerrar digite número negativo>:'))
while num >=0:
  print('Metade de ',num, ' = ',num/2)
  num = float(input('Digite outro número: '))
```

##Contadores e acumuladores
Contadores e acumuladores são variáveis usadas para contar/acumular valores. Para usá-las é interessante observar 3 momentos específicos:
* Inicialização - Deve ser feita antes do laço. O valor inicial de um contador/acumulador deve ser o elemento neutro da operação de incremento que será utilizada. Ou seja, caso a variável tenha seja um contador, a operação será uma soma, caso a variável seja um acumulador, se a operação de incremento for soma, a variável deve ser inicializada com 0 (zero), se for multiplicação, deve-se inicializar a operação com 1.
* Incremento - Deve ser feito dentro do laço;
* Uso - Na maioria dos casos será feito após o encerramento do laço, que será quando a contagem ou o valor acumulado estará completo.
