##Laços infinitos e **break**
Algumas vezes um programador inexperiente criará um laço infinito (loop infinito) disfuncional ao cometer erros, como esquecer de mudar o valor da variável de teste ou mesmo o erro de mudar esse valor de forma que a condição de encerramento do laço nunca seja alcançada.  
Em algumas situações, entretanto, pode ser interessante criar um laço infinito funcional desde que dentro do laço ocorra uma condição que leve ao encerramento do mesmo. Para isso usa-se a instrução **break** que faz com que o programa saia de dentro de um laço.
Vejamos o exemplo a seguir, no qual vemos um código que exibe o dobro de números até que o usuário digite um número negativo, que também deve ter seu dobro exibido.

```python
#Exemplo de loop infinito funcional
while True:
  num = int(input('Digite um número inteiro: '))
  print(f'O dobro de {num} é {num*2}')
  if num < 0:
    break
print('Fim do programa...')

#Para fomentar seu aprendizado, faça mudanças no programa e verfique como ele é executado
```

Observe que no exemplo anterior em lugar do **while** ter um teste ele tem apenas True que é o valor retornado por um teste verdadeiro. Vemos que nele a condição de saída é alcançada ao ser executada uma instrução **if**.

Em Python, esse tipo de construção é relativamente comum, pois permite que a condição de saída seja checada em qualquer lugar do bloco de instruções e não apenas antes do início do bloco. Outras linguagens de programação, como por exemplo **C, C++ e Java** possuem uma instrução, chamada **do... while**, em que o bloco da repetição é executado e o teste é feito após essa execução.

## Finalizando laços com continue
Eventualmente, em um bloco de uma repetição há a necessidade de avançar para uma próxima iteração sem executar o restante das instruções da iteração atual. Nesse caso deve-se usar a instrução **continue**. Essa instrução faz com que a execução do programa avance para a próxima iteração sem executar o restante do bloco de instruções.

Uma utilização dessa instrução pode ser vista no exemplo a seguir.

```python
#Programa exemplo que lê vários números exibindo a raiz quadrada dos números positivos.
# Para encerrar deve ser digitado 0 (zero).

while True:
  num = int(input('Digite um número inteiro <para sair digite 0>: '))
  if num == 0:
    break
  elif num < 0:
    continue
  print(f'A raiz quadrada de {num} é {num ** (1/2)}')
print('Encerrando programa...')
```

#Iteração sobre elementos de uma lista

Imagine que você quer exibir individualmente uma mensagem de boas vindas às pessoas cadastrada na lista ListaNomes. Como fazer?

Uma possibilidade é usar uma estrutura de repetição que será executada para cada elemento existente em uma lista. Esse é um dos possíveis usos da estrutura de iteração definida **for** que pode ser usado como no exemplo a seguir:


```python
listaNomes = ['Ana', 'Bia', 'Gil', 'Ze', 'Joana', 'João', 'Eliane']
for nome in listaNomes:
  print(f'Olá {nome}, seja bem vindo(a) ao grupo!')
  print('Sua presença nos alegra!!!\n')
print('Fim do programa...')


#Para fomentar seu aprendizado, faça mudanças no programa e verfique como ele é executado
```

Algumas observações importantes devem ser feitas a respeito do exemplo dado:
* A estrutura **for** gera uma quantidade limitada de iterações que corresponde a exata quantidade de elementos da lista, ou seja, pode-se entender que o bloco de instruções é repetido enquanto há itens na lista;
* Ao longo da repetição, a variável **nome** assume cada um dos valores da lista;
* Tal qual na repetição com **while**, observe que o início e fim do bloco de instruções a ser repetido com for também é definido pela indentação (que em Python é obrigatória), ou seja, pelo fato das instruções estarem afastadas da margem.


Vejamos outro exemplo de uso do for para exibir o dobro de cada elemento da lista de números:

```python
listaNumeros = [21,432,6,712, 9]
for num in listaNumeros:
  print(f'O dobro de {num} é {num*2}')
```

Observe que dessa vez a variável num assumiu o valor de cada item da variável listaNumeros, permitindo que ela fosse usada para calcular e exibir o dobro de cada um desses números.

Agora vamos pensar em outra possível situação. A fórmula de conversão de graus Fahrenheit para Celsius é C = (F-32) * 5/9. Dessa forma a resposta a uma questão que peça a criação de uma função que receba uma temperatura em Fahrenheit e retorne essa temperatura em Celcius seria a seguinte.

```python
#Função de conversão
def tempFParaC(temp):
  return (temp - 32) * 5/9

#Programa para testar a função
tFahr = float(input('Digite uma temperatura em Fahrenheit:'))
tCelsius = tempFParaC(tFahr)
print(f'{tFahr:.1f} ºF = {tCelsius:.1f} ºC')
```

##Desafio...
...e se quiséssemos usar essa função para escrever um código que mostrasse uma tabela de conversão de 0ºF a 100ºF, com a temperatura aumentando de 2 em 2 graus. Como poderíamos fazer?

Uma possibilidade seria usar a estrutura de repetição while, vista anteriormente. Conforme pode ser visto no código a seguir:

```python
#Função de conversão
def tempFParaC(temp):
  return (temp - 32) * 5/9

#Gerando tabela de conversão usando while
print('Tabela de conversão de Fahrenheit para Celsius\n')
cont = 0
while cont <= 100:
  print(f'{cont:4.1f} ºF = {tempFParaC(cont):4.1f} ºC')
  cont += 2;
```

###Sim, vê-se que conseguimos escrever o programa usando o while. Mas será que usar while é a forma mais adequada de se implementar esse tipo de código?

Há uma estrutura de repetição mais adequda a situações em que **sabemos quantas vezes um bloco de instruções deve ser repetido, antes** de iniciar a repetição.

Esse padrão de programação se encaixa perfeitamente com a estrutura **for** vista anteriormente, entretanto para usá-la nessa situação é interessante que antes aprendamos a gerar uma faixa de valores.

#Função range( )
A função **range()** pode ser usada para gerar automaticamente uma faixa de valores. Na verdade a tradução literal de range seria "faixa".
Há três variações do uso dessa instrução:
* Recebendo como parâmetro, um único número inteiro que representa a quantidade de números inteiros positivos gerados a partir do 0 (zero);
* Recebendo como parâmetro 2 valores inteiros que representam respectivamente valor inicial e valor final da faixa a ser gerada;
* Recebendo como parâmetros 3 valores inteiros que representam respectivamente valor inicial e valor final da faixa a ser gerada e incremendo a ser usado na construção dos valores da faixa, cujos valores formarão uma progressão aritmética.




```python
# Exemplo de uso de range com 1 parâmetro que indica quantidade de elementos da faixa gerada
# Observe que no exemplo a seguir x assumirá cada um dos valores
# da faixa gerada pela função range()

print('Exemplo do uso de range com 1 parâmetro')
for x in range(5):
  print(x)
print('Observe que o bloco de instruções foi repetido 5 vezes')
# Importante perceber que como os valores gerados começam do 0, o último valor é
# igual a quantidade de valores - 1.

```

```python
# Exemplo do uso de range com 2 parâmetros. Observe que o primeiro parâmetro indica o valor inicial da faixa
# e o segundo parâmetro indica o valor final
print('Exemplo do uso de range com 2 parâmetros')
for x in range(5, 10):
  print(x)
print('Observe que o bloco de instruções foi repetido 5 vezes')
```

No exenplo anterior, observe que a variável x assume os valores de 5 a 9 e não de 5 a 10 (como seria de se esperar). Isso ocorre para manter a coerência com a execução da versão de range( ) que recebe apenas 1 parâmetro.

Vejamos a seguir um exemplo de uso da versão de range que recebe 3 parâmetros:

```python
# Exemplo do uso de range com 3 parâmetros. Observe que o primeiro parâmetro indica o valor inicial da faixa
# o segundo parâmetro indica o valor final e o terceiro parâmetro indica o fator de incremento. Esse fator de incremento,
# inclusive, pode ser negativo.
print('Exemplo do uso de range com 3 parâmetros')
for x in range(5, 20, 2):
  print(x)
print('Fim do programa...')
```

Voltando ao desafio proposto. Escreva um programa que mostre uma tabela de conversão de 0ºF a 100ºF, com a temperatura aumentando de 2 em 2 graus.

```python
def tempFParaC(temp):
  return (temp - 32) * 5/9

print('Tabela de conversão')
for cont in range(0,101,2):
  aux = tempFParaC(cont)
  print(f'{cont:5.1f} ºF = {aux:4.1f} ºC')
```

Observe que para a conversão de 100ºF ser feita foi colocado como valor máximo do range 101 e não 100.
