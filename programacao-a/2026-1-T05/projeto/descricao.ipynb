# Projeto de Programação A - 2026-1

Neste projeto será elaborado um programa para construir desenhos no estilo das apps Google Drawings e Libre Office Draws.

O projeto deverá utilizar o Tkinter, a biblioteca padrão de Python para construção de GUIs. É exigido **Python 3.10 ou superior** (o projeto usa recursos de tipagem como `list[int]` e `X | None`). No Linux, o Tkinter pode exigir o pacote `python3-tk` (confira com `python -m tkinter`).

Será disponibilizado uma implementação inicial simples e incompleta, estruturada imperativamente. Esse esqueleto já traz o básico para desenhar com o mouse no `Canvas` (ex.: traçar linhas/mão livre) e serve apenas de ponto de partida; cabe às equipes evoluí-lo ao longo das entregas.

Ao final do projeto, o programa deverá contemplar (no mínimo) os seguintes tipos de figura: **Linha**, **Mão livre** (rabisco), **Retângulo**, **Oval**, **Polígono** e **Figura composta** (grupo). Cada entrega abaixo indica quando cada tipo e funcionalidade deve estar pronto.

O projeto deverá ser desenvolvido em equipes de dois a quatro alunos (preferencialmente quatro). Trabalhos individuais terão penalidade de 50% da nota. 

As equipes deverão:
  - Refatorar o projeto para uma estrutura de programa Orientada a Objetos.
  - Adicionar funcionalidades.
  - Aplicar alguns padrões arquiteturais e de projeto.
  - Desenvolver o sistema utilizando as ferramentas Git e Github para trabalho colaborativo.

## Objetivos de aprendizagem

Ao final do projeto, espera-se que o aluno seja capaz de:
  - Modelar um domínio com uma hierarquia de classes e polimorfismo (Orientação a Objetos).
  - Aplicar a arquitetura **MVC** e os padrões de projeto **State**, **Composite** e **Command**.
  - Persistir e recuperar o estado da aplicação (serialização / Salvar e Abrir).
  - Escrever **testes unitários** para a camada de modelo (ver Entrega 4).
  - Trabalhar de forma colaborativa com **Git e Github** (branches, commits, tags, divisão de tarefas).

## Regras de entrega e avaliação

As entregas devem ser feitas criando **tags de versionamento** do projeto no Git/Github (ver a tag indicada em cada entrega).

  + **Commits individuais:** ao longo de *todas* as entregas, cada integrante deve registrar seus próprios commits no Github (não vale um único integrante commitando o trabalho de todos). O histórico será conferido com `git log --author`.
  + **Registro de divisão de tarefas:** cada equipe deve manter um arquivo (ex.: `CONTRIBUTING.md`) indicando quem fez o quê em cada entrega.
  + **Checkpoints de demonstração:** a cada entrega (ou a cada duas), haverá uma breve arguição presencial em que *cada* integrante apresenta e explica a parte que desenvolveu.
  + **Repositório:** o repositório deve conter `README.md` com instruções de execução, `.gitignore` adequado e a organização de pastas recomendada (ver Entrega 3).
  + **Atrasos** nas entregas serão penalizados: cada atraso irá tirar 1 ponto.


# Entregas

> As datas abaixo seguem os dias de aula (terças e quintas). Atenção: **terça 23/06 é feriado** (sem aula). Comece cedo e distribua o trabalho.

## Entrega 1 (18/06, quinta - tag: `imperativa.1`)
*Objetivo:* familiarizar-se com o Tkinter e com o fluxo Git/Github da equipe.

Leitura do colab sobre Git e Github e criação do projeto no Github.

Partindo do esqueleto fornecido (que já permite traçar linhas/mão livre), adicionar funcionalidades que permitam:
  + Desenhar retângulos.
  + Desenhar ovais.
  + Que os desenhos individuais possam ter bordas com cor.
  + Que os desenhos individuais tenham cor de preenchimento.

Nesta entrega ainda é aceitável (e esperado) usar código imperativo, com condicionais do tipo `if ferramenta == "retangulo"`. Esse código condicional será eliminado mais adiante (Entrega 4, padrão **State**).

## Entrega 2 (25/06, quinta - tag: `OO.1`)
*Objetivo:* substituir o código imperativo por uma modelagem Orientada a Objetos.

Refatorar o sistema para seguir uma abordagem OO.
  + Definir uma hierarquia de classes (`Figuras`) para contemplar os diversos tipos de desenhos: linha, mão livre, retângulo, oval, etc.
  + Adequar o programa para utilizar a hierarquia de `Figuras`.
  + Adicionar desenhos de polígonos.
  + Separar o código em módulos (Classes `Figuras` em um arquivo separado, por exemplo).

## Entrega 3 (02/07, quinta - tag: `OO.MVC.1`)
*Objetivo:* separar responsabilidades aplicando a arquitetura MVC.

Refatorar para utilizar o padrão MVC.
  + Definir as classes do modelo (Figuras, Desenho, ...).
  + Definir classe ou classes para a visão.
  + Definir classes para os controladores.

Recomendação: usar a seguinte estrutura de arquivos e pastas

    nome-do-projeto/
    ├── .git/                     # Controlado pelo Git
    ├── .gitignore                # Arquivos e pastas ignorados pelo Git
    ├── README.md                 # O que é o projeto e como executar
    ├── tests/                    # Testes automatizados do modelo (ver Entrega 4)
    ├── src/                      # Código-fonte principal do projeto
    │   └── nome_do_projeto/      # Pacote python
    │       ├── main.py           # O programa principal
    │       ├── modelo            # Classes do Model
    │       │   └── ...           #
    │       ├── visao             # Classes da View
    │       │   └── ...
    │       └── controlador       # Classes Controllers
    │           └── ...

## Entrega 4 (07/07, terça - tag: `OO.State.1`)
*Objetivo:* eliminar condicionais de tipo com o padrão **State**, persistir o desenho e iniciar os testes.

  + Utilizar o padrão **State** para evitar código condicional nos métodos do(s) controlador(es) (a escolha da ferramenta — linha, retângulo, etc. — vira um estado).
  + Adicionar as funcionalidades **Salvar** e **Abrir**.
    - Definir um formato de serialização (sugestão: **JSON**).
    - A responsabilidade de serializar/desserializar deve ficar nas próprias `Figuras` (cada figura sabe se converter para/de um dicionário).
  + Escrever **testes unitários** (`pytest` ou `unittest`) para a camada de modelo, cobrindo:
    - geometria das figuras (criação, mover, teste de clique/seleção);
    - serialização: round-trip *figura → dicionário → figura* e o ciclo Salvar/Abrir.
  + Os testes devem ficar em `tests/` e crescer junto com o modelo nas entregas seguintes (Composite na Entrega 6, comandos na Entrega 7).

## Entrega 5 (09/07, quinta - tag: `OO.Selecao.1`)
*Objetivo:* manipular figuras já desenhadas.

Permitir a seleção de figuras e oferecer as seguintes funcionalidades (que atuam sobre as figuras selecionadas):
 + Mover figura.
 + Apagar figura.
 + Copiar/Colar (CTRL-C/CTRL-V).
 + Mover para frente e para atrás.
 + Mudar cores.
 + Seleção múltipla (de várias figuras).
    + Mover várias figuras, apagar várias figuras, ...

As tarefas desta entrega devem ser divididas entre os integrantes do grupo. Cada aluno deverá registrar seus commits individuais no Github.

## Entrega 6 (14/07, terça - tag: `OO.Compostas.1`)
*Objetivo:* tratar uma figura composta como uma figura qualquer, usando o padrão **Composite**.

Permitir criar figuras compostas (agrupar várias figuras em uma só) e também desfazer o agrupamento (desagrupar). Aplicar o padrão de projeto **Composite**: uma figura composta deve respeitar o mesmo contrato (`desenha`, mover, etc.) das figuras simples.

## Entrega 7 (16/07, quinta - tag: `OO.UndoRedo.1`)
*Objetivo:* desfazer e refazer ações com o padrão **Command**.

Implementar **Undo** e **Redo** aplicando o padrão de projeto **Command**: cada ação que altera o desenho (criar, mover, apagar, mudar cor, agrupar, ...) vira um objeto comando com `executar()` e `desfazer()`, mantidos em pilhas de undo/redo.
