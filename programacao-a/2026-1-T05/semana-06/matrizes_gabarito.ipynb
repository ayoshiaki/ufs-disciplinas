## §1 Matriz como lista de listas

*Tempo estimado: ~15 min*

Em Python, uma matriz $m \times n$ é uma **lista de $m$ listas**, cada uma com $n$ elementos. Acessamos um elemento com `M[linha][coluna]`.

- `len(M)` retorna o número de **linhas**.
- `len(M[0])` retorna o número de **colunas** (assumindo que todas as linhas têm o mesmo tamanho).


````python
M = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

print('Matriz inteira:', M)
print('Linha 0:', M[0])
print('M[1][2] =', M[1][2])
print('Dimensões:', len(M), 'x', len(M[0]))
````

## §2 Criação de matrizes

*Tempo estimado: ~20 min*

Para criar uma matriz $m \times n$ de zeros usamos uma **list comprehension**:

```python
M = [[0] * n for _ in range(m)]
```

**Atenção à armadilha** `[[0] * n] * m`: isso cria $m$ referências para **a mesma lista**, não $m$ listas independentes. Modificar uma linha modifica todas!


````python
# Modo correto: m listas independentes
M = [[0] * 3 for _ in range(2)]
M[0][0] = 1
print('Correto:', M)

# Modo errado: m referências para a mesma lista
N = [[0] * 3] * 2
N[0][0] = 1
print('Errado: ', N)
````

> **Exercício 2.1 (fácil)** — Escreva `zeros(m, n)` que retorna uma matriz $m \times n$ de zeros. Teste com `zeros(2, 3)`.

````python
def zeros(m, n):
    return [[0] * n for _ in range(m)]

print(zeros(2, 3))
````

> **Exercício 2.2 (fácil)** — Escreva `identidade(n)` que retorna a matriz identidade $n \times n$ (1 na diagonal, 0 fora). Teste com `identidade(4)`.

````python
def identidade(n):
    I = [[0] * n for _ in range(n)]
    for i in range(n):
        I[i][i] = 1
    return I

for linha in identidade(4):
    print(linha)
````

> **Exercício 2.4 (médio)** — Escreva `tabela_multiplicacao(n)` que retorna uma matriz $n \times n$ com `M[i][j] = (i+1) * (j+1)`. Imprima `tabela_multiplicacao(5)`.

````python
def tabela_multiplicacao(n):
    M = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            M[i][j] = (i + 1) * (j + 1)
    return M

for linha in tabela_multiplicacao(5):
    print(linha)
````

## §3 Percorrer matrizes

*Tempo estimado: ~15 min*

Problemas com matrizes tipicamente exigem **dois laços aninhados**: um para linhas, outro para colunas.

```python
for i in range(len(M)):
    for j in range(len(M[0])):
        # use M[i][j]
```

Quando só precisamos dos valores (sem os índices), o laço externo `for linha in M:` é mais limpo.

````python
M = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]

# Estilo 1: percorre os valores
for linha in M:
    for x in linha:
        print(x, end=' ')
    print()

print()

# Estilo 2: percorre os índices
for i in range(len(M)):
    for j in range(len(M[0])):
        print(f'M[{i}][{j}]={M[i][j]}', end='  ')
    print()
````

> **Exercício 3.1 (fácil)** — Escreva `soma_total(M)` que retorna a soma de todos os elementos da matriz. Teste com `[[1,2,3],[4,5,6]]` (deve retornar 21).

````python
def soma_total(M):
    total = 0
    for linha in M:
        for x in linha:
            total += x
    return total

print(soma_total([[1, 2, 3], [4, 5, 6]]))  # 21
````

> **Exercício 3.2 (médio)** — Escreva `soma_diagonal(M)` que retorna a soma da diagonal principal de uma matriz **quadrada**. Teste com a identidade 4×4 (deve retornar 4).

````python
def soma_diagonal(M):
    total = 0
    for i in range(len(M)):
        total += M[i][i]
    return total

I4 = [[1 if i == j else 0 for j in range(4)] for i in range(4)]
print(soma_diagonal(I4))  # 4
````

> **Exercício 3.3 (médio)** — Escreva `imprimir_bonita(M)` que imprime a matriz com cada elemento ocupando 4 caracteres alinhados à direita (use `f'{x:4d}'`). Teste com a tabela de multiplicação 5×5 da seção anterior.

## §4 Operações clássicas

*Tempo estimado: ~20 min*

Operações típicas com matrizes seguem o padrão: **criar matriz resultado** do tamanho certo e **preencher** com laço duplo.

Este é o último passo antes de NumPy: depois desta seção, vamos ver como a biblioteca NumPy faz tudo isto de forma muito mais concisa e rápida.

> **Exercício 4.1 (fácil)** — Escreva `soma_matrizes(A, B)` que retorna $C = A + B$, assumindo que $A$ e $B$ têm a mesma dimensão. Teste com $A = [[1,2],[3,4]]$ e $B = [[5,6],[7,8]]$ (deve dar `[[6, 8], [10, 12]]`).

````python
def soma_matrizes(A, B):
    m, n = len(A), len(A[0])
    C = [[0] * n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            C[i][j] = A[i][j] + B[i][j]
    return C

print(soma_matrizes([[1, 2], [3, 4]], [[5, 6], [7, 8]]))
````

> **Exercício 4.2 (médio)** — Escreva `transposta(M)`: a transposta de uma matriz $m \times n$ é uma matriz $n \times m$ onde $T[j][i] = M[i][j]$. Teste com $[[1,2,3],[4,5,6]]$ (deve dar `[[1, 4], [2, 5], [3, 6]]`).

````python
def transposta(M):
    m, n = len(M), len(M[0])
    T = [[0] * m for _ in range(n)]
    for i in range(m):
        for j in range(n):
            T[j][i] = M[i][j]
    return T

print(transposta([[1, 2, 3], [4, 5, 6]]))
````

> **Exercício 4.3 (médio)** — Escreva `posicao_do_maximo(M)` que retorna a tupla `(i, j)` com a posição do maior elemento. Em caso de empate, retorne a primeira ocorrência (linha por linha). Teste com $[[3,1,4],[1,5,9],[2,6,5]]$ (deve dar `(1, 2)`).

````python
def posicao_do_maximo(M):
    i_max, j_max = 0, 0
    for i in range(len(M)):
        for j in range(len(M[0])):
            if M[i][j] > M[i_max][j_max]:
                i_max, j_max = i, j
    return (i_max, j_max)

print(posicao_do_maximo([[3, 1, 4], [1, 5, 9], [2, 6, 5]]))
````

> **Exercício 4.4 (difícil)** — Escreva `eh_simetrica(M)` que retorna `True` se $M$ é quadrada e $M[i][j] = M[j][i]$ para todo $i, j$. **Dica:** basta verificar a metade superior (acima da diagonal). Teste com `[[1,2,3],[2,4,5],[3,5,6]]` (True) e `[[1,2],[3,4]]` (False).

````python
def eh_simetrica(M):
    n = len(M)
    if n != len(M[0]):
        return False
    for i in range(n):
        for j in range(i + 1, n):  # apenas metade superior
            if M[i][j] != M[j][i]:
                return False
    return True

print(eh_simetrica([[1, 2, 3], [2, 4, 5], [3, 5, 6]]))  # True
print(eh_simetrica([[1, 2], [3, 4]]))                  # False

# Verificar só j > i evita comparar cada par duas vezes (e M[i][i] consigo
# mesmo).
````

## §5 Por que NumPy?

*Tempo estimado: ~15 min*

**NumPy** é a biblioteca padrão para computação numérica em Python. Ela oferece o tipo `ndarray` (matriz N-dimensional) com **operações vetorizadas**, que são tipicamente **10× a 100× mais rápidas** que laços com listas.

Convenção universal: `import numpy as np`.

Funções de criação:

| Função | O que faz |
|---|---|
| `np.array([[1,2],[3,4]])` | Converte lista de listas em array |
| `np.zeros((m, n))` | Matriz $m \times n$ de zeros |
| `np.ones((m, n))` | Matriz de uns |
| `np.arange(start, stop, step)` | Vetor com valores em progressão |
| `np.linspace(a, b, n)` | $n$ pontos igualmente espaçados entre $a$ e $b$ |


````python
M = np.array([[1, 2, 3], [4, 5, 6]])
print(M)
print('shape:', M.shape)   # (2, 3)
print('dtype:', M.dtype)   # int64
print('ndim:', M.ndim)     # 2
````

## §6 Indexação, slicing 2D, máscaras booleanas

*Tempo estimado: ~20 min*

Em NumPy, o acesso a elementos é com **vírgula** dentro dos colchetes: `M[i, j]` (em vez de `M[i][j]` das listas).

Slicing funciona em cada dimensão:
- `M[:, 0]` — toda a primeira coluna.
- `M[1, :]` — toda a linha 1.
- `M[1:3, 0:2]` — submatriz das linhas 1–2 e colunas 0–1.

**Máscaras booleanas:** `M > 5` retorna um array booleano da mesma forma; `M[M > 5]` retorna um vetor 1D com os elementos que satisfazem a condição.

````python
M = np.arange(1, 13).reshape(3, 4)
print('M =')
print(M)
print()
print('M[1, 2] =', M[1, 2])
print('M[:, 0] =', M[:, 0])
print('M[1, :] =', M[1, :])
print('M[0:2, 1:3] =')
print(M[0:2, 1:3])
print('M[M > 5] =', M[M > 5])
````

## §7 Vetorização e broadcasting

*Tempo estimado: ~22 min*

**Vetorização**: operadores como `+`, `*`, `**` aplicam-se elemento-a-elemento sobre arrays de mesma forma. Não precisamos de laços.

**Broadcasting**: NumPy permite operar arrays de **formas compatíveis, mas diferentes**. Um escalar 'se expande' para cada elemento. Um vetor de forma `(n,)` casa com cada linha de uma matriz `(m, n)`. Um vetor coluna `(m, 1)` casa com cada coluna.


````python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[10, 20, 30], [40, 50, 60]])

print('A + B =')
print(A + B)
print('A * 2 =')
print(A * 2)
print('A**2 =')
print(A**2)

# Broadcasting com vetor (3,)
v = np.array([100, 200, 300])
print('A + v =')
print(A + v)
````

````python
M = np.array([[1.0, 2.0], [3.0, 4.0]])
M_centrada = M - M.mean()
print(M_centrada)
print('Soma:', M_centrada.sum())  # ~0
````

## §8 Agregações com axis, reshape, transpose

*Tempo estimado: ~18 min*

Agregações: `M.sum()`, `M.mean()`, `M.max()`, `M.min()`, `M.std()` retornam um único número (sobre todos os elementos).

Com `axis`, agregamos por eixo:

- `axis=0`: colapsa **linhas** → resultado tem uma entrada por **coluna**.
- `axis=1`: colapsa **colunas** → resultado tem uma entrada por **linha**.

**`argmax()`** retorna o **índice** (linear, ou pelo eixo) do maior valor.

**`reshape(m, n)`** reorganiza o array em outras dimensões (sem copiar dados).
**`M.T`** é a transposta.

````python
M = np.array([[1, 2, 3], [4, 5, 6]])
print('soma total:', M.sum())
print('soma por coluna (axis=0):', M.sum(axis=0))
print('soma por linha   (axis=1):', M.sum(axis=1))
print('argmax (linear):', M.argmax())
print('reshape em (3, 2):')
print(M.reshape(3, 2))
print('transposta M.T:')
print(M.T)
````

## §10 Operações por pixel

*Tempo estimado: ~18 min*

Como toda imagem é uma matriz, **todas as operações de NumPy** se aplicam:

- **Negativo**: `255 - img` (inverte cada pixel).
- **Brilho**: somar uma constante. Cuidado com **estouro** (`uint8` só vai de 0 a 255): use `np.clip` ou trabalhe em `int16`/`float`.
- **Threshold (binarização)**: `np.where(cinza > limiar, 255, 0)` — produz só preto e branco.

Carregue a imagem original (caso não esteja na memória):

## §11 Filtro de média 3×3

*Tempo estimado: ~15 min*

Um **filtro de média 3×3** suaviza a imagem: cada pixel é substituído pela média dos 9 pixels da vizinhança 3×3 ao seu redor (incluindo ele mesmo).

Vamos implementar duas versões: a **ingênua** com laço duplo, e a **vetorizada** com slicing — e comparar tempos.

Trabalhamos sobre a versão em **tons de cinza** (matriz 2D). As bordas (linha 0, última linha, coluna 0, última coluna) ficam com 0 — não há vizinhança 3×3 completa.

````python
img = iio.imread('imageio:chelsea.png')
cinza = img.mean(axis=2)
print('shape:', cinza.shape)
````

> **Exercício 11.1 (médio)** — Implemente `borrar_lento(M)` que retorna a imagem borrada usando **dois laços aninhados** sobre os pixels internos (de 1 a $h-2$ e 1 a $w-2$). Para cada pixel, some os 9 vizinhos e divida por 9. Teste em `cinza` e exiba.

````python
def borrar_lento(M):
    h, w = M.shape
    out = np.zeros_like(M)
    for i in range(1, h - 1):
        for j in range(1, w - 1):
            soma = 0.0
            for di in (-1, 0, 1):
                for dj in (-1, 0, 1):
                    soma += M[i + di, j + dj]
            out[i, j] = soma / 9
    return out

borrada_lenta = borrar_lento(cinza)
plt.imshow(borrada_lenta, cmap='gray')
plt.axis('off')
plt.title('Borrada (versão lenta)')
plt.show()
````

> **Exercício 11.2 (difícil)** — Implemente `borrar_rapido(M)` **sem laços**, usando soma de **9 fatias deslocadas** da matriz. O exemplo abaixo mostra o padrão:

```
M[:-2, :-2]   M[:-2, 1:-1]  M[:-2, 2:]
M[1:-1, :-2]  M[1:-1, 1:-1] M[1:-1, 2:]
M[2:, :-2]    M[2:, 1:-1]   M[2:, 2:]
```

Cada fatia tem dimensão $(h-2, w-2)$. Atribua a soma das 9 fatias dividida por 9 a `out[1:-1, 1:-1]`. Compare o tempo com `borrar_lento` usando `%timeit`.

````python
def borrar_rapido(M):
    M = M.astype(np.float64)
    out = np.zeros_like(M)
    out[1:-1, 1:-1] = (
        M[:-2, :-2]   + M[:-2, 1:-1]  + M[:-2, 2:] +
        M[1:-1, :-2]  + M[1:-1, 1:-1] + M[1:-1, 2:] +
        M[2:, :-2]    + M[2:, 1:-1]   + M[2:, 2:]
    ) / 9
    return out

borrada_rapida = borrar_rapido(cinza)
plt.imshow(borrada_rapida, cmap='gray')
plt.axis('off')
plt.title('Borrada (versão rápida)')
plt.show()

# Comparação de tempo
%timeit borrar_lento(cinza)
%timeit borrar_rapido(cinza)

# Cada slice deslocado captura uma posição da vizinhança 3x3 para todos os
# pixels internos de uma vez. A soma das 9 fatias é a soma das vizinhanças,
# vetorizada.
````
