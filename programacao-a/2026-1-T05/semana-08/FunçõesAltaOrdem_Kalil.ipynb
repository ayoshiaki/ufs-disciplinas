# Map

A função map() permite aplicar uma função a todos os elementos de um iterável (lista, tupla, conjunto, etc.) de forma elegante e declarativa.

Ela trabalha diretamente com iteradores e retorna um objeto iterável contendo os resultados da transformação aplicada.

## Sintaxe



````python
map(funcao, iteravel)
````

Ou com multiplos iteráveis

````python
map(funcao, iteravel1, iteravel2, ...)
````

## Funcionamento

O map():

1. percorre os elementos do iterável;
2. aplica a função informada;
3. retorna um iterador com os resultados.

## Exemplo básico

Vamos desenvolver uma função qe retorna os elementos de uma lista com os valores dobrados

### Sem map

````python
numeros = [1, 2, 3, 4]

resultado = []

for n in numeros:
    resultado.append(n * 2)

print(resultado)
````

### Com map

````python
numeros = [1, 2, 3, 4]

resultado = map(lambda n: n * 2, numeros)

print(list(resultado))
````

## Relação com funções lambda

O uso de map normalmente acontece com funções lambda, pois um dos seus argumentos é uma função

````python
nomes = ["ana", "carlos", "maria"]

resultado = map(lambda nome: nome.upper(), nomes)

print(list(resultado))
````

## map retorna um iterador

Na função map, python retorna um iterador, não uma lista

````python
numeros = [1, 2, 3]

m = map(lambda x: x * 2, numeros)

print(m) # <map object at ...>
````

Para visualizar os valores:

````python
print(list(m))
````

## Consumo do iterador

Um iterador pode ser consumido apenas uma vez. Na execução do programa é possível observar esse comportamento. Isso acontece porque o iterador já foi percorrido, o que torna o resultado vazio na segunda execução.

````python
m = map(lambda x: x * 2, [1, 2, 3])

print(list(m))
print(list(m))
````

## map com múltiplos iteráveis

Map consegue percorrer dois ou mais iteráveis simultaneamente.

````python
a = [1, 2, 3]
b = [10, 20, 30]

resultado = map(lambda x, y: x + y, a, b)

print(list(resultado))
````

# Filter

A função filter permite selecionar apenas os elementos que atendem a uma condição.

A função filter recebe:

* uma função predicado (*predicate*)
* um iterável

E retorna apenas os elementos para os quais o predicado é verdadeiro.

## Sintaxe

````python
filter(funcao, iteravel)
````

Onde:

* funcao: função que retorna True ou False
* iteravel: lista, tupla, string, range etc

Nesse contexto, a função filter:

* percorre todos os elementos do iterável
* aplica uma condição em cada elemento
* mantém apenas os valores cujo resultado seja True



## Exemplo básico

Filtrando números pares

````python
def eh_par(x):
    return x % 2 == 0

numeros = [1, 2, 3, 4, 5, 6]

resultado = filter(eh_par, numeros)

print(list(resultado))
````

````python
numeros = [1, 2, 3, 4, 5, 6]

resultado = filter(lambda x: x %2 == 0, numeros)

print(list(resultado))
````

Da mesma forma que a função map, a função filter retorna um iterador e não uma lista diretamente

## filter (None, iterável)

Existe um comportamento especial para a função filter:

````python
filter(None, iteravel)
````

Nesse caso, o python remove todos os elementos considerados falsos:

* 0
* False
* None
* ''
* listas vazias, etc

````python
dados = [0, 1, "", "Python", None, True]

resultado = filter(None, dados)

print(list(resultado))
````

## Mais exemplos



````python
alunos = [
    ("Ana", 8),
    ("Carlos", 5),
    ("Maria", 9),
    ("João", 4)
]

aprovados = filter(lambda aluno: aluno[1] >= 7, alunos)

print(list(aprovados))
````

# Reduce

A função reduce é usada para reduzir uma sequência de valores a um único resultado através da aplicação cumulativa de uma função.

Exemplos:

* somar todos os números
* multiplicar todos os elementos
* concatenar strings
* encontrar maior valor
* calcular produto acumulado

## Sintaxe

O reduce não é *built-in* no python. Ele fica no módulo *functools* e obirgatoriamente para ser usada, precisa ser importada no código

````python
from functools import reduce
````

Sua utilização segue a seguinte sintaxe:

````python
reduce(funcao, iteravel[, valor_inicial])
````

Onde:
* funcao: função que recebe 2 argumentos
* iteravel: sequência de elementos
* valor_inicial: valor opcional para iniciar a acumulação

## Exemplo básico


````python
from functools import reduce

numeros = [1, 2, 3, 4]

resultado = reduce(lambda x, y: x + y, numeros)

print(resultado)
````

Internamente, acontecem os seguintes passos:



```
((1 + 2) + 3) + 4 = 10
```



## Acumulador

O primeiro parâmetro da função representa o acumulador.

````python
lambda acumulador, elemento: ...
````

````python
from functools import reduce

numeros = [1, 2, 3, 4]

def soma(acumulador, numero):
    print(f'acumulador={acumulador}, numero={numero}')
    return acumulador + numero

resultado = reduce(soma, numeros)

print(resultado)
````

### Usando valor inicial

````python
from functools import reduce

numeros = [1, 2, 3]

resultado = reduce(lambda x, y: x + y, numeros, 10)

print(resultado)
````

O processo de execução seria:



```
(((10 + 1) + 2) + 3) = 16
```



## Mais exemplos

### Multiplicando elementos

````python
from functools import reduce

numeros = [1, 2, 3, 4]

produto = reduce(lambda x, y: x * y, numeros)

print(produto)
````

### Concatenando strings

````python
from functools import reduce

palavras = ['Python', 'é', 'funcional']

frase = reduce(lambda x, y: x + ' ' + y, palavras)

print(frase)
````

### Encontrando maior valor

````python
from functools import reduce

numeros = [10, 7, 25, 3]

maior = reduce(lambda x, y: x if x > y else y, numeros)

print(maior)
````
