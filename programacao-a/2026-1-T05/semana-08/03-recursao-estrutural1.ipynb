# 1. Recursão primitiva em listas - funções puras

O princípio da indução não se aplica apenas aos números naturais. Em particular aplica-se também a estruturas definidas indutivamente.

Podemos considerar listas como sendo estruturas indutivas. Uma lista `lst` é
+ a lista vazia `[]` (caso base); ou
+ formada pelo primeiro elemento, `lst[0]`, e seguida de um **resto**, `lst[1:]`.

A caraterística indutiva de `lst` emerge porque o seu resto também é uma lista.

Podemos resolver problemas que envolvem listas usando uma abordagem indutiva, de forma similar a como fizemos com problemas com números naturais. Assim, para resolver um problema $P$ sobre listas usando **recursão primitiva**, seguimos estes dois passos:
1. Damos a solução para o caso da lista ser vazia (caso base).
2. Supondo, como hipótese indutiva (HI), que temos a solução para o resto da lista, construímos a solução para a lista completa.

## 1.1 Soma dos elementos de uma lista

Como um primeiro exemplo, definamos uma função `somaLista` que calcula a soma de todos os elementos de uma lista, assim
1. `somaLista`([])` é igual com zero (caso base).
2. Supondo que temos o valor de `somaLista(ns[1:])` (HI), podemos obter o valor de`somaLista(ns)` com a soma  `ns[0] + somaLista(ns[1:])`.

A correspondente definição em Python é

````python
def somaLista(ns) :
  if ns == [] :
    return 0                           # caso base
  else :
    return ns[0] + somaLista(ns[1:])   # solução indutiva

somaLista([20, 10, 30, 5])
````

Observe que a suposição da HI aparece na definição como uma chamada recursiva passando como argumento o resto da lista em questão.

Em termos de eficiência, a definição anterior, apesar de correta, não é uma definição aceitável em Python. Seria adequada se estivéssemos programando em uma linguagem funcional como Ocaml, Lisp ou Haskell, onde o conceito e implementação de listas é bastante diferente de Python.

O problema desta definição está no uso do slice `ns[1:]` como argumento na chamada recursiva. Um slice produz uma nova lista, copiando elementos da lista original. Isto significa que a cada chamada recursiva estaremos criando uma nova lista, o que torna a definição muito ineficiente.

Uma forma de evitar o uso de slices é definirmos uma função auxiliar `somaListaDesde` com um argumento extra para indicar a posição inicial da lista a partir da qual queremos fazer o cômputo da soma. Assim, `somaListaDesde(ini)` calculará a soma da porção de `ns` que inicia em `ini` e vai até o fim de `ns`.

Podemos pensar, então, que estamos definindo `somaListaDesde` por recursão primitiva no tamanho da porção que inicia em `ini` e vai até o fim. Uma chamada recursiva com `ini+1`, diminui em um o tamanho da porção.

````python
def somaLista(ns) :
  # Calcula a soma dos elementos de ns a partir da posição ini em diante
  def somaListaDesde(ini) :
    if ini == len(ns) :
      return 0                                    # caso base (a porção é vazia)
    else :
      return ns[ini] + somaListaDesde(ini+1)  # solução indutiva

  return somaListaDesde(0)


somaLista([20, 1, 4, -5])
````

Talvez tenha se surprendido por que a definição de `somaListaDesde` está indentada, ou seja, está dentro da definição de `somaLista`. Neste caso, temos que a definição de `somaListaDesde` é local à definição de `somaLista` e portanto só pode ser usada dentro do corpo de `somaLista`.

Uma outra alternativa para a definição de `somaLista` é explorar a possibilidade que Python oferece de definir argumentos com valor default. Assim, a função `somaLista` terá dois parâmetros, a lista e a posição inicial, porém o segundo parâmetro será opcional com zero (0) como valor default.    



````python
def somaLista(ns, ini=0) :
  if ini == len(ns) :
    return 0
  else:
    return ns[ini] + somaLista(ns, ini+1)

print(somaLista([10, 20, 15, -5]))
print(somaLista([10, 20, 15, -5], ini=2))
````

A chamada somaLista([10, 20, 15, -5]) implicitamente passará 0 como segundo argumento, pois este é o valor default declarado no cabeçalho da definição da função. Assim ela calculará a soma de todos os elementos da lista.

Já a chamada somaLista([10, 20, 15, -5], ini=2) passa explicitamente o valor de 2 como segundo argumento. Aqui a função calculará a soma da lista a partir da posição 2.

Vejamos mais uma versão de `somaLista`, também por indução no tamanho de uma porção de `ns`, mas agora com o argumento extra indicando o final da porção. Agora, a porção vai do início de `ns`, posição 0, até a pósição indicada.

````python
def somaLista(ns) :
  # Calcula a soma dos elementos de ns desde a posição 0 até fim
  def somaListaAte(fim) :
    if fim == -1 :
      return 0                              # caso base (a porção é vazia)
    else :
      return somaListaAte(fim-1) + ns[fim]  # solução indutiva

  return somaListaAte(len(ns)-1)


somaLista([10, 20, 15, -5])
````

Note que a chamada recursiva `somaListaAte(fim-1)` diminui em um o tamanho da porção.

## 1.2 Quantidade de pares

Como outro exemplo, definamos uma função que calcule a quantidade de pares em uma lista de inteiros `ns`. Usemos indução na porção de `ns` que começa a partir do `i`.

1. Se `i >= len(ns)`, há zero pares na porção, pois ela é vazia (caso base).
2. Se `i < len(ns)`
    + Se `ns[i]` é par, a quantidade de pares a partir da posição `i` é um a mais do que a quantidade de pares a partir da posição `i+1`.
    + Caso contrário, a quantidade de pares, tanto a partir de `i` como a partir de `i+1`, é a mesma.

````python
def par(n) :
  return n % 2 == 0

def quantidadePares(ns, i = 0) :
  if i >= len(ns) :
    return 0  # caso base
  else :
    # solução indutiva
    if par(ns[i]) :
      return 1 + quantidadePares(ns, i+1)
    else :
      return quantidadePares(ns, i+1)

quantidadePares([2, 3, 2, 4, 1, 7, 10, 12])
````

# 2. Recursão primitiva - procedimentos

As definições anteriores são de funções puras.
Obviamente, podemos usar indução também para definir procedimentos.

Por exemplo, a impressão dos elementos de uma lista `lst` usando recursão na porção de `lst` iniciada em `i` é dada por

1. Se a porção for vazia (`i >= len(lst)`), não temos que fazer nada.
2. Se não for vazia, imprimimos `lst[i]` e logo a seguir, indutivamente, imprimimos todo o resto da lista a partir da posição `i+1`.



````python
def imprimirLista(lst, i = 0) :
  if i < len(lst) :
    print(lst[i])
    imprimirLista(lst, i+1)

imprimirLista(["Maria", "Pedro", "Joao", "Jonatas", "Carlos"])
````

Note que aqui o caso base requer não fazer nada. Note também que a ordem em que estão as linhas 3 e 4 é relevante na solução. Começa-se imprimindo o primeiro da lista para depois imprimir o restante da lista.

## 2.2 Dobrar os elementos de uma lista numérica

Dada uma lista de números `ns` queremos alterar todos seus elementos dobrando-os (multiplicando por dois). Indutivamente, na porção de `ns` a partir de `i`, temos

1) Se `i >= len(ns)`, a porção é vazia e, portanto, não precisamos fazer nada
2) se `i < len(ns)`, dobramos `ns[i]` e, recursivamente, dobramos todo o resto de `ns` a partir da posição `i+1`.

````python
def dobrar(ns, i=0) :
  if i < len(ns) :
    ns[i] *= 2
    dobrar(ns, i+1)

nums = [2, 4, 0, 1, 5, 10]
dobrar(nums)
nums
````

# 3. Estudo de caso: Avaliação de polinômios

Abordemos o problema de computar o valor de um polinômio de grau $n$

$$p(x) = a_n x^n + a_{n-1} x^{n-1} + ... + a_1 x + a_0$$

Suponhamos que um polinômio é representado em Python por uma lista contendo os seus coeficientes. Assim, o polinômio acima será representado pela lista $[a_0, a_1, ..., a_{n-1}, a_{n}]$.

Pensando recursivamente, talvez a solução mais natural seja
1. Caso base: se o polinômio é de grau zero, a solução é $a_0$.
2. A avaliação do polinômio $p(x)$ é igual ao termo de grau $n$, $a_n x^n$, mais a avaliação do polinômio de grau $n-1$, $a_{n-1} x^{n-1} + ... + a_1 x + a_0$

````python
def avaliaPolinomio(a, x) :

  def avalia(n) :
    if n == 0 :
      return a[0]
    else :
      return a[n]*x**n + avalia(n-1)

  return avalia(len(a)-1)

avaliaPolinomio([1, 2, 1, 2], 3)
````

## 3.1 Técnica de fortalecer a Hipótese Indutiva

O operador de exponenciação `**` é mais custoso, em termos de eficiência, do que uma multiplicação. Inclusive, se a sua implementação interna fosse muito ingênua, ela faria $n-1$ multiplicações. De qualquer forma, mesmo implementada eficientemente, ela continua mais cara do que uma simples multiplicação. Será que podemos dispensar o uso do operador `**`?

Obviamente, para ter efetividade, teríamos que substituir o uso deste operador por algo que calcule a cada chamada recursiva $x^k$ com custo apenas de uma multiplicação.

Uma técnica frequentemente usada é **fortalecer a Hipótese Indutiva**.
Em particular, para a avaliação de polinômios, podemos escolher uma hipótese indutiva mais forte assumindo
+ que temos a solução para o polinômio de grau $n-1$, sem considerar o último termo; e, ao mesmo tempo,
+ que temos calculado $x^{n-1}$.

Com esta HI mais forte, podemos calcular $x^n$ apenas com uma multiplicação por $x$ ao valor já calculado $x^{n-1}$. E, com o valor de $x^{n}$, podemos calcular a avaliação do polinômio completo somando o termo $a_n x^{n}$ à avaliação do polinômio de grau $n-1$ da HI.

Em termos de implementação, a definição recursiva agora retorna dois valores. Um é o valor do polinômio e outro é o valor da potência de $x$ do último termo, isto é $x^{n}$. Assim, temos

````python
def avaliaPolinomio(a, x) :
  def avalia(n) :
    if n == 0 :
      return a[0], 1
    else :
      p , xn = avalia(n-1)
      xn = x*xn
      return a[n]*xn + p, xn

  return avalia(len(a)-1)[0]

avaliaPolinomio([1, 2, 1, 2], 3)
````

## 3.2 Regra de Horner

Considere novamente o polinômio
$$p(x) = a_n x^n + a_{n-1} x^{n-1} + ... + a_1 x + a_0$$

Fatorando o $x$ nos $n$ primeiros termos temos que

$$p(x) = (a_n x^{n-1} + a_{n-1} x^{n-2} + ... + a_1) x + a_0$$

Observe que, no lado direito da equação, temos, entre parênteses, um polinômio de grau $n-1$. Podemos então fazer a chamada recursiva para avaliar este polinômio. Assim, temos

````python
# Avalia o polinômio de grau (n-i) com os coeficientes em a[i:].
# Ou seja, avalia:  a[n]*x**(n-i) + a[n-1]*x**(n-i-1) + .. a[i]
def horner(a, x, i = 0) :
  if i == len(a) :
    return 0
  else :
    return horner(a, x, i+1)*x + a[i]

horner([1, 2, 1, 2], 3)
````

<!-- REVISAR -->

Como vimos ao longo deste documento, pensar de forma recursiva (indutiva) é uma ferramenta poderosa. Ela permite-nos resolver problemas complexos quebrando-os em partes menores, como demonstrado na elegância da Regra de Horner, por exemplo.

No entanto, no mundo real da engenharia de software com Python, a recursão pode esbarrar em dois problemas práticos:
1. **Sobrecarga de memória:** Cada chamada recursiva cria um novo contexto na memória do computador. Se houverem muitas chamadas recursivas a quantidade de memória total necessária pode ser grande.
2. **Limite de profundidade:** O Python tem um limite de chamadas recursivas empilhadas (geralmente 1000), o que pode originar um `RecursionError` em listas muito grandes.

Qualquer algoritmo pensado recursivamente pode ser traduzido para uma versão iterativa,  utilizando laços `while` ou `for`.
Se bem que a tradução nem sempre é trivial, a excelente notícia é que, em muitos casos, ela é bastante direta.

Às vezes, dada uma solução recursiva, pode-se implementar uma solução iterativa baseada em cômputo para a frente. É o caso da maioria dos problemas vistos até aqui.
Por exemplo, vejamos o exemplo da função `somaLista` vista no início deste capítulo.

```Python
def somaLista(ns) :
  def somaListaAte(fim) :
    if fim == -1 :
      return 0
    else :
      return somaListaAte(fim-1) + ns[fim]

  return somaListaAte(len(ns)-1)
```

A tabela de cômputo para frente começa no caso base, com `fim == -1` e avança incrementando `fim` de um em um, calculando `somaListaAte(fim)`até que `fim` chegue a ser o último índice de `ns`. Assim, por exemplo, para `ns = [10, 20, 15, -5]`, temos a seguinte tabela de cômputo para a frente.

| `fim` | `somaListaAte(fim)` | Justificativa|
|---:|---:|:--:|
|-1 | 0 | caso base
|0 | 10 | `somaListaAte(-1) + ns[0]` |
|1| 30 |`somaListaAte(0) + ns[1]` |
|2 | 45 | `somaListaAte(1) + ns[2]` |
|3 | 40 | `somaListaAte(2) + ns[3]`

Observe que para calcular o valor de `somaLista` de uma linha qualquer que não seja a inicial, precisamos apenas do valor de `somaLista` da linha anterior. Podemos então definir uma função iterativa que simule o cálculo linha a linha desta tabela assim


````python
def somaListaIterativo(ns) :
  soma = 0 # Usamos esta variável para calcular somaLista de c/linha da tabela
           # Inicializamos com o valor do caso base (primeira fila)
  for fim in range(len(ns)) :
    soma = soma + ns[fim]   # calcula o valor de somaLista(fim)
                            # usando o valor anterior (em soma)
  return soma
````

Note a correspondência matemática direta: a variável `soma` atua como o recetor dos valores que antes ficavam suspensos na pilha de chamadas da recursão. Em particular, a linha 6 deste código
```Python
soma = soma + ns[fim]
```
reproduz a expressão correspondente da versão recursiva:
```Python
return somaListaAte(fim-1) + ns[fim]
```
