# 🐍 **Dicionários em Python**

## **1. Motivação: Por que usar Dicionários?**

Imagine que você precisa armazenar os preços de diferentes frutas.

Usando Listas: Você teria uma lista para os nomes ['maçã', 'banana', 'uva', 'laranja'] e outra para os preços [5.50, 3.20, 4.50, 6.15]. Para saber o preço da banana, você teria que descobrir em qual posição a "banana" está na primeira lista e acessar a mesma posição na segunda. Complexo e lento!

Tente escrever um código fonte em Python utilizando listas que leia o nome da fruta e responda para o usuário qual é o preço da mesma.

````python
nomes = ['maçã', 'banana', 'uva' ,'laranja']
precos = [5.50, 3.20, 4.50, 6.15]

# Passo 1: solicite o nome da fruta

# Passo 2: busque o nome da fruta na lista de nomes

# Passo 3: sabendo em qual posição da lista está o nome da fruta, acesse o preço correspondente na lista de preços e o imprima
````

O que achou da solução com listas?

Observe que para achar o nome da fruta, foi feita uma busca do nome da mesma na lista de nomes e com base na posição do nome nesta lista foi deduzida a posição na lista de preços.

Pense também na manutenção dos preços. Para alterar um preço, este mesmo processo de buscar o nome e encontrar a posição deve ser realizado.

E para remover uma fruta, temos que fazer o processo novamente e ainda excluir o nome da lista nomes e o preço da lista precos.




**Usando Dicionários: Você cria um mapeamento direto. A "chave" é a fruta e o "valor" é o preço. O acesso é instantâneo.**

````python
# A estrutura é {chave1: valor1, chave2: valor2, ...}
precos = {
    "maçã": 5.50,
    "banana": 3.20,
    "uva": 4.50,
    "laranja": 6.15
}

print(f"O preço da uva é: R$ {precos['uva']}")
````

Perceba que para acessar o preço da uva, basta informar a 'uva' (a chave) entre colchetes e esta operação é realizada de forma muito mais eficiente do que a abordagem apresentada com listas.

Os dicionários em Python são implementados pela classe **dict**. Você pode pensar no dicionário como se fosse uma tabela:

| fruta (chave) | preço (valor associado) |
| :---: | :---: |
| maçã | 5.50 |
| banana | 3.20 |
| uva | 4.50 |
| laranja | 6.15 |



##**2. Métodos Essenciais da Classe dict**


O Python oferece métodos integrados para manipular esses dados de forma eficiente. Vamos explorar os principais:

### **2.1. Criando um dicionário**


Para criar um dicionário, há vária alternativas. Uma delas é criar usando o contrutor **dict()** . Outra forma é com abre e fecha chave **{}**. Nestes dois casos, o dicionário está vario.

```
dic1 = dict()  # dic1 está vazio
dic2 = {}      # dic2 também está vazio
```

Caso se deseje criar um dicionário contendo valores iniciais, podemos especificar entre abre e fecha chaves os pares (chave, valor) separados entre si por vírgulas, sendo que a chave e o valor são separados por um caractere :

Por exemplo, criando um dicionários com nomes e telefones:


```
contatos = {
    "Ana": "(79)99988-7766",
    "Pedro": "(82)99933-4567",
    "Beto": "(79)99955-4433"
}
```

Uma outra alternativa mais compacta é utilizando o construtor dict e passando os parâmetros diretamente:

```
contatos = dict(Ana = "(79)99988-7766",
                Pedro = "(82)99933-4567",
                Beto = "(79)99955-4433")
```


###**2.2. Acesso Seguro com o método get()**


Tentar acessar uma chave que não existe com colchetes causa um erro (KeyError). O método **get()** evita isso retornando **None** ou um **valor padrão**.

Vamos pensar em um outro exemplo. Que tal criar um dicionário para guardar os contatos telefônicos de seus amigos? A chave seria o nome e o valor associado seria o número do telefone.

````python
contatos = {
    "Ana": "(79)99988-7766",
    "Pedro": "(82)99933-4567",
    "Beto": "(79)99955-4433"
}

# Acessando um nome que existe usando get
telefone_pedro = contatos.get("Pedro", "Contato não encontrado")
print(telefone_pedro)

# Tentando buscar um nome que não existe usando get
telefone_carlos = contatos.get("Carlos", "Contato não encontrado")
print(telefone_carlos)

# Tente buscar um nome que não existe sem usar get
print(contatos["Pedro"])
````

###**2.3. Acessando chaves, valores e itens**


Muitas vezes precisamos iterar apenas sobre os nomes ou apenas sobre os dados. Para isso, a classe dict oferece os seguintes métodos:
- **keys()** -> Retorna as chaves
- **values()** -> Retorna os valores
- **items()** -> Retorna os pares (chave, valor) como uma sequência de tuplas

````python
print("Chaves:", contatos.keys())
print("Valores:", contatos.values())
print("Pares:", contatos.items())
````

**Exemplo 1:** Mostrando todos os nomes contidos no dicionário de contatos e seus respectivos números de telefone com auxílio do método keys().

````python
# o método keys retorna uma sequência com os nomes (chaves) e é possível acessar o valor usando o get ou com a sintaxe de colchetes
for nome in contatos.keys():
    print(f"{nome} -> {contatos[nome]}")
````

````python
for nome, fone in contatos.items() :
  print(f"{nome} -> {fone}")
````

### **Verificando se uma chaves está no dicionário**

Para verificar se uma há uma chaves no dicionário podemos fazer

````python
"Ana" in contatos.keys()
````

No entanto, uma forma mais curta é

````python
"Ana" in contatos
````

**Exemplo 2**: Também é possível acessar as chaves se utilizarmos apenas o dicionário no laço for.

````python
# o método keys retorna uma sequência com os nomes (chaves) e é possível acessar o valor usando o get ou com a sintaxe de colchetes
for nome in contatos:
    print(f"{nome}: {contatos[nome]}")
````

**Exemplo 3:** Mostrando todos os valores contidos no dicionário de contatos com auxílio do método values().

````python
# o método values retorna uma sequência com os telefones (valores) não sendo possível saber a qual chave corresponde
for telefone in contatos.values():
    print(f"{telefone}")
````

**Exemplo 4:** Mostrando todos os nomes contidos no dicionário de contatos e seus respectivos números de telefone com auxílio do método items().

````python
# o método items retorna uma sequência com os pares (chave, valor) do dicionário os quais são atribuídos às variáveis de iteração nome e telefone.
for nome, telefone in contatos.items():
    print(f"{nome}: {telefone}")
````

###**2.4. Mudando valores no dicionário**


Muitas vezes precisamos modificar um valor associado a uma chave no dicionário.

Por exemplo, podemos precisar alterar o telefone de algum dos seus amigos no dicionário de contatos. Para isto, você pode utilizar:

````python
# Mudando o telefone de Ana
contatos['Ana'] = '(71)93234-2555'

# Podemos também acrescentar elementos no dicionário
contatos['Joao'] = "(79) 98888 7777"
# Conferindo o conteúdo do dicionário
print(contatos)
````

###**2.5. Mudando valores no dicionário com update**


Em algumas situação precisamos atualizar vários valores associados em conjunto e esses novos dados estão armazenados em um outro dicionário.

Neste caso, podemos utilizar o método **update()**, que recebe um dicionário como argumento e os valores contidos neste dicionário serão modificados ou incluídos no dicionário original.

Observe que em um dicionário não é possível ter chaves duplicadas, portanto durante o update, quando há chaves duplicadas, o valor associado à chave já existente será substituído pelo passado no update.

````python
contatos = {'Ana': '(71)93234-2555', 'Pedro': '(82)99933-4567', 'Beto': '(79)99955-4433'}
print(f'Contatos antes: {contatos}')

# Pedro já existe e Maria é um contato novo
contatos_atualizados = {'Pedro':'(75)99887-6655', 'Maria': '(11)99811-1223'}

# Este comando vai mudar o telefone de Pedro e incluir Maria com seu telefone no dicionário
contatos.update(contatos_atualizados)

print(f'Contatos atualizados: {contatos}')
````

###**2.6. Removendo uma chave do dicionário**

Pense que você poderia ter incluído um nome por engano no dicionário. Como fazer para removê-lo do dicionário?

Para isso a classe dict oferece o método **pop()**, o qual recebe a chave e executa a retirada do par (chave, valor) correspondente do dicinário.

O método pop retorna o valor que estava associado à chave como retorno da chamada, podendo ser utilizado.

````python
contatos = {'Ana': '(71)93234-2555', 'Pedro': '(75)99887-6655', 'Beto': '(79)99955-4433', 'Maria': '(11)99811-1223'}

# removendo 'Maria' e guardando seu telefone
telefone_maria = contatos.pop('Maria')

print(f'Telefone da Maria: {telefone_maria}')
print(f'Contatos após remoção: {contatos}')
````

O que acontece se você tentar remover com o pop uma chave que não existe no dicionário?

Mude o nome de 'Maria' para outro nome que não exista no dicionário, execute novamente e descubra o que acontece!

Uma outra alternativa para remoção é usando a palavra-chave **del**. Por exemplo, considerando o exemplo dos contatos acima, poderíamos remover o contato de 'Maria' por meio do comando:


```
del contatos['Maria']
```

### **2.7 Outros métodos úteis em dicionários**

O método **clear()** limpa todos os dados do dicionário, deixando-o totalmente vazio.

O método **copy()** retorna uma cópia do dicionário atual, o que pode ser útil se você quiser preservar o estado inicial dele antes de fazer modificações.


````python
contatos = {'Ana': '(71)93234-2555', 'Pedro': '(75)99887-6655', 'Beto': '(79)99955-4433', 'Maria': '(11)99811-1223'}

# Realiada a cópia do conteúdo do dicionário contatos para copia_contatos
copia_contatos = contatos.copy()

# O dicionário original com os contatos é limpo
contatos.clear()

print(f'Cópia de contatos: {copia_contatos}')
print(f'Contatos após remoção: {contatos}')
````

### **2.8 Acessando chaves de dicionários de forma ordenada**

Os dicionários são implementados de uma forma em que a busca pelo valor associado à chave é feito de uma forma bastante eficiente, já que utiliza tabelas Hash. Como consequência disso, não é possível ordenar os dados usando o método sort disponível na classe **list**.

Porém, podemos utilizar a função **sorted()** para ordenar as chaves de um dicionário e obter como resultado uma lista. Observe que o dicionário não sofre qualquer mudança, ou seja, a ordenação não ocorre no dicionário mas na lista que é retornada pela função sorted.

Por exemplo, suponha que você deseja ver os seus contatos na ordem alfabética. Pode observar que ao imprimir os nomes contidos no dicionário contatos, a ordem em que aparece é a de insercão. A partir da versão 3.7, este é o comportamento padrão.

Vejamos um exemplo de como poderíamos utilizar a função sorted para acessar os contatos em ordem alfabética logo abaixo:

````python
contatos = {'Ana': '(71)93234-2555', 'Pedro': '(75)99887-6655', 'Beto': '(79)99955-4433', 'Maria': '(11)99811-1223'}

contatos_ord = sorted(contatos)
print(f'Tipo de list_contatos {type(contatos_ord)}')
for nome in contatos_ord:
    print(f"{nome}: {contatos[nome]}")
````

### **3.1. Categorização**

Uma das aplicações mais comuns de dicionários é fazer a contagem de algo por categoria, sendo a categoria a chave do dicionário e a quantidade de vezes em que ocorreu sendo o valor.

Por exemplo, suponha que você precisa contar quantas vezes cada palavra aparece em um texto. Você poderia utilizar um dicionário para isso.

Não seria interessante que fossem mostrados as palavras na ordem da que mais aparece para a que menos aparece em uma tabela? Como seria possível fazer isso? Vimos como poderíamos ordenar pela chave do dicionário, mas como fazer para ordenar pelo valor associado?

Uma possibilidade é utilizar o parâmetro **key** da função **sorted**, passando uma função que retorna qual será o valor utilizado como critério de ordenação para cada item. Como cada item retornado pelo método **items** é uma tupla com a chave e o valor associado, retorna-se o *item[1]*, já que este corresponde à quantidade de ocorrências da palavra e é o segundo valor contido na tupla de cada item.

Adicionalmente, utilizamos o parâmetro **reverse** para que a ordem seja decrescente (reversa).

````python
palavras_ord_qtd = sorted(conts.items(), key=lambda item: item[1], reverse=True)
print('Ordenado pela quantidade:')
for palavra, qtd in palavras_ord_qtd:
   print(f'{palavra:<10}:{qtd:>3}')
````

### **3.2 Armazenamento e troca de dados de tipos heterogêneos**

Um cenário em que dicionários são utilizados é na trocada de dados via requisições e respostas a APIs REST (*Representational State Transfer ou Transferência de Estado Representacional*), cujo formato de dados normalmente é [JSON (JavaScript Object Notation - Notação de Objetos JavaScript](https://www.json.org/json-pt.html).




Por exemplo, suponha que você obteve a seguinte resposta em JSON de uma API REST que fornece dados sobre o clima.

````python
{
  "temp": 25,
  "date": "2026-03-02",
  "time": "14:00",
  "condition": "Cloud",
  "description": "Nublado",
  "currently": "dia",
  "humidity": 60,
  "forecast": [
    {"date": "03/03", "max": 28, "min": 18, "condition": "rain"},
    {"date": "04/03", "max": 26, "min": 17, "condition": "storm"}
  ]
}
````

O conteúdo acima pode ser facilmente transformado em uma estrutura de dados em Python do tipo dicionário, contendo as chaves temp, date, time, condition, description, currently, humidity e forecast. Observe que esta última está associada a uma valor do tipo lista de dicionários com as chaves date, max, min e condition.

````python
import json

# Este conteúdo poderia ter sido obtivo via chamada direta a API REST
conteudo = '''
{
  "temp": 25,
  "date": "2026-03-02",
  "time": "14:00",
  "condition": "Cloud",
  "description": "Nublado",
  "currently": "dia",
  "humidity": 60,
  "forecast": [
    {"date": "03/03", "max": 28, "min": 18, "condition": "rain"},
    {"date": "04/03", "max": 26, "min": 17, "condition": "storm"}
  ]
}
'''
# O método loads carrega os dados contidos no parâmetro do tipo string e cria uma
# estrutura de dados correspondente, que pode conter dicionários e listas aninhadas
clima_dic = json.loads(conteudo)

print(type(clima_dic))
print(type(clima_dic['forecast']))

print(clima_dic)
for chave, valor in clima_dic.items():
  print(chave, valor)
````
