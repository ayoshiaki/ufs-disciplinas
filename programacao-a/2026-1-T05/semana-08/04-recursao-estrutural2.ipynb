# Estruturas indutivas

Existem várias estruturas que por natureza são indutivas. Por exemplo, uma árvore genealógica. Por simplicidade, consideremos apenas genealogia de mãe. Graficamente, podemos representar uma árvore genealógica assim:

![alt text](https://drive.google.com/uc?export=view&id=1lAhpWIUnplcum6iLa78BakBlJPbK9kF8)


Nesta figura, acima está a mãe e embaixo as filhas(os). Assim, Maria é mãe de Joana, Pedro, Patricia e Marcos, por exemplo.
Observe que qualquer árvore genealógica é indutiva, no sentido de que a árvore de uma mãe contém as (sub)árvores das filhas. Sendo assim, muitas funções que manipulam árvores genealógicas naturalmente podem ser definidas usando recursão estrutural (recursão na estrutura da árvore).

A árvore do exemplo anterior poderia ser representada em Python usando uma dupla com o nome da mãe que está na *raiz* da árvore e uma lista dos descendentes diretos (filhos), assim:

```
( "Maria", [ ("Joana", [ ("Lucio", []),
                         ("Jõao", [])
                       ]
             ),
             ("Pedro",[]),
             ("Patricia", [ ("Marina", [ ("Marcelo", []),
                                         ("Tomás", [])
                                       ]
                            )
                          ]
             ),
             ("Marcos",[])
           ]
)
```

Uma função que calcule todos os descendentes diretos e indiretos da *raiz* de uma árvore genealógica pode ser definida assim:
1) Caso base: Se a *raiz* não tem filhos, o resultado é o conjunto vazio
2) Se a raiz tem filhos, os filhos são descendentes e, recursivamente, também os descendentes de todos os filhos.





````python
def descendentes(a) :
  if a[1] == [] :
    return []
  else :
    des = []
    for filho in a[1] :
      des.append(filho[0])
      des.extend(descendentes(filho))
    return des

arv = ( "Maria", [ ("Joana", [ ("Lucio", []),
                               ("Jõao", [])
                             ]
                   ),
                   ("Pedro",[]),
                   ("Patricia", [ ("Marina", [ ("Marcelo", []),
                                               ("Tomás", [])
                                             ]
                                  )
                                ]
                   ),
                   ("Marcos",[])
                 ]
      )

descendentes(arv)
````
