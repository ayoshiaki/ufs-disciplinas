# 🐍 **Conjuntos em Python**

## **1. Motivação: Por que usar Conjuntos?**

Imagine que você está desenvolvendo um software que executa em um equipamento de rede de computadores chamado _Firewall_. Este equipamento é um dispositivo ou software de segurança de rede que monitora e controla o tráfego de entrada e saída com base em regras de segurança predefinidas.

Uma das funcionalidades deste equipamento/software é impedir que os usuários acessem determinados sites, ou seja, deve haver uma lista de sites proibidos.

Este software também poderia classificar os sites por nível de prioridade, dando mais banda de rede para quem estiver acessando sites relacionados ao trabalho e menos banda de rede para sites menos relacionados ao trabalho (exemplos: Instagram, Youtube, Facebook, Netflix e outros...).

Raciocinando como um programador Python, seria natural você pensar em utilizar uma lista para armazenar a relação dos sites. É uma abordagem que de fato vai funcionar. Porém, o desempenho não seria muito bom. Isso ocorre porque a busca na lista é sequencial e teria um custo linear O(N), ou seja, proporcional ao número de sites armazenados nela.

Uma estrutura de dados que seria melhor para esta aplicação seria o Conjunto (**set**) disponível em Python. Um **set** em Python é implementado com uma **Tabela Hash**. Você vai aprender mais sobre esta estrutura na disciplina de Algoritmos e Estruturas de Dados 1.

Por hora, podemos adiantar que os dados são armazenados nas posições da tabela _hash_ de acordo com o valor calculado por uma função _hash_, que é aplicada sobre o valor (endereço do site) e retorna sua posição. Esta abordagem tem a vantagem de ter um custo constante O(1), isto é, independentemente do tamanho do set, o custo será o mesmo. Lembre-se de que nas listas seria O(N).

Vejamos como a lista de sites proibidos poderia ser implementada com a estrutura de dados **list**:

````python
sites_proibidos = ['a.com', 'b.com.br', 'c.net' ,'d.edu.br', 'jogos.net']
print(type(sites_proibidos))
site = input('Digite o endereço do site: ')

if site in sites_proibidos:
  print('Este site está proibido!')
  print('Acesso negado!')
else:
  print('Acesso permitido!')
````

O que achou da solução com listas?

Em termos de programação é fácil de entender e implementar. Porém, se você utilizasse um **set** teria um **desempenho melhor**.

Então vamos ver como seria este mesmo exemplo implementado com um conjunto (**set**):

````python
sites_proibidos = {'a.com', 'b.com.br', 'c.net' ,'d.edu.br', 'jogos.net'}
print(type(sites_proibidos))
site = input('Digite o endereço do site: ')

if site in sites_proibidos:
  print('Este site está proibido!')
  print('Acesso negado!')
else:
  print('Acesso permitido!')
````

Qual é a diferença entre os dois exemplos?

**Dica:** a sintaxe de criação de um set é ligeiramente diferente da lista. Ao invés de colchetes [ ], nos conjuntos utilizamos as chaves { } para envolver os valores.



Uma outra coisa que você precisaria pensar é na atualização dos valores. Pense que a relação de sites é enorme e que eventualmente você vai receber uma solicitação para colocar um site na lista. Porém, antes de inserí-lo é preciso confirmar se ele já se encontra na relação de sites proibidos para evitar inserir um valor duplicado. Portanto, para conferir se o site já está na lista, é preciso percorrer toda a lista, o que tem um custo proporcional ao tamanho dela.

Já no caso do set, ele tem como característica não permitir que sejam incluídos valores duplicados. Quando se chama o método **add** passando um novo valor para o conjunto, este só será incluído no conjunto se não estiver presente. Esta operação é realizada de maneira bastante eficiente porque ao tentar inserir um valor, é utilizada a função _hash_ que calcula a posição e o custo de checar se o valor já existe no set é constante, ou seja, não aumenta de forma proporcional ao tamanho no conjunto.

````python
sites_proibidos = {'a.com', 'b.com.br', 'c.net' ,'d.edu.br', 'jogos.net'}
site = input('Digite o endereço do site que deseja incluir: ')
sites_proibidos.add(site)
print('Conjunto após a inserção', sites_proibidos)
````

Conjuntos são objetos iteráveis. Portanto, podemos usar laços `for` para iterar sobre todos os elementos de um conjunto.

````python
for s in sites_proibidos :
  print(s)
````

Então até o momento temos **duas aplicações de Conjuntos**:
- Verificar se um valor está presente em um conjunto de valores
- Armazenar valores evitando duplicação

## **2. Principais métodos de Conjuntos**

Os sets em Python possuem os seguintes métodos que permitem manipular os valores contidos nos mesmos:
- **add**: adiciona o valor no conjunto, se já não estiver lá
- **discard**: descarta o valor passado, retirando-o do conjunto. Não gera erro quando não consegue.
- **remove**: remove o valor passado do conjunto. Gera erro se não conseguir
- **pop**: retira um valor arbitrário do conjunto
- **update**: atualiza o conjunto com uma sequência de valores
- **clear**: limpa o conjunto (remove todos os valores)
- **copy**: cria uma cópia do conjunto

**Vejamos um exemplo utilizando os principais métodos básicos da classe set:**

````python
nums = {1,2,3,4,5}
print(f'1: {nums}')

nums.add(6)
nums.add(10)
nums.add(4) # já está no conjunto, logo nada muda
print(f'2: {nums}')
````

````python
nums.update({4,9,11}) # atualizando o conjunto com os dados de outro conjunto
print(f'3: {nums}') # observe que o 4 não fica repetido
nums.update([7,8,9]) # atualizando o conjunto com os dados de uma lista
print(f'4: {nums}') # observe que o 9 não fica repetido
````

````python
nums.remove(6)
print(f'5: {nums}')

# Retire o comentário abaixo e verifique o que acontece
nums.remove(6) # gera erro pois o 6 não existe
````

````python
nums.discard(10)
print(f'6: {nums}')

nums.discard(10) # não gera erro
print(f'7: {nums}')
````

````python
while len(nums) > 0:
  retirado = nums.pop()
  print(f'Retirado o valor {retirado}')
  print(f'Após o pop: {nums}')
````

## **3. Sets x Listas**




Implementar conjuntos com listas transforma operações que normalmente são instantâneas em processos que exigem "varrer" a lista repetidamente. Vamos exemplificar como seria implementar algumas operações de conjuntos utilizando listas para efeito apenas de comparação.





### **3.1. Implementando operações básicas de conjuntos com listas**

Para que uma lista se comporte como um conjunto, devemos **garantir a unicidade dos elementos**.

**Pertencimento (in)**

Para verificar se um item está no conjunto, o Python percorre a lista do início ao fim até encontrá-lo.

```python
def pertence(lista, elemento):
    for item in lista:
        if item == elemento:
            return True
    return False
```

**União (|)**

A união combina todos os elementos de A e B, mas evita duplicatas.

```python
def uniao(lista_a, lista_b):
    resultado = lista_a[:] # Copia A
    for item in lista_b:
        if not pertence(resultado, item): # O(n) a cada inserção
            resultado.append(item)
    return resultado
```

**Interseção (&)**

Apenas os elementos que estão em ambos os conjuntos.

```python
def intersecao(lista_a, lista_b):
  inter = []
  for a in list_a :
    if pertence(lista_b, a) :
      inter.append(a);
  return inter
```


### **3.2 Comparando a complexidade de conjuntos com listas**

A grande diferença entre usar set e list reside na eficiência. Em um set real, a busca é, em média, _O(1)_. Com listas, a busca é _O(n)_.

Considerando duas listas ou dois conjuntos com tamanhos **n** e **m**, teríamos as seguintes complexidades para as principais operações:

| Operação | Implementação com Lista | Implementação com Set (Nativo)
| :--- | :--- | :--- |
| Busca (**in**) | O(n)      | O(1) |
| União          | O(n x m)  | O(n + m) |
| Intersecção    | O(n x m ) | O(min(n,m)) |
| Remoção        | O(n)      | O(1) |

Utilizar listas para operações de conjunto é extremamente ineficiente para grandes volumes de dados. Se você tem duas listas de 10.000 itens, a **interseção via listas** pode exigir até 100.000.000 de comparações, enquanto o set faria algo próximo a 10.000 operações. Isso ocorre porque o set utiliza o hash do objeto para encontrar sua posição na memória quase instantaneamente.


**Complexidade de Codificação**

-	Listas: Exigem loops aninhados (for dentro de for) ou verificações manuais de duplicidade. Isso aumenta a chance de erros (bugs) e torna o código mais "sujo".
-	Sets: A sintaxe é declarativa. Operações como set_a & set_b são legíveis e seguem a notação matemática.
