# Decomposição em subproblemas

A **Decomposição em Subproblemas**, também conhecido como **Decomposição Top-Down** ou **Decomposição Funcional**, é um método usado na Engenharia de Software para lidar com a complexidade dos problemas. Decomposição é o processo de separar um problema grande/complexo em subproblemas menores, mais acessíveis.

A Decomposição em Subproblemas atua por etapas, em conjunto com abstração. Para abordar um problema inicial, assume-se que há soluções para os subproblemas, ou seja, abstraem-se os detalhes de como se resolvem os subproblemas, deixando-os para próximas etapas. A solução então é construída compondo as soluções supostas dos subproblemas. Desta maneira, a composição foca somente em partes relevantes, abstraindo detalhes que são deixados para depois.

Em particular, em programação imperativa, as ferramentas fundamentais que permitem definir abstrações de expressões e comandos são as funções e procedimentos. Assim, na abordagem top-down, a suposição da existência a soluções aos subproblemas normalmente é uma suposição de existência de algumas funções, uma para cada subproblema.

Como exemplo, considere que deseja-se vender ingressos para entrar no Cinema. Para vender um ingresso é preciso verificar se a idade do cliente permite assistir ao filme, calcular o preço e imprimir o ingresso.
Uma pessoa pode assistir o filme se a sua idade é maior ou igual à classificação do mesmo. Adicionalmente, por lei, existe a meia-entrada.

Seguindo o método da decomposição em subproblemas, a primeira etapa é escrever o programa principal ou uma função que resolve o problema abstraindo detalhes de subproblemas. Neste caso, os subproblemas serão:
+ Verificar se a idade permite assistir o filme
+ Calcular o preço do ingresso.
+ Imprimir o ingresso.

Assumindo que existe uma função para cada subproblema com nomes `verificaIdade`, `calculaPreco` e `imprimirIngresso`, a solução para o problema é a função a seguir:

```python
def venderIngresso(idadeCliente, temCarteira, classificacaoFilme) :
  if verificaIdade(idadeCliente, classificacaoFilme) :
    p = calculaPreco(temCarteira)
    imprimirIngresso(p)
```

Ao assumir a existência das três funções estamos *abstraindo* os detalhes das suas definições, as quais não são relevantes no momento. Certamente, precisaremos dar as definições de `verificaIdade`, `calculaPreco` e `imprimirIngresso`, mas por enquanto não nos preocupamos por elas, apenas assumimos que existem.

O próximo passo é enfrentar a solução dos subproblemas, um de cada vez. Se o subproblema da vez for simples o suficiente da-se uma solução direta. Porém, se ele resultar complexo, aplica-se nele o mesmo método de decomposição e abstração.

# Um Assunto por Vez - Separação de assuntos

Programadores iniciantes frequentemente tentam codificar situações envolvendo várias condições e ações utilizando uma única construção linguística com partes complexas, tal como um if com ifs dentro de cada caso (ifs aninhados). Por exemplo, suponha que para calcular a conta de energia elétrica tem-se como regras que:
1) Os dados para calcular a conta são a quantidade de kilowatts hora (kwh) e a tarifa de um kwh.
2) O preço mínimo de qualquer conta é R$ 35; e
3) Quando o consumo for maior que 300KWh será cobrada uma taxa extra de 10% no valor da conta.

Uma implementação que trata com ambas as regras ao mesmo tempo utilizando ifs aninhados é:

```python
def contaEnergia(kwh, tarifa) :
   if kwh > 300 :
      conta = kwh * tarifa * 1.1
      if conta < 35 :
         conta = 35
   else :
      conta = kwh * tarifa
      if conta < 35 :
         conta = 35
  return conta
```

Em geral, `if`s aninhados resultam em programas difíceis de entender e alterar, e deveriam ser evitados na medida do possível.

Um dos princípios da Engenharia de Software é a *Separação de Assuntos*. Assim, recomenda-se que, quando podemos identificar assuntos que podem ser tratados independentemente ou sequencialmente, o código deve considerar um assunto por vez.

No exemplo anterior, podemos projetar uma solução mais simples se separamos as regras 2) e 3) para serem aplicadas sequencialmente. O programa pode realizar em primeiro lugar o cálculo correspondente à regra 3) e só depois, em sequência, o cálculo relativo à regra 2), assim:

```python
def contaEnergia(kwh, tarifa) :
   if kwh > 300 :
      conta = kwh * tarifa * 1.1
   else :
      conta = kwh * tarifa

   if conta < 35 :
      conta = 35

   return conta
```

Podemos melhorar ainda um pouco mais, obtendo uma solução mais clara e simples, quando tratamos também em separado o cálculo normal da conta e o reajuste caso o consumo exija os 10% extras, assim:

```python
def contaEnergia(kwh, tarifa) :
   # Calcula a tarifa base, depois aplica a taxa extra (se aplicável), e só no fim garante o limite mínimo
   conta = kwh * tarifa

   if kwh > 300 :
      conta *= 1.1

   if conta < 35 :
      conta = 35

   return conta
```

# Padrões algorítmicos

Um **padrão algorítmico** é um método, estratégia ou técnica de solução de problemas. Um padrão não resolve um problema específico, mas pode ser aplicado para resolver problemas que satisfazem certas características. Por outro lado, a escolha de um padrão conveniente para resolver um problema depende da natureza do problema, estabelecida por como podemos atingir os objetivos do problema. Um padrão tampouco é código pronto, mas sim uma *receita* para resolver algum tipo de problemas.

Em cursos de computação, o foco de estudo de disciplinas de algoritmos são precisamente padrões algorítmicos. Padrões comuns estudados nestas disciplinas são, por exemplo, Dividir para Conquistar, Programação Dinâmica e Método Guloso. Aqui, no contexto de uma disciplina introdutória de Programação, o centro são padrões mais elementares e que normalmente não estão classificados na literatura.

## Seleção de Resultado

Desejamos calcular algo que possui um número fixo relativamente pequeno de possíveis resultados e cada resultado depende de uma condição separada. Para escolher a resposta, podemos usar Seleção Se ou Não, ou Seleção Sequencial. Normalmente, Seleção Sequencial é mais adequado. Em geral, uma solução usando Solução de Resultado terá esta forma:

```python
if condição_1 :
  resultado = res1
elif condição_2 :  
  resultado = res2
...
else :
  resultado = res_n
```

Por exemplo, queremos calcular o maior dentre três números a, b e c. A resposta só pode ser a, b ou c e cada resposta se corresponde com uma condição. Usando Seleção Sequencial podemos esboçar a solução assim:

```python
if a é o maior dos três  :
   maior = a
elif b é o maior dos três  :
   maior = b
else : # c é o maior dos três
  maior = c
```

Este apenas é um esboço da solução, pois as condições estão descritas em português, não em Python. O interpretador Python não poderá executar isto e acusará erros sintáticos. No entanto, esta mistura na escrita de código Python com português, chamada de pseudo-código, é uma ferramenta útil de *abstração* que facilita a construção da solução. Neste caso estamos abstraindo as condições. Por enquanto, não nos preocupamos nos seus detalhes, pois no momento queremos uma solução abstrata.

Para terminar a solução, vamos agora aos detalhes.
```python
if a >= b and a >= c  :
   maior = a
elif b >= a and b >= c  :
   maior = b
else :
  maior = c
```

Esta solução está correta, porém podemos simplificar a segunda condição, aquela associada à resposta `b`, pois ela tem uma parte que é redundante. Quando esta condição for avaliada sabemos que a primeira condição é falsa. Ou seja, sabemos que `a` não pode ser a solução e portanto não precisamos da comparação `b >= a`. Com isto temos a seguinte solução mais eficiente, concisa e clara.

```python
if a >= b and a >= c  :
   maior = a
elif  b >= c  :
   maior = b
else :
  maior = c
```


## Cálculo incremental
Aplica-se quando a resposta pode ser calculada incrementalmente, passo a passo analisando um dado de cada vez. Se consideramos a porção até o momento analisada, a resposta está completa com relação a esta porção. Para o caso geral em que a resposta depende dos dados $a_1, a_2, ..., a_n$, a solução segue o esquema:

+ Calcule a resposta para $a_1$
+ Calcule a resposta para $a_1$ e $a_2$ usando a resposta já obtida para $a_1$.
+ ...   
+ Calcule a resposta para $a_1, a_2, ..., a_n$ usando a resposta para $a_1, a_2, .., a_{n-1}$.

Por exemplo, queremos calcular o maior dentre `a1`, `a2` e `a3`. Usando Cálculo Incremental, colocando a resposta numa variável `maior`, a solução inicial considera somente `a1`, e assim inicializamos maior com `a1` (o maior do conjunto formado somente por `a1` é `a1`). Logo, podemos ir atualizando a variável maior passo a passo à medida que analisamos `a2` e `a3`. Se algum destes valores for maior que `maior`, atualizamos `maior` com o novo valor.



O código python disto é:
```python
maior = a1

if maior < a2 :
   maior = a2

if maior < a3 :
   maior = a3
```

Da forma em que está apresentado aqui, Cálculo Incremental é adequado para quando a sequência $a_1, a_2, ..., a_n$ tem poucos elementos, no máximo três ou quatro. Estudaremos uma apresentação mais geral deste padrão quando abordarmos comandos de repetição.
