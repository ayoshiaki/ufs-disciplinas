# Comandos Condicionais
Comandos condicionais são úteis para escrever programas que atuam de formas diferentes, dependendo se alguma ou algumas condições são satisfeitas.
Comandos condicionais são também conhecidos como comandos de seleção, pois permitem selecionar ações a serem executadas de acordo com condições estabelecidas.

Python apresenta três tipos de comandos de seleção: Seleção Se ou Não, Seleção Alternativa e Seleção Sequencial.



## Seleção Se ou Não (`if`)
É a forma mais simples de seleção. Permite executar uma ação no caso em que uma condição for satisfeita. Ela se aplica em situações em que uma ação é necessária ou desnecessária dependendo de uma condição. Por exemplo, um programa de controle poderia acionar um alarme ou não, dependendo se um sensor detecta a presença de alguém.


```python
...
if presencaDetectada :
  acionarAlarme()
...
```

A ação `acionarAlarme()` será executada se a condição `presencaDetectada` for verdadeira. Já se a condição `presencaDetectada` for falsa, nenhuma ação será realizada.

Observe o uso do caractere `:` após a condição. Este símbolo é obrigatório. Observe também que a ação `acionarAlarme()` a ser executada quando `presencaDetectada` for verdadeira está indentada.

Eventualmente, pode haver uma sequência de ações a ser executada quando a condição é satisfeita. Neste caso, a sequência deve formar um bloco de comandos, todo ele indentado com todas as linhas alinhadas na mesma coluna.

Como um exemplo mais concreto, considere o programa abaixo que calcula o valor total de uma compra considerando que há desconto quando se tem um cupom de desconto.

````python
# Teste o script quando há desconto e quando não há desconto.
unidades = 2
preco = 10.5
temCupomDesconto = True  # altere para False quando não há desconto.

total = unidades * preco
if temCupomDesconto :
  total = total*0.9

print("O total é: ", total)
````

## Seleção Alternativa (`if`/`else`)
Uma outra forma de seleção aplica-se quando há duas possibilidades de ações e quem determina qual é escolhida é a condição. A sintaxe neste caso é


```python
if condição :
  bloco_de_comandos_1
else :
  bloco_de_comandos_2
```

Neste caso, se `condição` for `True`, será executado `bloco_de_comandos_1`. Já se `condição` for `False`, será executado `bloco_de_comandos_2`.

Como na Seleção Se ou Não, a sintaxe requer o caractere `:` após a `condição` do `if`. Observe também que o `:` também deve vir logo após a palavra reservada `else`. Finalmente, a indentação das alternativas é obrigatória.

---
**Terminologia**: Numa Seleção Alternativa, as alternativas são chamadas de **ramos**.

---

A título de ilustração, considere o script a seguir que, seguindo as regras de aprovação de disciplinas na UFS, imprime "Aprovado" ou "Reprovado" dependendo se a nota do aluno for maior ou igual a cinco.

````python
# teste com diversos valores de nota
nota = 5.7

if nota >= 5 :
  print("Aprovado")
else :
  print("Reprovado")
````

---
**Antipadrão**: Utilizar duas Seleções Se ou Não quando Seleção Alternativa é mais adequado.

O exemplo anterior que imprime "Aprovado" ou "Reprovado" poderia ser escrito usando duas Seleções Se ou Não, assim

````python
# teste com diversos valores de nota
nota = 5.7

if nota >= 5 :
  print("Aprovado")
if nota < 5 :
  print("Reprovado")
````

Neste caso, foi necessário escrever a negação da condição para ser aprovado, `nota < 5`. Apesar do programa estar correto, este gastará mais tempo de cômputo, pois duas condições terão que ser avaliadas. Pior ainda, se no futuro a condição de aprovação mudar será necessário alterar as duas condições, o que é mais suscetível a erros.

Ainda, nem sempre o antipadrão produz código equivalente com o original. Isto pode acontecer, por exemplo, quando a condição realiza algum efeito colateral como alterar alguma variável ou fazer alguma operação de Entrada/Saída.

## Seleção Sequencial (`if`/`elif`/`else`)
Existem casos em que é necessário selecionar uma dentre várias possíveis ações e a escolha não depende de uma condição única, mas sim de condições separadas que devem ser verificadas sequencialmente. Por exemplo, a situação de aprovação em disciplinas na UFPE é conceituada como "Aprovado", "Reprovado" ou "Vai para a Final", dependendo do valor da nota. Dada a nota, o programa a seguir retorna a situação de um aluno.



````python
# Teste com diversos valores de nota
nota = 8.3

if nota >= 7 :
   print("Aprovado")
elif nota >= 5 :
   print("Vai para a final")
else :
   print("Reprovado")
````

O programa escolhe imprimir uma das três possibilidades. Para isto, primeiro testa se `nota >= 7`. Se for verdadeiro escolhe imprimir "Aprovado". Contudo, se for falso, testa se `nota >=5` para escolher se imprime "Vai para a final" ou "Reprovado".

Vale salientar que a estrutura de uma Seleção Sequencial é fácil de entender, pois permite identificar rapidamente as ações alternativas junto com suas respectivas condições.

Cada ação é **guardada** por sua própria condição. As condições são avaliadas sequencialmente até encontrar a primeira que é verdadeira para então executar a sua ação correspondente.

Condições posteriores devem ser simplificadas usando o fato que as condições anteriores são falsas. No código anterior, por exemplo, para o caso em que a nota está entre 5 e 7 somente é necessário verificar se `nota >= 5`, pois no momento da sua avaliação já se sabe que `nota >= 7` é falso, ou seja, sabe-se que `nota < 7`.

+ Use Seleção Se ou Não quando tem como opções realizar ou não uma ação.

## Condicionais aninhados
Frequentemente, programas requerem tratar múltiplas condições. Uma maneira de lidar com isto é usando **comandos condicionais aninhados**, ou seja, comandos condicionais dentro de outro comando condicional. Por exemplo, seguindo o exemplo do desempenho em uma disciplina, considere agora que a aprovação em uma disciplina depende não só da nota, mas também da frequência às aulas e que as opções são: "Aprovado", "Reprovado por falta", "Reprovado por nota" e "Reprovado por nota e falta". O seguinte script imprime a situação do aluno.

````python
nota = 7.5
frequencia = 80  # em porcentagem

if nota >= 6:
    if frequencia >= 75:
        print("Aprovado")
    else:
        print("Reprovado por falta")
else:
    if frequencia >= 75:
        print("Reprovado por nota")
    else:
        print("Reprovado por nota e falta")
````

Neste script temos um comando condicional externo, com condição `nota >= 6`, como opções de seleção outros comandos também condicionais, ambos com condição `frequência >= 75`.

---
**Cuidado:**
Apesar de não sempre, comandos condicionais aninhados podem ser um problema. Normalmente, a situação se complica quando há vários níveis de aninhamento, pois nestes casos os programas frequentemente apresentam as seguintes deficiências.
+ **Complexidade de Leitura**: Muitos níveis de aninhamento tornam o código difícil de seguir. Quanto maior o aninhamento, menos claro se torna o fluxo do programa. O código torna-se confuso e difícil de seguir mentalmente, aumentando a chance do desenvolvedor perder o contexto. Tradicionalmente estes códigos são chamados de "*código espaguete*".
+ **Manutenção Difícil**: Entender todas as combinações possíveis de verdadeiro/falso em vários níveis aninhados é complicado. Isso dificulta a identificação de onde uma lógica falhou, tornando a depuração uma tarefa trabalhosa. Também, alterar uma lógica de um condicional interno pode quebrar com a lógica externa esperada.
+ **Risco de Erro**: Há mais riscos de erro devido ao incremento da complexidade cognitiva do código.

````python
nota = 7.5
frequencia = 80  # em porcentagem

# Solução explicitando todas as quatro possibilidades
if nota >= 6 and frequencia >= 75:
    print("Aprovado")
elif nota >=6 and frequencia < 75:
    print("Reprovado por falta")
elif nota < 6 and frequencia >= 75:
    print("Reprovado por nota")
else:
    print("Reprovado por nota e falta")

# Crítica: condições desnecessárias. Podemos simplificar

# Simplificando as condições

if nota >= 6 and frequencia >= 75:
    print("Aprovado")
elif nota >= 6 :                  # Como a condição anterior é falsa, se nota >=6 teremos obrigatoriamente que frequencia < 75
    print("Reprovado por falta")
elif frequencia >= 75 :           # Como a condição anterior é falsa, sabemos então que nota < 6
    print("Reprovado por nota")
else :
    print("Reprovado por nota e falta")

# + Solução mais eficiente, pois as condições são mais simples
# + Programa mais curto
# - Talvez um pouco mais díficil de entender -- pode compensar acrescentando comentários
````

## Simplificar ou não simplificar condições numa seleção sequencial?

Na perspectiva da Engenharia de Software queremos que os programas sejam legíveis, fáceis de entender, fáceis de fazer manutenção e eficientes, dentre outras qualidades.

Certamente a simplificação das condições poderá fazer os condicionais mais eficientes, pois expressões booleanas originalmente redundantes não precisarão ser avaliadas. Mas quanto a legibilidade e facilidade de entendimento, o que podemos dizer?

Como visto anteriormente, numa seleção sequencial o código é avaliado de cima para baixo. Se o fluxo de execução chegou em um elif, é uma garantia absoluta e lógica de que a condição anterior era falsa. Portanto, não precisamos, e não devemos, testar novamente o que a própria estrutura já descartou.

Para entender por que as condições simplificadas ganham tanto em legibilidade quanto em manutenção, podemos analisar os dois aspectos:

+ **Manutenção de código**: Esse é o principal motivo. Imagine que você usou uma condição redundante como `elif nota < 7 and nota >= 5:`. Se no semestre seguinte a regra da escola mudar e a aprovação direta passar a ser com nota 6, você terá que alterar o código em dois lugares: no limite do primeiro `if` e no limite do `elif`. Se esquecer de atualizar a redundância, seu código terá um bug (notas entre 6 e 7 cairão num "limbo" ou darão erro). Na condição simplificada, você altera apenas o topo da cascata e o resto se adapta automaticamente.

+ **Legibilidade e clareza**: Embora iniciantes achem que escrever o limite inferior e superior (ex: "se for menor que 7 E maior que 5") deixa o texto mais claro, na prática de desenvolvimento isso cria uma poluição desnecessária. Profissionais leem o if/elif como uma cascata natural. Menos código escrito significa menos carga cognitiva para quem está lendo.

Resumindo, a redundância cria a falsa sensação de segurança na leitura, mas cobra um preço alto na hora de atualizar o sistema. É muito melhor treinar desde cedo a simplificação de condições e confiar na estrutura em cascata do `if`/`elif`/`else`.

## **Seção opcional:** Seleção por Casamento de Padrões (`match`/`case`)

Desde a versão 3.10, Python oferece outra forma de seleção condicional denominada **casamento de padrões**. O mecanismo geral é bastante poderoso. Em particular ele se aplica bem quando a seleção é feita pela escolha dentre alguns valores fixos. Por exemplo, a seguir definimos uma função que recebe um string que representa um mês e retorna a quantidade de dias que o mês tem. Nesta, a escolha do que imprimir é feita usando casamento de padrões.

````python
def diasDoMes(mes) :
  match mes:
    case "abril" | "junho" | "setembro" | "novembro":
      print("30 dias")
    case "fevereiro":
      print("28 ou 29 dias")
    case "janeiro" | "março" | "maio" | "julho" | "agosto":
      print("31 dias")
    case _ :
      print(f"mês {mes} inválido")

diasDoMes("novembro")
diasDoMes("março")
diasDoMes("fevereiro")
diasDoMes("xyz")
````

Para um estudo mais aprofundado de casamento de padrões, veja, por exemplo, https://realpython.com/structural-pattern-matching/.

### Exercício
Uma leve variante deste exercício está disponível em https://www.thehuxley.com/problem/3897.

**Descrição**

Considere as expressões abaixo para cálculo das médias aritmética (A), harmônica (H) e geométrica (G).

Média aritmética: $A = \frac{a+b+c}{3}$

Média harmônica: $H = \frac{3}{\frac{1}{a}+\frac{1}{b}+\frac{1}{c}}$

Média geométrica $G = (abc)^{\frac{1}{3}}$

Elabore um programa que contenha três funções, cada uma para calcular uma das médias descritas acima. Todas as funções receberão como parâmetros os valores de $a$, $b$ e $c$.


**Formato de entrada**

Um caracter representando o tipo de média (A - aritmética, H - harmônica, G - geométrica) e três números inteiros. Cada um dos dados é inserido em uma linha.

**Formato de saída**

A média correspondente calculada com três casas decimais.

````python
# Solução com seleção sequencial

# Este é um bom exemplo para mostrar que pensamento bottom-up também é útil
# Pelo enunciado, é claro que temos três fórmulas, cada uma para um conceito
# específico e bem estabelecido. Então, de cara, é interessante definir uma
# função para cada conceito.
def mediaAritmetica(a, b, c) :
  return (a+b+c)/3

def mediaHarmonica(a, b, c) :
  return 1/(1/a + 1/b + 1/c)

def mediaGeometrica(a, b, c) :
  return (a*b*c)**1/3

tipoMedia = input()
a = int(input())
b = int(input())
c = int(input())
if tipoMedia == "A" :
  print(f"{mediaAritmetica(a, b, c):.3f}")
elif tipoMedia == "H" :
  print(f"{mediaHarmonica(a, b, c):.3f}")
else :
  print(f"{mediaGeometrica(a, b, c):.3f}")
````

````python
# Solução com casamento de padrões

def mediaAritmetica(a, b, c) :
  return (a+b+c)/3

def mediaHarmonica(a, b, c) :
  return 1/(1/a + 1/b + 1/c)

def mediaGeometrica(a, b, c) :
  return (a*b*c)**1/3

tipoMedia = input()
a = int(input())
b = int(input())
c = int(input())
match tipoMedia :
  case "A" :  print(f"{mediaAritmetica(a, b, c):.3f}")
  case "H" :  print(f"{mediaHarmonica(a, b, c):.3f}")
  _        :  print(f"{mediaGeometrica(a, b, c):.3f}")
````
