## SeleĂ§ĂŁo Sequencial (`if`/`elif`/`else`)
Existem casos em que ĂŠ necessĂĄrio selecionar uma dentre vĂĄrias possĂ­veis aĂ§Ăľes e a escolha nĂŁo depende de uma condiĂ§ĂŁo Ăşnica, mas sim de condiĂ§Ăľes separadas que devem ser verificadas sequencialmente. Por exemplo, a situaĂ§ĂŁo de aprovaĂ§ĂŁo em disciplinas na UFPE ĂŠ conceituada como "Aprovado", "Reprovado" ou "Vai para a Final", dependendo do valor da nota. Dada a nota, o programa a seguir retorna a situaĂ§ĂŁo de um aluno.



````python
# Teste com diversos valores de nota
nota = 8.3

if nota >= 7 :
   print("Aprovado")
elif nota >= 5 :
   print("Vai para a final")
else :
   print("Reprovado")
````

O programa escolhe imprimir uma das trĂŞs possibilidades. Para isto, primeiro testa se `nota >= 7`. Se for verdadeiro escolhe imprimir "Aprovado". Contudo, se for falso, testa se `nota >=5` para escolher se imprime "Vai para a final" ou "Reprovado".

Vale salientar que a estrutura de uma SeleĂ§ĂŁo Sequencial ĂŠ fĂĄcil de entender, pois permite identificar rapidamente as aĂ§Ăľes alternativas junto com suas respectivas condiĂ§Ăľes.

Cada aĂ§ĂŁo ĂŠ **guardada** por sua prĂłpria condiĂ§ĂŁo. As condiĂ§Ăľes sĂŁo avaliadas sequencialmente atĂŠ encontrar a primeira que ĂŠ verdadeira para entĂŁo executar a sua aĂ§ĂŁo correspondente.

CondiĂ§Ăľes posteriores devem ser simplificadas usando o fato que as condiĂ§Ăľes anteriores sĂŁo falsas. No cĂłdigo anterior, por exemplo, para o caso em que a nota estĂĄ entre 5 e 7 somente ĂŠ necessĂĄrio verificar se `nota >= 5`, pois no momento da sua avaliaĂ§ĂŁo jĂĄ se sabe que `nota >= 7` ĂŠ falso, ou seja, sabe-se que `nota < 7`.

## Condicionais aninhados
Frequentemente, programas requerem tratar mĂşltiplas condiĂ§Ăľes. Uma maneira de lidar com isto ĂŠ usando **comandos condicionais aninhados**, ou seja, comandos condicionais dentro de outro comando condicional. Por exemplo, seguindo o exemplo do desempenho em uma disciplina, considere agora que a aprovaĂ§ĂŁo em uma disciplina depende nĂŁo sĂł da nota, mas tambĂŠm da frequĂŞncia Ă s aulas e que as opĂ§Ăľes sĂŁo: "Aprovado", "Reprovado por falta", "Reprovado por nota" e "Reprovado por nota e falta". O seguinte script imprime a situaĂ§ĂŁo do aluno.

````python
nota = 7.5
frequencia = 80  # em porcentagem

if nota >= 6:
    if frequencia >= 75:
        print("Aprovado")
    else:
        print("Reprovado por falta")
else:
    if frequencia >= 75:
        print("Reprovado por nota")
    else:
        print("Reprovado por nota e falta")
````

Neste script temos um comando condicional externo, com condiĂ§ĂŁo `nota >= 6`, como opĂ§Ăľes de seleĂ§ĂŁo outros comandos tambĂŠm condicionais, ambos com condiĂ§ĂŁo `frequĂŞncia >= 75`.

---
**Cuidado:**
Apesar de nĂŁo sempre, comandos condicionais aninhados podem ser um problema. Normalmente, a situaĂ§ĂŁo se complica quando hĂĄ vĂĄrios nĂ­veis de aninhamento, pois nestes casos os programas frequentemente apresentam as seguintes deficiĂŞncias.
+ **Complexidade de Leitura**: Muitos nĂ­veis de aninhamento tornam o cĂłdigo difĂ­cil de seguir. Quanto maior o aninhamento, menos claro se torna o fluxo do programa. O cĂłdigo torna-se confuso e difĂ­cil de seguir mentalmente, aumentando a chance do desenvolvedor perder o contexto. Tradicionalmente estes cĂłdigos sĂŁo chamados de "*cĂłdigo espaguete*".
+ **ManutenĂ§ĂŁo DifĂ­cil**: Entender todas as combinaĂ§Ăľes possĂ­veis de verdadeiro/falso em vĂĄrios nĂ­veis aninhados ĂŠ complicado. Isso dificulta a identificaĂ§ĂŁo de onde uma lĂłgica falhou, tornando a depuraĂ§ĂŁo uma tarefa trabalhosa. TambĂŠm, alterar uma lĂłgica de um condicional interno pode quebrar com a lĂłgica externa esperada.
+ **Risco de Erro**: HĂĄ mais riscos de erro devido ao incremento da complexidade cognitiva do cĂłdigo.

## **SeĂ§ĂŁo opcional:** SeleĂ§ĂŁo por Casamento de PadrĂľes (`match`/`case`)

Desde a versĂŁo 3.10, Python oferece outra forma de seleĂ§ĂŁo condicional denominada **casamento de padrĂľes**. O mecanismo geral ĂŠ bastante poderoso. Em particular ele se aplica bem quando a seleĂ§ĂŁo ĂŠ feita pela escolha dentre alguns valores fixos. Por exemplo, a seguir definimos uma funĂ§ĂŁo que recebe um string que representa um mĂŞs e retorna a quantidade de dias que o mĂŞs. Nesta, a escolha do que imprimir ĂŠ feita usando casamento de padrĂľes.

````python
def diasDoMes(mes) :
  match mes:
    case "abril" | "junho" | "setembro" | "novembro":
      print("30 dias")
    case "fevereiro":
      print("28 ou 29 dias")
    case "janeiro" | "marĂ§o" | "maio" | "julho" | "agosto":
      print("31 dias")
    case _ :
      print(f"mĂŞs {mes} invĂĄlido")

diasDoMes("novembro")
diasDoMes("marĂ§o")
diasDoMes("fevereiro")
diasDoMes("xyz")
````
