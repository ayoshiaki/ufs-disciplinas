Como outro exemplo, escrevamos um programa que leia a altura e o raio da base de um cilindro e que então calcule qual é a área do cilindro.
<!--  Apresentar de forma top down: começar primeiro com o return e depois as linhas para areaLateral e areaBase
      
      Destacar o import caso não tenha sido introduzido antes
-->


```python
from math import pi

def areaCilindro(a, r) :
    areaLateral = 2*pi*r*a
    areaBase = pi*r**2
    return areaLateral + 2*areaBase

altura = float(input())
raio = float(input())
print(areaCilindro(altura, raio))
```

Usando procedimentos, definamos um script que apresenta duas casas.
<!-- Apresentar de forma top-down.
     1) Explicar qual é o resultado que esperamos
     2) Iniciar a explicação do programa pelo script principal
     3) Depois casa e só depois parede e teto.

     Destacar parametrização. Conseguimos genericidade.
     Destacar, no final, que a construção foi feita passo a passo, abstraindo através de funções auxiliares.
-->

```python
def parede(c) :
    print(c*9)
    print(c*9)
    print(c*3 + " "*3 + c*3)
    print(c*3 + " "*3 + c*3)

def teto(c) :
    print(" "*4 + c)
    print(" "*2 + c*5)
    print(c*9)

def casa(p, t) :
    teto(t)
    parede(p)

casa("#", "%")
print()
casa("+", "*")
```

Algumas observações sobre o código anterior.
  + O código foi dividido ou estruturado usando funções.
  + O uso de nomes significativos para as funções é crucial para o entendimento do programa por parte de outros programadores.
  + O programa principal, linhas 16-17, expressa claramente a intenção, apresentar duas casas separada por uma linha em branco.
  + Da mesma forma, a intenção da função `casa` é clara. Primeiro apresenta-se o teto e depois a parede.
  + As funções `parede` e `teto` estão parametrizadas pelo caracter que será usado na apresentação. Isto permite reusar o mesmo código para casas com caracteres diversos para apresentar tanto a parede como o teto.
  + Para alguém que lê o código, o entendimento é modular.

Sobre entendimento modular, a recomendação é iniciar a leitura do programa pelo programa principal. No caso, as linhas 16-17. Se apenas interessa um entendimento de mais alto nível, a leitura pode terminar aqui. Porém, se interessam os detalhes, a leitura pode continuar para a definição de `casa`. A definição de `casa` diz que primeiro apresenta-se o teto e logo depois a parede. Por sua vez, a apresentação de teto e  parede estão encapsuladas em duas funções separadas. Se o leitor quiser mais detalhes, ele pode então partir para a leitura das definições das funções correspondentes, `teto` e `parede`.
Seguindo esta abordagem de leitura, o entendimento completo do programa se dá por camadas, primeiro obtem-se um entendimento geral e abstrato, de alto nível. Depois, podem se entender detalhes, descendo a um nível mais baixo, olhando as respectivas definições.  

A divisão do programa em funções não só permite entender partes do programa por separado, mas também permite testar e depurar peças de código por separado.
