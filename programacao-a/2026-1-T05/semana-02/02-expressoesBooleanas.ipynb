# Booleanos
Python oferece o tipo `bool` para os valores de verdade, tĂ­picos da lĂłgica matemĂĄtica. O tipo tem apenas dois valores: `True`e `False`.
Python tambĂŠm oferece operadores booleanos que se correspondem aos conectivos lĂłgicos da matemĂĄtica.
+ Para a conjunĂ§ĂŁo (e lĂłgico) usa-se `and` como operador binĂĄrio infixo.
+ Para a disjunĂ§ĂŁo (ou lĂłgico) usa-se `or` como operador binĂĄrio infixo.
+ Para a negaĂ§ĂŁo (nĂŁo lĂłgico) usa-se  `not` como operador unĂĄrio prefixo.

Os cĂłdigos a seguir apresentam o resultado destes operadores para todas as combinaĂ§Ăľes possĂ­veis de valores de verdade.


````python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
````

````python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
````

````python
print(not True)
print(not False)
````

## Operadores relacionais
Operadores relacionais retornam como resultado valores booleanos.
Dentre eles temos operadores de comparaĂ§ĂŁo `<`, `<=`, `>`, `>=`, `==` e `!=`.


````python
print(1 <= 3)
print(1 >= 3)
print(1 == 3)
print(1 != 3)
````

## Operadores lĂłgicos sĂŁo preguiĂ§osos: avaliaĂ§ĂŁo de curto-circuito.

Observe que para que `A and B` seja `False`, basta que A seja `False`.
Quando Python avalia que `A and B`, avalia primeiro `A` e se este for `False` termina com resultado final `False`. Observe que `B` nĂŁo ĂŠ avaliado neste caso. Este tipo de avaliaĂ§ĂŁo ĂŠ denominada de **avaliaĂ§ĂŁo de curto-circuito**.

Similarmente, para que `A or B` seja `True`, basta que `A` seja `True` e, neste caso, `B` nĂŁo precisa ser avaliado.

Veremos, mais na frente, que esta caracterĂ­stica preguiĂ§osa dos operadores lĂłgicos tĂŞm algumas sutilezas e que, em alguns casos especiais, a comutatividade do `and` e do `or` nĂŁo ĂŠ vĂĄlida.

## SimplificaĂ§ĂŁo de expressĂľes booleanas

### Prefira afirmaĂ§Ăľes a negaĂ§Ăľes

Em geral, expressĂľes que afirmam algo sĂŁo mais diretas e fĂĄceis de compreender que aquelas que afirmam a sua negaĂ§ĂŁo. Por exemplo, considere as duas seguintes afirmaĂ§Ăľes equivalentes
  + JoĂŁo nĂŁo ĂŠ de maior idade
  + JoĂŁo ĂŠ de menor idade

Qual ĂŠ mais direta? Qual exige menos esforĂ§o mental para entender?

Cada operadores relacional tem outro que se corresponde com a negaĂ§ĂŁo da relaĂ§ĂŁo. Isto permite evitar aplicaĂ§Ăľes do operador `not`.

Assim, por exemplo
| Ao invĂŠs de | Prefira |
| ------------| --------|
| ```not (idade > 18)``` | ```idade <= 18```|
| ```not x == y``` | ```x != y``` |
| ```not x != y``` | ```x == y```|

### NĂŁo compare uma expressĂŁo booleana com `True` ou com `False`

A expressĂŁo booleana `E == True` ĂŠ equivalente com `E`. Obviamente, a opĂ§ĂŁo `E` ĂŠ mais simples.

A expressĂŁo booleana `E == False` ĂŠ equivalente com `not E`. Prefira usar `not E`.

### Utilize propriedades algĂŠbricas da lĂłgica matemĂĄtica para simplificar expressĂľes.

A seguir temos algumas propriedades algĂŠbricas frequentemente Ăşteis. Nelas usamos `<==>` como sĂ­mbolo para a equivalĂŞncia lĂłgica. Este operador nĂŁo existe em Python.

**NegaĂ§ĂŁo dupla**
```python
 not (not A) <==> A
```

**Leis de De Morgan**
```python
not (A and B) <==> not A or not B
```

**Leis Distributivas/FatoraĂ§ĂŁo**
```python
A and (B or C) <==> A and B  or  A and C
A or (B and C) <==> (A or B) and (A or C)
```

Quando a igualdade ĂŠ lida de esquerda para a direita, ambas leis sĂŁo vistas como *distributivas*. A primeira diz que o `and` distribui sobre o `or` enquanto que a segunda diz que o `or` distribui sobre o `and`.

Agora, quando as igualdades sĂŁo lidas de direita para a esquerda, ambas as leis podem ser vistas como um tipo de  *fatoraĂ§ĂŁo algĂŠbrica* ou *colocar em evidĂŞncia*.

**Leis Comutativas**
```python
A and B <==> B and A
A or B <==> B or A
```

**Elemento neutro**
```python
True and A <==> A and True <==> A
False or A <==> A or False <==> A
```

**Elemento absortivo ou anulador**
```python
False and A <==> A and False <==> False
True or A <==> A or True <==> True
```

### Encadeamento de opeadores relacionais

Na matemĂĄtica ĂŠ comum abreviar expressĂľes relacionais usando *encadeamentos*. Por exemplo para exprimir os relacionamentos $x < y$ e $y < z$, na matemĂĄtica abrevia-se assim $x < y < z$. Encadeamentos como estes sĂŁo permitidos somente apenas se todos os operadores "apontam" na mesma direĂ§ĂŁo, pois
como as relaĂ§Ăľes de ordem sĂŁo transitivas, a combinaĂ§ĂŁo tambĂŠm expressa uma relaĂ§ĂŁo entre o primeiro e Ăşltimo valor. Assim por exemplo, se temos $x < y \leq z$, alĂŠm de termos que $x < y$ e $y \leq z$, por transitividade temos tambĂŠm que $x < z$.

Python permite encadeamentos de operadores relacionais. No entanto, Python ĂŠ mais permissivo de tal forma que expressĂľes como `x > y < z` tambĂŠm sĂŁo vĂĄlidas. Neste caso devemos interpretar que `y` ĂŠ menor que `x` e `z` ao mesmo tempo. Em outras palavras, a expressĂŁo ĂŠ equivalente com `x > y and y < z`.
