## Funções matemáticas

Na matemática uma função é uma relação entre um conjunto de entradas e um conjunto de saídas, onde cada entrada está relacionada a exatamente uma saída. Assim, podemos considerar que para cada valor no conjunto de entrada uma função associa ou *mapeia* este valor a um único valor do conjunto de saída.
A **figura** a seguir ilustra uma função cujos conjuntos de entradas e saídas são os números naturais.






A função $f$ na figura anterior é definida matematicamente usando a seguinte notação

$$\begin{split}
     &f:\mathbb{N} \to \mathbb{N}\\
     &f(x) = \lfloor x/2 \rfloor
  \end{split}
$$

A primeira linha declara que o nome da função que está sendo definida é $f$ e que ambos os conjuntos de entradas e saídas são o mesmo conjunto dos números naturais, $\mathbb{N}$. A segunda linha descreve o mapeamento através de uma equação. Ela diz que para qualquer $x \in \mathbb{N}$, $f$ mapeia $x$ para o piso de $x/2$, também um natural.

Em Python podemos definir funções matemáticas, porém utilizando uma notação diferente.
Assim, por exemplo, para a função $f$ anterior temos


````python
def f(x) :
  return x//2
````

Aqui, na primeira linha, a palavra reservada `def` inicia a definição da função. Logo a seguir, em `f(x) :` declara-se que o nome da função é `f` e que recebe uma entrada chamada `x`. Na segunda linha, o comando `return` `x//2` diz que o valor de saída que `f` **associa à** entrada `x` é `x//2`, ou seja, a divisão inteira de `x` por `2`.

Diferente da definição matemática rigorosa, a nossa função em Python não restringe nativamente a entrada apenas a números naturais. O valor de `x` poderia ser, por exemplo, um número decimal (`float`). No entanto, para o nosso propósito de simular a função matemática $f$, assumiremos que o usuário passará um valor inteiro maior ou igual a zero.

### *Boa prática: uso de nomes significativos*

Em matemática, prioriza-se a **concisão**, por isso as funções normalmente tem nomes curtos como $f$, $g$ ou $h$. Em computação prioriza-se legibilidade, e assim o mais comum é usar nomes mais significativos tais como `metade`, `dobro` ou `tarifa_de_luz`. Por exemplo, a função anterior é melhor definida com o nome `metade`, assim

````python
def metade(x) :
  return x//2
````

 Em casos como a definição anterior, nomear as variáveis com nomes curtos como `x`, `y`, `i` ou `j` pode ser adequado. Porém, quando variáveis são usadas em códigos mais longos, também recomenda-se o uso de nomes significativos.

-----


Retornando para a definição matemática de $f$ dada anteriormente, observe que o nome $x$ da variável é um nome arbitrário, escolhido apenas para denotar algum número natural qualquer. Assim, uma definição equivalente, usando outro nome para a variável, é

$$\begin{split}
     &f:\mathbb{N} \to \mathbb{N}\\
     &f(y) = \lfloor y/2 \rfloor
  \end{split}
$$

Similarmente em Python, os nomes das variáveis e parâmetros são escolhidos arbitrariamente pelo programador. Assim, por exemplo, as seguintes definições são equivalentes.

```python
    def metade(k):                          def metade(x) :
      return k//2                             return x//2
```

### *Notação e Terminologia*
O script a seguir define uma função e a usa no programa principal.


````python
def metade(x) :
  return x//2

i = 5
m = metade(12*i)
print(m)
````

Nas linhas 1 e 2 estamos **definindo** ou **declarando** a função `metade`.

A linha 1 é dita **cabeçalho** ou **assinatura** da função. Nela, a variável `x` é chamada de **parâmetro da função**. Observe que no final da linha temos um símbolo `:`. Este símbolo é obrigatório no final do cabeçalho.

A linha 2 é o **corpo** da função. Nela o comando `return` **retorna como saída** ou **retorna como resultado** `x//2`. Observe que o corpo não está alinhado com a coluna de início da palavra `def`, ele inicia alguns espaços à direita. Dizemos que o corpo está **indentado**. A indentação é obrigatória. Se o corpo da função contiver várias linhas, todas devem estar indentadas e alinhadas dentro do corpo.

Na linha 5, a função `metade` está sendo usada para calcular a saída para uma entrada concreta. Dizemos que a função `metade` está sendo **chamada** ou **instanciada**. Em particular, a chamada `metade(12*i)` é feita com **argumento** `12*i`.


---
**Cuidado**:
 + Misturar espaçõs e caracteres de tabulação para indentar pode causar erros. A dica é sempre usar espaços, nunca tabulação.
 + Não esqueça do caractere `:` ao final da linha do cabeçalho de uma função.

## Procedimentos

Funções Pyhton não precisam ser puramente matemáticas. De fato, no contexto da programação, uma função é um bloco de comandos com um nome definido e que realiza um cômputo. Assim, além de possivelmente retornar algum valor da saída, uma função também pode encapsular ações tais como alterar o conteúdo de variáveis e realizar entrada e saída. Em programação, estas funções são denominadas de **procedimentos**.

Em capítulo anterior vimos algumas funções predefinidas de Python. Algumas como `int()` e `float()` são puras, ou seja, matemáticas. Outras, como `input()` e `print()`, realizam ações. Também temos visto que podemos definir nossas próprias funções matemáticas, como `metade`.  Mas também podemos definir nossos próprios procedimentos. Por exemplo, a seguinte função permite escrever na tela uma sequência de caracteres `"-"`.

````python
def imprimeHorizontal(n):
    print(n*"-")

imprimeHorizontal(3)
imprimeHorizontal(6)
imprimeHorizontal(9)
````

Dividir o código de um programa em funções tem consequências importantes desde o ponto de vista de engenharia de software. Programas se tornam estruturados, legíveis e mais fáceis de serem testados e depurados. Não menos importante, evita-se repetição de código.

Seguindo com o exemplo anterior, a seguir temos um programa que imprime dois retângulos.

````python
def imprimeHorizontal(n):
    print(n*"-")

def imprime2BarrasVerticais(n) :
    print("|" + (n-2)*" " + "|")

def imprimeRetangulo() :
    imprimeHorizontal(30)
    imprime2BarrasVerticais(30)
    imprime2BarrasVerticais(30)
    imprime2BarrasVerticais(30)
    imprime2BarrasVerticais(30)
    imprimeHorizontal(30)

imprimeRetangulo()
print()
imprimeRetangulo()
````

Observe a indentação na definição da função `imprimeRetangulo`. Neste caso, o corpo da definição contém várias linhas, todas indentadas com o mesmo alinhamento. Grupos de comandos indentados desta forma são denominados **blocos de comandos**.




## Escopo

Dentro de um programa, o nome de uma variável ou função passa por dois momentos: a sua **definição** (criação) e o seu **uso** (chamada). No caso de uma função, a definição acontece logo após a palavra reservada `def` enquanto que os usos acontecem nas chamadas à função. Por exemplo, no programa anterior que imprime um retângulo, `imprimeRetangulo` na linha 4 é uma definição, enquanto que a chamada na linha 12, é um uso.

No caso de variáveis, elas são definidas quando acontece a primeira atribuição e são usadas após. Assim, por exemplo, considerando o seguinte trecho de código

````python
i = 5
i += 1
j = 12*i
print(i, j)
````

a variável `i` está sendo definida na primeira linha e usada em todas as seguintes.

Com relação a parâmetros de uma função, a presença de um parâmetro no cabeçalho da função é uma definição, enquanto que as outras chamadas, no corpo da função, são de uso. Assim por exemplo, na seguinte definição

````python
def hipotenusa(a, b) :
    return (a**2 + b**2)**0.5
````

as presenças de `a` e `b` na linha 1 são definições, enquanto que as da linha 2 são usos.

A distinção entre definição e uso dentro de um programa é fundamental para entender o conceito de escopo. O trecho do código onde uma definição é válida e pode ser acessada é chamado de **escopo**. Em capítulos posteriores explicaremos mais cuidadosamente as regras de escopo de Python. No momento, basta termos algumas noções gerais.



+ Definições de funções em Python tem como escopo o bloco em que estão sendo definidas.


+ O escopo de um parâmetro de uma função é o corpo da definição da função.

+ Variáveis também têm escopo. Uma variável é definida na primeira vez que é atribuída e o seu escopo é o bloco de código em que se encontra.



Para ilustrar escopos e introduzir algumas terminologias, considere o seguinte código


````python
x = 10

def f(x, y) :
    m = (x+y)/2
    k = 5
    return k+m

def g(y) :
    return f(x, y)

print(g(10)+x)
````

A variável `x` definida na linha 1 têm como escopo o programa completo. Variáveis com o programa completo como escopo são ditas **globais**. Esta variável é usada no corpo da função `g`, na linha 9, assim como também na linha 11.

O parâmetro `x` da função `f`, na linha 3, têm como escopo o corpo da função `f`. Observe que neste caso, a definição do parâmetro se **sobrepõe** à definição da variável global `x` e, consequentemente, a variável `x` da linha 4 se referencia a este parâmetro, não à variável global.

Similarmente, o parâmetro `y` de `f` tem como escopo o corpo da função. Logo, a variável `y` mencionada na linha 4 se refere a este parâmetro.

O escopo das variáveis `m` e `k`, definidas no corpo de `f`, linhas 4 e 5, respectivamente, têm como escopo o corpo completo de `f`. Em casos assim, dizemos que `m` e `k` são **variáveis locais** a `f`.

A função `g` tem como parâmetro `y` com escopo o corpo de `g`.

Quanto ao escopo das funções `f` e `g`, ambos são globais, podendo ambas serem chamadas em qualquer parte do programa.

## Tempo de vida de parâmetros e variáveis

O tempo de vida de uma variável é um conceito relacionado com a execução do programa. Variáveis globais têm como tempo de vida a duração completa do programa. Variáveis locais e parâmetros de uma função têm como tempo de vida a duração da execução da chamada à função. Quando uma função é chamada, criam-se os parâmetros e variáveis locais, os quais persistem **até** o final da execução da chamada à função.

Em particular, com relação aos parâmetros, quando uma função é chamada
  + criam-se as variáveis correspondentes aos parâmetros,
  + avaliam-se os argumentos com os quais estão sendo chamada a função e atribuem-se os valores resultantes aos respectivos parâmetros,
  + executa-se o corpo da função e, finalmente,
  + eliminam-se da memória as variáveis relativas aos parâmetros.

Por exemplo, voltemos ao seguinte programa, já visto acima, para ilustrar o tempo de vida de algumas variáveis

````python
def metade(x) :
  return x//2

i = 5
m = metade(12*i)
print(m)
````

1) A variável global `i` tem como tempo de vida a execução do programa inteiro.
2) Quando o programa executa a linha 5, a chamada a `metade` cria uma variável `x`, parâmetro da função, e a inicializa com o valor de `12*i`, no caso 60. Logo então, executa-se o corpo de `metade` para retornar 30, a chamada a função termina e, como consequência, a variável `x` deixa de **existir**.
3) O valor retornado pela chamada a `metade` é atribuído a `m`, criada neste momento e com tempo de vida o que resta da execução do programa.

## Por que dividir um programa em funções
Em programação, uma função é um bloco de código reutilizável que realiza uma tarefa específica. As funções permitem que você organize seu código em partes menores e mais gerenciáveis, facilitando a leitura, manutenção e reutilização do código. Dentre os benefícios de estruturar um programa com funções temos:

* **Facilita a leitura e o entendimento:** O código se torna mais modular e intuitivo.
* **Facilita o teste e a depuração:** É mais fácil encontrar e corrigir erros em blocos isolados do que em um script gigante.
* **Evita a repetição de código:** Você escreve a definição uma vez e a utiliza quantas vezes precisar.
* **Permite a abstração e decomposição:** Ajuda a quebrar um problema complexo em partes menores e mais simples de resolver.

## Bibliotecas

Além das funções predefinidas, tais como `int()`, `float()`, `print()` e outras, Python oferece um repertório amplo de funções para diversos propósitos. Estas funções, junto com definições de outros elementos, são providas em **bibliotecas** especializadas. Algumas oficiais, como a biblioteca `math`, e outras que apesar de não serem oficiais são muito populares.

Para usar elementos de uma biblioteca, é preciso usar o comando `import` seguido do nome da biblioteca. Por exemplo, para usar a constante `pi`, é preciso importar a biblioteca `math`. Junto com a constante `pi`, há várias funções matemáticas. Algumas delas são `sin`, `cos`, `tan`, `asin`, `acos`, `atan, `log`, `log2`, `log10`, `pow`, `sqrt`, `floor` e `ceil`. A seguir estão alguns usos destas funções. Experimente usar outras e descubra o que elas fazem.

````python
import math

print(math.pi)

x = math.pow(3,4)
print(x, math.sqrt(x))

print(math.floor(2.8))

print(math.ceil(2.8))
````

Observe que, no exemplo anterior, para usar elementos da biblioteca `math` precisamos prefixar os nomes com `math.`. Assim, por exemplo, para nos referir à constante `pi` da bilblioteca `math`, usamos `math.pi`. Esta forma de nomeação é dita *qualificada* com o nome da biblioteca e é a mais recomendada, pois evita colisão de nomes iguais que, por ventura, sejam definidos em várias bibliotecas e/ou no programa atual.

Outra forma de importar é explicitando os elementos que estão sendo importados, utilizando a palavra reservada `from`. Neste caso, os elementos da biblioteca não precisam ser qualificados pelo nome do módulo. O código a seguir ilustra que podemos referenciar `pi` e `sqrt` da bilbioteca `math` sem necessidade de qualificá-los.


````python
from math import sqrt, pi

def distance(x1, y1, x2, y2):
  return sqrt( (x2-x1)**2 + (y2-y1)**2 )

print(distance(1, 2, 4, 6))
print(pi)
````
