# **Busca Binária**

## **1. Motivação**

Pense como seria para encontrar um valor em uma lista. Uma possibilidade é procurar partindo da posição 0 até chegar ao final da lista. Esta é uma estratégia bastante simples e funciona.

Se você fosse implementar em Python, poderia adotar um algoritmo semelhante ao mostrado abaixo:


```python
def buscaSequencial(lista, alvo):
    """
    Percorre a lista e retorna o índice do elemento alvo.
    Caso não encontre, retorna -1.
    """
    for i in range(len(lista)):
        if lista[i] == alvo:
            return i  # Elemento encontrado, retorna o índice

    return -1  # Elemento não encontrado


nomes = ['Pedro', 'Maria', 'Ana', 'Flavia', 'Carlos']
nome = input("Digite o nome a ser buscado: ")
while nome.strip() != "":
    pos = buscaSequencial(nomes, nome)
    if pos >= 0:
       print(f'O nome {nome} está na posição {pos}')
    else:
       print(f'Nome {nome} não encontrado')
    nome = input("Digite o nome a ser buscado (ou ENTER para finalizar): ")
```

Python oferece um método predefinido na classe **list** chamado **index** que possui um comportamento semelhante à função **buscaSequencial**, mas ao invés de retornar o valor -1 quando não encontra o valor buscado, lança uma exceção do tipo **ValueError**.

##**2. Busca Binária**


Como você faz para procurar uma palavra em um dicionário? Obviamente você não começa a fazer a busca pela primeira página e segue fazendo a busca na sequência...

Provavelmente você abre o dicionário em uma página e vê a primeira palavra que aparece. Se a palavra que você está buscando vem antes dela na ordem alfabética, concentra a busca no lado esquerdo do dicionário, ou seja, nas páginas anteriores. Caso contrário, faz a busca do lado direito. Este processo vai sendo repetido gradativamente até chegar na parte certa.

Observe que neste algoritmo a cada análise você está eliminando metade dos valores, fazendo com que o algoritmo seja bastante eficiente.




```python
def buscaBinaria(lista, alvo):
    inicio = 0
    fim = len(lista) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2  # Encontra o índice do meio
        chute = lista[meio]

        if chute == alvo:
            return meio  # Encontrou o alvo!

        if chute > alvo:
            fim = meio - 1  # O alvo está na metade da esquerda
        else:
            inicio = meio + 1 # O alvo está na metade da direita

    return -1  # O elemento não existe na lista


nums = [1, 3, 5, 7, 8, 9, 10, 15, 20, 22, 25, 30]
print(f"Posição do 3: {buscaBinaria(nums, 3)}")
print(f"Posição do 22: {buscaBinaria(nums, 22)}")
print(f"Posição do 16: {buscaBinaria(nums, 16)}")
```

### **2.3. Análise de Desempenho**

A diferença de eficiência torna-se exponencial conforme o número de elementos ($n$) aumenta:

**Busca Sequencial ($O(n)$):** O algoritmo percorre item por item do início ao fim. Se você tem 1 milhão de itens, no pior caso, fará 1 milhão de comparações. O tempo de execução cresce de forma linear com o tamanho da lista.

**Busca Binária ($O(\log n)$):** O algoritmo divide o espaço de busca pela metade a cada passo. Para os mesmos 1 milhão de itens, a busca binária encontrará o elemento em, no máximo, 20 comparações.

| 2 | 500.000  |

Então, quando devo usar cada uma?

**Escolha a Busca Sequencial se:**
- A lista de dados é pequena.
- A lista está desordenada e você só precisa realizar uma busca pontual (não vale o custo de ordenar a lista apenas para buscar uma vez).
- Você está trabalhando com estruturas de dados que não permitem acesso aleatório (como uma Lista Simplesmente Encadeada).

**Escolha a Busca Binária se:**
- A lista de dados é grande.
- A lista já está ordenada ou será ordenada uma única vez para múltiplas consultas posteriores.
- O desempenho é crítico e você tem acesso direto aos índices (como em um Array/Vetor).
