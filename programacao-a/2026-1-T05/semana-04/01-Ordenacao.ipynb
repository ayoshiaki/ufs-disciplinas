##**2. Método de Ordenação por Seleção**


O primeiro método que vamos aprender é o de Seleção. Este método se baseia na ideia de **selecionar** o menor valor dentre os valores que não estão ordenados e colocá-lo como o último dentre os valores que já estão ordenados.

Neste método se inicia com uma porção da lista com todos os valores que não estão ordenados e outra porção da lista com os valores ordenados, que sempre começa vazia.

A cada passo é escolhido o menor valor dentre os não ordenados e o mesmo é colocado no final da lista de valores ordenados. Desta forma, esta lista vai sendo ordenada gradativamente até conter todos os valores da lista não ordenada e esta estar vazia.

| passo | porção ordenada | porção ainda não ordenada |
| :--: | ---: | :--- |
| 1 | [   |  50, 10, 55, 32, **7** ] |
| 2 | [ 7 ,  | 50, **10**, 55, 32 ] |
| 3 | [ 7, 10,  | 50, 55, **32** ] |
| 4 | [ 7, 10, 32,  | **50**, 55 ] |
| 5 | [ 7, 10, 32, 50,  | **55** ] |
| 6 | [ 7, 10, 32, 50, 55,  | ] |


Observe que a porção da lista ordenada vai aos poucos recebendo os valores em ordem crescente e ao final a lista estará ordenada. Já a não ordenada vai perdendo os valores e no final ficará vazia.

Para evitar criar duas listas diferentes, o algoritmo de ordenação pode utilizar uma única lista, dividindo-a em duas partes: a **ordenada** e a **não ordenada**.

Desta forma, a tabela abaixo mostra como será implementada a solução com uma única lista. Isto é conhecido como ordenação _in-place_, ou seja, que é feita no próprio local de armazenamento. Outra observação importante é que, ao posicionar o elemento na lista ordenada, é feita uma troca com o valor que estava naquela posição, de forma que seja evitada a movimentação de outros valores desnecessariamente, o que seria bastante ineficiente.


| passo | parte ordenada | parte não ordenada | lista |
| :--: | :---: | :---: | :---: |
| 1 | [ ]  | [ 50, 10, 55, 32, 7 ] | [ 50, 10, 55, 32, 7 ] |
| 2 | [ 7 ]  | [ 10, 55, 32, 50 ] | [ **7**, 10, 55, 32, 50 ] |
| 3 | [ 7, 10 ]  | [ 55, 32, 50 ] | [ **7, 10**, 55, 32, 50 ]
| 4 | [ 7, 10, 32 ]  | [ 55, 50 ] | [ **7, 10, 32**, 55, 50 ] |
| 5 | [ 7, 10, 32, 50 ]  | [ 55 ] | [ **7, 10, 32, 50**, 55 ] |
| 6 | [ 7, 10, 32, 50, 55 ]  | [ ] | [ **7, 10, 32, 50, 55** ] |

### **2.1. Implementação do método de ordenação por Seleção em Python**


Para facilitar o trabalho e modularizar o código, foi criada a função **ordenarSelecao** que recebe a lista a ser ordenada como parâmetro. Ela utiliza outra função **indiceDoMenorDesde**, chamada repetidamente para encontrar o menor valor na lista desde uma certa posição até que toda a lista esteja ordenada.

De forma resumida, para uma lista de tamanho N, a função ordenarSelecao executa um laço for N-1 vezes, sendo que em cada iteração pega o menor valor entre os não ordenados e troca pelo valor contido na posição que corresponde à iteração atual do for, que começa com **0** e vai até o **len(l)-2**.


```python
def ordenarSelecao(l):
   for i in range(len(l)-1):
       pos_menor = indiceDoMenorDesde(i, l)
       print(f'O menor é o {l[pos_menor]} e está na posição {pos_menor}')
       if i != pos_menor:  # não troca de posição se for igual
          print(f'Trocando {l[i]} por {l[pos_menor]} na posição {i}')
          l[i], l[pos_menor] = l[pos_menor], l[i]
          print(l)

def indiceDoMenorDesde(inicio, l):
   pos_menor = inicio
   for pos in range(pos_menor+1, len(l)):
     if l[pos] < l[pos_menor]:
       pos_menor = pos
   return pos_menor

nums = [50, 10, 55, 32, 7]
#nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenarSelecao(nums)
print('Fim:', nums)
```

##**3. Método de Ordenação por Inserção**


O método de ordenação por Inserção também é bastante simples de entender. Seu funcionamento também se baseia na ideia de ter uma lista ordenada inicialmente vazia e aos poucos pegar valores da lista não ordenada para colocar na lista ordenada de forma que esta se mantenha ordenada.

Porém, como a lógica do método de ordenação por inserção é colocar o valor na lista ordenada de maneira que ela continue ordenada, deve ser encontrada a posição exata para a inserção do valor. Para que isso seja possível, todos os valores maiores que ele são deslocados uma posição para a direita para abrir espaço para a inserção.

A cada passo um valor é inserido na parte ordenada na lista, deslocando para a direita uma posição todos os valores que forem maiores que ele. Isto é repetido até que não haja mais valores na parte não ordenada da lista.

| passo | valor a ser inserido | parte ordenada | parte não ordenada | lista |
| :--: | :--: | :---: | :---: | :---: |
| 1 |    | [ 50 ]| [ 10, 55, 32, 7 ] | [ **50**, 10, 55, 32, 7 ] |
| 2 | 10 | [ 10, 50 ] | [ 55, 32, 7 ] | [ **10, 50**, 55, 32, 7 ] |
| 3 | 55 | [ 10, 50, 55 ] | [ 32, 7 ] | [ **10, 50, 55**, 32, 7 ]
| 4 | 32 | [ 10, 32, 50, 55 ]   | [ 7 ] | [ **10, 32,, 50, 55**, 7 ] |
| 5 |  7 | [ 7, 10, 32, 50, 55 ]  | [ ] | [ **7, 10, 32, 50, 55** ] |

### **3.1. Implementação do método de ordenação por Inserção em Python**

```python
def ordenarInsercao(l):
    # Percorre da posição 1 até o final da lista
    for i in range(1, len(l)):
        valor = l[i]
        print(f'Valor: {valor}')
        j = i - 1
        # Move os elementos maiores que o valor para uma posição adiante
        while j >= 0 and l[j] > valor:
            print(f'Copiando {l[j]} da posição {j} para a posição {j + 1}')
            l[j + 1] = l[j]
            j -= 1
            print(l)
        l[j + 1] = valor  # coloca o valor na posição que mantém a ordem
        print(f'Colocando {valor} na posição {j+1}\n{l}')
    return l

nums = [50, 10, 55, 32, 7]
#nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenarInsercao(nums)
print('Fim:', nums)
```

##**4. Método de Ordenação Bolha (Bubble Sorte)**


A ideia geral do algoritmo bolha é fazer comparações de valores vizinhos da esquerda para a direita e trocar os valores de posição sempre que o da esquerda for maior que o da direita. Isso faz com que na primeira passada o maior valor seja colocado na última posição da lista. Na segunda passada, será colocado no penúltimo o segundo maior valor. Daí vem o nome do algoritmo, já que o valor borbulha e vai até a posição.

De forma geral, o algoritmo consiste em pegar o elemento com índice 0 e comparar com o elemento de índice 1. Se o elemento de índice 0 for maior que o elemento de índice 1, eles serão trocados.

Em seguida, a comparação será realizada entre o elemento de índice 1 e o elemento de índice 2. Esse processo irá continuar até o último elemento.

O mesmo processo acontecerá nas iterações seguintes, até que os elementos estejam ordenados em ordem crescente.

### **4.1. Implementação do método de ordenação Bolha em Python**

```python
def ordenarBolha(lista):
    n = len(lista)
    for i in range(n):
        print('Passo', i+1)
        # Flag para otimizar, caso a lista já esteja ordenada
        trocou = False
        for j in range(0, n-i-1):
            if lista[j] > lista[j+1]:
                print('Trocando', lista[j], 'por', lista[j+1])
                lista[j], lista[j+1] = lista[j+1], lista[j]
                print('Após a troca:', lista)
                trocou = True
        if not trocou:
            break  # Se nenhuma troca foi feita, a lista está ordenada
    return lista


nums = [50, 10, 55, 32, 7]
#nums = [3,1,10,5,15,20,50,40,30,2,4]
print('Inicio:', nums)
ordenarBolha(nums)
print('Fim:', nums)
```
