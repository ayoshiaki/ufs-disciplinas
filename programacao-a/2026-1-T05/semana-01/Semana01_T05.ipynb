# Nomeação de variáveis em Python

# Valores

+ Programas manipulam valores tais como: números inteiros, cadeias de caracteres, imagens e outros.

+ Alguns valores são **primitivos**, ou seja, são atômicos no sentido em que não são formados pela composição de outros valores. Por exemplo um número.

+ Alguns valores são **compostos**, ou seja, são formados pela composição de outros valores. Por exemplo, uma sequência de caracteres (ou **cadeia de caracteres**) é um valor cujos componentes são caracteres.

+ Valores são classificados em **tipos**. Um **tipo** é um conjunto de valores com características comuns de tal forma que há funções ou operações que operam uniformemente sobre eles. <br>
Por exemplo, o tipo dos números `float` é o conjunto de números reais representáveis com 64 bits seguindo o padrão IEEE 754. Para este tipo há operações de soma, multiplicação, divisão e mais algumas outras.


## Alguns tipos predefinidos em Python:**

### `int`
+ Números inteiros sem limites (potencialmente infinitos - o limite é a memória).
+ **Operações**: `+`, `*`, `//`, `%`, `**`, `==`, `!=`, `<`, `<=`, `>`, `>=`.
+ **Funções**: `int(x)` e `abs(x)`.
+ **Constantes Literais**: 12, -230, 4500, 4_500, 0b10111, 0o756, ...

### `float`
+ Números reais representáveis com o padrão IEEE 754.
+ **Operações**:  `+`, `*`, `/`, `%`, `**`, `==`, `!=`, `<`, `<=`, `>`, `>=`.
+ **Funções**: `int(x)` e `abs(x)`.
+ **Constantes literais**: 2.56, -0.5, 1e10, ...

### `str`
+ Cadeias de caracteres. Conhecidas como *strings*.
+ **Operações**: `+`, `*`, `[_]`, `==`, `!=`, `<`, `<=`, `>`, `>=`, `in`, `not in`.
+ **Função**: `str(x)`.
+ **Constantes Literais**: "João da Silva", 'Tarifa básica'.<br>

#Como saber o tipo de um valor
A função type(  ) informa o tipo associado a uma variável naquele momento, como visto no exemplo a seguir: print(type(numero)).
Nesse caso o nome da variável testada seria **numero**.
Um exemplo de uso dessa função pode ser visto no código a seguir.


```python
numero = 5
nome = 'computador'
valor = 5.5
print('Tipos das variáveis:')
print(type(numero))
print(type(nome))
print(type(valor))
print('Ao mudar o conteúdo da variável, seu tipo também muda')
valor = 'Aracaju'
print(type(valor))
```

#Leitura em Python
A função input(  ) é uma das principais formas de se ler valores em Python. Essa função sempre lê valores do tipo string. Caso se deseje ler valores de outro tipo deve ser usada uma função de conversão.
* A função int(  ) pode ser usada para converter strings para números inteiros
* A função float(  ) pode ser usada para se converter strings para números com parte fracionária.
* A função str(  ) pode ser usada para converter valores numéricos em string.

Vejamos os exemplos a seguir.

```python
#Exemplo 1: Leitura de string
nome = input('Digite seu nome:')
print('O nome digitado foi ' + nome)
#Observe que o texto e o conteúdo da variável foram concatenados na saída usando o operador '+'

#Exemplo 2: Leitura de inteiro
quant = int(input('Digite a quantidade de alunos:'))
print('A quantidade de alunos é ')
print(quant)

#Exemplo 3: Leitura de float
nota = float(input('Digite a nota do aluno:'))
print('A nota do aluno é ')
print(nota)

# Podemos refazer as últimas linhas do exemplo 2 para que o texto e a nota sejam
# exibidos em um só print
print('A quantidade de alunos é ' + str(quant))

# Você consegue refazer o Exemplo 3, acima, para a exibição de texto e variável ser feita em uma única linha?

```

##Precedência de operadores
Operadores em Python são executados de acordo com a seguinte ordem de precedência:


1.   **
2.   *, /, //
3. %
4. +, -

Em Python, quando dois ou mais operadores têm a mesma precedência, a execução segue a associatividade definida para aquele grupo de operadores. Na maioria dos casos a associatividade é da esquerda para a direita, com exceção do operador de exponenciação (**) que é da direita para a esquerda. A ordem de precedência pode ser alterada pela colocação de parênteses. Podem ser colocados quantos pares de parênteses forem necessários.

1) Leia dois números e exiba a média

```python
# Observe que o enunciado fala que é necessário ler 2 números. Como não há restrição
# de que esses números devem ser inteiros, serão lidos números do tipo float
# Uma vez que input( ) sempre retorna string, a leitura será feita e seu retorno será
# convertido para float.

#Entradas
num1 = float(input('Digite um número:'))
num2 = float(input('Digite outro número:'))

#Processamento
media = (num1 + num2) /2

#Saída
print('Média = ',media)
```

2) Leia nome e ano de nascimento de alguém e exiba quantos anos ele tem aproximadamente

```python
# Entradas
nome = input('Digite seu nome: ') # Ao ler o nome não é necessário fazer nenhuma conversão, pois o valor lido já é string

nascimento = int(input('Digite o ano de seu nascimento: ')) # Ao ler a idade o valor retornado da função input será convertido para inteiro, pois
                                                           # o tipo da variável ano é naturalmente inteiro
#Processamento e saída
print(nome,' tem aproximadamente ', 2026-nascimento ,' anos')
```

```python
# O mesmo programa pode ser feito de uma forma mais detalhada:

#Entradas
nome = input('Digite seu nome:')
anoProv = input('Digite o ano de seu nascimento')

#Processamento
nascimento = int(anoProv)
idade = 2026 - nascimento

#Saída
print(nome,' tem aproximadamente ', idade ,' anos')
```

#Atribuições compostas
* É possível codificar de forma concisa uma operação aritmética e uma atribuição, conforme visto no exemplo a seguir:  num += 10
* Dessa forma o valor da variável é usado, a operação aritmética é feita e o resultado é guardado na própria variável. Assim, caso antes dessa instrução o valor de num fosse 5, após sua execução o valor de num seria 15.
* Nesse tipo de construção pode ser usado qualquer operador aritmético e (qualquer variável)
* Exemplos:
* valor *=2
* idade +=1
* numero /=5



```python
num = 5
print('Valor anterior de num = ',num)
num += 10
print('Novo valor de num = ',num)

valor = 8
print('Valor anterior da variável valor = ',valor)
valor /= 2
print('Novo valor da variável valor = ',valor)
```
