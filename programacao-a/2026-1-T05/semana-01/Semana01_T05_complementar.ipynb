### ⚠️ Cuidado: Precisão limitada!

Floats não são números reais — são *aproximações*.
Isso pode causar surpresas:


````python
# Isso NÃO é um bug do Python — é uma propriedade dos floats
0.1 + 0.2
````

````python
# Compare: o resultado é exatamente 0.3?
0.1 + 0.2 == 0.3
````

Isso acontece porque 0.1 e 0.2 não têm representação exata em binário,
assim como 1/3 não tem representação decimal finita (0.333...).

**Regra prática:** nunca compare floats com `==`. Veremos alternativas mais adiante.

# Parte 2: Meios de Composição (~80 min)

> *Os meios de composição permitem que elementos simples se combinem para formar elementos compostos.*

Agora que conhecemos os elementos primitivos, vamos aprender a **combiná-los**
em **expressões** mais complexas usando **operadores**.

Em Python, uma **expressão** é qualquer trecho de código que produz um valor.
O interpretador **avalia** (*evaluates*) a expressão e retorna o resultado.


## 2.1 Operadores Aritméticos

| Operador | Operação | Exemplo | Resultado |
|----------|----------|---------|-----------|
| `+` | Adição | `7 + 3` | `10` |
| `-` | Subtração | `7 - 3` | `4` |
| `*` | Multiplicação | `7 * 3` | `21` |
| `/` | Divisão (real) | `7 / 3` | `2.3333...` |
| `//` | Divisão inteira | `7 // 3` | `2` |
| `%` | Módulo (resto) | `7 % 3` | `1` |
| `**` | Exponenciação | `7 ** 3` | `343` |


````python
# Adição e Subtração
print(7 + 3)   # 10
print(7 - 3)   # 4
````

````python
# Multiplicação
print(7 * 3)   # 21
````

````python
# Divisão real: SEMPRE retorna float
print(7 / 3)    # 2.3333...
print(6 / 3)    # 2.0 (float, mesmo que o resultado seja "inteiro")
print(type(6 / 3))
````

````python
# Divisão inteira: descarta a parte decimal (trunca em direção a -∞)
print(7 // 3)    # 2
print(-7 // 3)   # -3 (não -2! trunca para baixo)
````

````python
# Módulo (resto da divisão inteira)
print(7 % 3)     # 1 (pois 7 = 2×3 + 1)
print(10 % 2)    # 0 (10 é par)
print(10 % 3)    # 1
````

### A relação fundamental entre `//` e `%`

Para quaisquer inteiros `a` e `b` (com `b ≠ 0`), vale:

$$a = (a \;\mathbin{//}\; b) \times b + (a \;\mathbin{\%}\; b)$$

A função `divmod(a, b)` retorna ambos de uma vez:


````python
# Verificando a relação: a = (a // b) * b + (a % b)
a, b = 17, 5
quociente = a // b   # 3
resto = a % b         # 2
print(f"{a} = {quociente} × {b} + {resto}")  # 17 = 3 × 5 + 2
print(f"Verificação: {quociente * b + resto}")  # 17
````

````python
# divmod retorna (quociente, resto) de uma vez
divmod(17, 5)
````

````python
# Exponenciação
print(2 ** 10)    # 1024
print(9 ** 0.5)   # 3.0 (raiz quadrada!)
print(27 ** (1/3))  # 3.0 (raiz cúbica — aproximada)
````

## 2.2 Precedência de Operadores Aritméticos

Assim como na matemática, os operadores têm uma ordem de prioridade:

| Prioridade | Operadores | Associatividade |
|------------|-----------|----------------|
| 1ª (mais alta) | `**` | Direita para esquerda |
| 2ª | `+x`, `-x` (unários) | — |
| 3ª | `*`, `/`, `//`, `%` | Esquerda para direita |
| 4ª (mais baixa) | `+`, `-` | Esquerda para direita |

**Na dúvida, use parênteses!** Eles tornam a intenção explícita e não custam nada.


````python
# Sem parênteses: * tem precedência sobre +
2 + 3 * 4       # 2 + 12 = 14, NÃO 20
````

````python
# Com parênteses: forçamos a ordem
(2 + 3) * 4     # 5 * 4 = 20
````

````python
# ** é associativo à direita:
print(2 ** 3 ** 2)     # 2 ** (3 ** 2) = 2 ** 9 = 512
print((2 ** 3) ** 2)   # 8 ** 2 = 64
````

## 2.3 Composição entre Tipos: Coerção e Conversão

Quando combinamos tipos diferentes, Python faz **promoção automática** (coerção)
para o tipo mais geral:

$$\text{int} \to \text{float} \to \text{complex}$$


````python
# int + float → float
resultado = 1 + 2.0
print(resultado, type(resultado))  # 3.0, float
````

````python
# int + complex → complex
resultado = 1 + 2j
print(resultado, type(resultado))  # (1+2j), complex
````

### Conversão explícita (casting)

Podemos converter entre tipos usando as funções construtoras:


````python
# Converter para inteiro (trunca a parte decimal)
print(int(3.7))      # 3
print(int(-3.7))     # -3
print(int('42'))     # 42 (string → int)
````

````python
# Converter para float
print(float(42))     # 42.0
print(float('3.14')) # 3.14 (string → float)
````

````python
# Converter para string
print(str(42))       # '42'
print(str(3.14))     # '3.14'
print(str(True))     # 'True'
````

````python
# Converter para bool
print(bool(0))       # False
print(bool(42))      # True  (qualquer número ≠ 0)
print(bool(''))      # False (string vazia)
print(bool('olá'))   # True  (string não-vazia)
print(bool(None))    # False
````

### Regra de *truthiness* em Python

Valores considerados **falsy** (equivalem a `False` quando convertidos para `bool`):
`0`, `0.0`, `0j`, `False`, `None`, `''` (string vazia), e coleções vazias.

**Todo o resto** é **truthy** (equivale a `True`).


## 2.4 Operadores Relacionais (Comparação)

Operadores relacionais comparam dois valores e retornam um **booleano** (`True` ou `False`).

| Operador | Significado |
|----------|-------------|
| `==` | Igual a |
| `!=` | Diferente de |
| `<` | Menor que |
| `>` | Maior que |
| `<=` | Menor ou igual a |
| `>=` | Maior ou igual a |

> ⚠️ Cuidado: `=` é **atribuição**, `==` é **comparação**.


````python
print(5 == 5)    # True
print(5 == 5.0)  # True (compara valor, não tipo)
print(5 != 3)    # True
````

````python
print(3 < 5)     # True
print(5 < 3)     # False
print(5 <= 5)    # True
````

### Comparação encadeada (uma elegância de Python!)

Python permite encadear comparações como na matemática:


````python
x = 5

# Em vez de: x > 0 and x < 10
print(0 < x < 10)     # True (x está entre 0 e 10)
print(1 <= x <= 5)    # True
print(0 < x < 3)      # False
````

### Comparação de strings

Strings são comparadas **lexicograficamente** (ordem do dicionário, baseada em Unicode):


````python
print('abc' < 'abd')    # True ('c' < 'd')
print('abc' < 'abcd')   # True (prefixo vem antes)
print('B' < 'a')        # True! (maiúsculas vêm antes de minúsculas em Unicode)
````

````python
# Verificando os valores Unicode
print(f"ord('A') = {ord('A')}")  # 65
print(f"ord('a') = {ord('a')}")  # 97
print(f"ord('B') = {ord('B')}")  # 66
````

## 2.5 Operadores Booleanos (Lógicos)

| Operador | Significado | Exemplo |
|----------|-------------|--------|
| `and` | E lógico (conjunção) | `True and False → False` |
| `or` | OU lógico (disjunção) | `True or False → True` |
| `not` | NÃO lógico (negação) | `not True → False` |

### Tabela-verdade

| `A` | `B` | `A and B` | `A or B` | `not A` |
|-----|-----|-----------|----------|---------|
| True | True | True | True | False |
| True | False | False | True | False |
| False | True | False | True | True |
| False | False | False | False | True |


````python
print(True and True)    # True
print(True and False)   # False
print(False or True)    # True
print(not True)         # False
````

### Combinando operadores relacionais e booleanos

É aqui que a composição ganha poder: podemos construir condições complexas.


````python
idade = 20
tem_carteira = True

# Pode dirigir?
pode_dirigir = idade >= 18 and tem_carteira
print(pode_dirigir)  # True
````

````python
nota = 6.5

# Aprovado: nota >= 7.0
# Recuperação: 4.0 <= nota < 7.0
# Reprovado: nota < 4.0

aprovado = nota >= 7.0
recuperacao = 4.0 <= nota < 7.0
reprovado = nota < 4.0

print(f"Nota: {nota}")
print(f"Aprovado: {aprovado}")
print(f"Recuperação: {recuperacao}")
print(f"Reprovado: {reprovado}")
````

### Avaliação em curto-circuito (*short-circuit evaluation*)

Python avalia expressões booleanas de forma "preguiçosa":
- `A and B`: se `A` é falso, não avalia `B` (resultado já é falso)
- `A or B`: se `A` é verdadeiro, não avalia `B` (resultado já é verdadeiro)

Na verdade, `and` e `or` retornam um dos **operandos**, não necessariamente `True`/`False`:


````python
# 'and' retorna o primeiro valor falsy, ou o último se todos truthy
print(1 and 2 and 3)   # 3  (todos truthy → retorna o último)
print(1 and 0 and 3)   # 0  (primeiro falsy)
print('' and 'olá')    # '' (primeiro falsy)
````

````python
# 'or' retorna o primeiro valor truthy, ou o último se todos falsy
print(0 or '' or 'padrão')  # 'padrão' (primeiro truthy)
print(0 or '' or None)      # None     (todos falsy → último)
````

### Precedência completa (do mais alto ao mais baixo)

| Prioridade | Operadores |
|------------|------------|
| 1ª | `**` |
| 2ª | `+x`, `-x` (unários) |
| 3ª | `*`, `/`, `//`, `%` |
| 4ª | `+`, `-` |
| 5ª | `==`, `!=`, `<`, `>`, `<=`, `>=` |
| 6ª | `not` |
| 7ª | `and` |
| 8ª | `or` |

Ou seja: aritméticos primeiro, depois comparações, depois lógicos.


````python
# Exemplo: tudo funciona sem parênteses por causa da precedência
x = 5
resultado = x + 1 > 3 and x * 2 < 20 or not x == 0

# Equivalente com parênteses explícitos:
resultado_explicito = ((x + 1) > 3) and ((x * 2) < 20) or (not (x == 0))

print(resultado)             # True
print(resultado_explicito)   # True
````

## 2.7 f-strings: O Meio de Composição Mais Elegante

A partir do Python 3.6, **f-strings** (formatted string literals)
são a forma mais legível e poderosa de compor strings com valores.

Basta prefixar a string com `f` e colocar expressões entre chaves `{}`:


````python
nome = 'Ada'
idade = 36

# A mágica das f-strings: expressões dentro de {}
f'{nome} tem {idade} anos'
````

````python
# Qualquer expressão válida pode ir dentro das chaves!
a = 7
b = 3
print(f'{a} + {b} = {a + b}')
print(f'{a} × {b} = {a * b}')
print(f'{a} ÷ {b} = {a / b:.4f}')  # .4f = 4 casas decimais
````

### Especificadores de formato em f-strings

Dentro das chaves, após `:`, podemos controlar a formatação:

| Formato | Descrição | Exemplo | Resultado |
|---------|-----------|---------|----------|
| `:.2f` | 2 casas decimais | `f'{3.14159:.2f}'` | `'3.14'` |
| `:.0f` | Sem casas decimais | `f'{3.7:.0f}'` | `'4'` |
| `:d` | Inteiro decimal | `f'{42:d}'` | `'42'` |
| `:>10` | Alinhado à direita (10 chars) | `f'{"oi":>10}'` | `'        oi'` |
| `:<10` | Alinhado à esquerda (10 chars) | `f'{"oi":<10}'` | `'oi        '` |
| `:^10` | Centralizado (10 chars) | `f'{"oi":^10}'` | `'    oi    '` |
| `:,` | Separador de milhares | `f'{1000000:,}'` | `'1,000,000'` |
| `:.2%` | Porcentagem | `f'{0.85:.2%}'` | `'85.00%'` |
| `:08b` | Binário com 8 dígitos | `f'{42:08b}'` | `'00101010'` |


````python
pi = 3.14159265

print(f'Pi ≈ {pi}')          # Sem formatação
print(f'Pi ≈ {pi:.2f}')      # 2 casas decimais
print(f'Pi ≈ {pi:.6f}')      # 6 casas decimais
````

````python
# Formatação de tabelas com alinhamento
print(f'{"Produto":<15} {"Preço":>10}')
print(f'{"-" * 15} {"-" * 10}')
print(f'{"Café":<15} {"R$ 8.50":>10}')
print(f'{"Pão de queijo":<15} {"R$ 5.00":>10}')
print(f'{"Água":<15} {"R$ 3.00":>10}')
````

````python
# Formatação numérica
populacao = 214000000
taxa = 0.0347

print(f'População: {populacao:,}')        # 214,000,000
print(f'Taxa de crescimento: {taxa:.2%}') # 3.47%
````

````python
# Representações em diferentes bases
numero = 255
print(f'Decimal:     {numero}')
print(f'Binário:     {numero:08b}')
print(f'Octal:       {numero:o}')
print(f'Hexadecimal: {numero:02x}')
````

### Debugging com f-strings (Python 3.8+)

O operador `=` dentro das chaves é perfeito para depuração:


````python
x = 42
y = 7
print(f'{x = }')          # x = 42
print(f'{x + y = }')      # x + y = 49
print(f'{x > 0 = }')      # x > 0 = True
````

# Parte 3: Meios de Abstração (~50 min)

> *Os meios de abstração permitem que elementos compostos sejam nomeados e manipulados como unidades.*

A abstração mais simples — e talvez mais poderosa — é dar um **nome** a um valor.
Isso é o que fazemos com **variáveis** e o operador de **atribuição** (`=`).


## 3.1 Variáveis e Atribuição

Em Python, uma variável é um **nome** que se refere a um **objeto** na memória.
O operador `=` associa (*bind*) o nome ao valor.

```
nome = expressão
```

O Python primeiro **avalia a expressão** à direita do `=`,
depois **associa** o resultado ao nome à esquerda.


````python
# Criando variáveis (abstração por nomeação)
pi = 3.14159265
raio = 5.0

# Usando as abstrações em composições
area = pi * raio ** 2
perimetro = 2 * pi * raio

print(f'Raio = {raio}')
print(f'Área = {area:.2f}')
print(f'Perímetro = {perimetro:.2f}')
````

### O poder da abstração

Compare estas duas versões do mesmo cálculo:

**Sem abstração** (críptico e repetitivo):
```python
print(3.14159265 * 5.0 ** 2)
print(2 * 3.14159265 * 5.0)
```

**Com abstração** (claro e reutilizável):
```python
pi = 3.14159265
raio = 5.0
print(pi * raio ** 2)
print(2 * pi * raio)
```

A segunda versão é mais legível, menos propensa a erros,
e fácil de modificar (basta mudar o valor de `raio` em um único lugar!).


## 3.2 Regras e Convenções para Nomes

### Regras obrigatórias (violar = `SyntaxError`)

1. Deve começar com **letra** ou **underscore** (`_`)
2. Pode conter letras, dígitos e underscores
3. **Não** pode ser uma palavra reservada de Python
4. É **case-sensitive**: `nome`, `Nome` e `NOME` são variáveis diferentes

### Convenções (PEP 8 — estilo Pythônico)

| Estilo | Uso | Exemplo |
|--------|-----|--------|
| `snake_case` | Variáveis e funções | `total_alunos`, `nota_final` |
| `UPPER_SNAKE_CASE` | Constantes | `PI`, `VELOCIDADE_LUZ` |
| `PascalCase` | Classes | `ContaBancaria`, `ListaAlunos` |
| `_prefixo` | Uso interno/privado | `_contador` |


````python
# Nomes válidos
idade = 25
nome_completo = 'Ada Lovelace'
_variavel_interna = 42
nota1 = 8.5
PI = 3.14159  # constante (por convenção)

print(idade, nome_completo, _variavel_interna, nota1, PI)
````

## 3.3 Formas de Atribuição

### Atribuição simples


````python
x = 10
y = 20
print(x, y)  # 10 20
````

### Atribuição múltipla (desempacotamento)


````python
# Atribuição múltipla
a, b, c = 1, 2, 3
print(a, b, c)  # 1 2 3
````

````python
# Troca de valores (swap) — elegância Pythônica!
x, y = 10, 20
print(f'Antes:  x={x}, y={y}')

x, y = y, x  # troca sem variável auxiliar!
print(f'Depois: x={x}, y={y}')
````

## 3.4 O Modelo de Referências de Python

É importante entender que em Python, variáveis são **nomes** (etiquetas, *labels*)
que apontam para **objetos** na memória. Não são "caixas" que contêm valores.

```
a = 42
```
Cria o objeto `42` na memória e faz o nome `a` apontar para ele.

```
b = a
```
Faz o nome `b` apontar para o **mesmo** objeto `42`. Não copia o valor.


````python
a = 42
b = a

# a e b apontam para o mesmo objeto?
print(f'a = {a}')
print(f'b = {b}')
print(f'a é b? {a is b}')  # True! Mesmo objeto.
print(f'id(a) = {id(a)}')
print(f'id(b) = {id(b)}')
````

````python
# Reatribuição: muda para onde o nome aponta
a = 42
b = a      # b aponta para 42
a = 100    # a agora aponta para 100; b continua em 42

print(f'a = {a}')  # 100
print(f'b = {b}')  # 42 (não mudou!)
````

Isso é especialmente importante para entender que **reatribuição não é mutação**.
Com tipos primitivos (int, float, str, bool), isso raramente causa confusão.
Mas se torna crucial quando trabalharmos com listas e outros objetos mutáveis.


## 3.5 Saída: `print()`

A função `print()` é o principal meio de **saída** — ela exibe valores no console.


````python
# print() aceita múltiplos argumentos, separados por espaço
print('Olá', 'mundo', '!')
````

````python
# O separador padrão é espaço, mas podemos mudar com sep=
print('2026', '03', '24', sep='-')   # 2026-03-24
print('A', 'B', 'C', sep=' → ')      # A → B → C
````

````python
# O terminador padrão é \n (nova linha), mas podemos mudar com end=
print('Carregando', end='')   # sem quebra de linha
print('...', end='')          # sem quebra de linha
print(' Pronto!')             # agora sim quebra a linha
````

````python
# Combinando print() com f-strings — o combo mais útil!
nome = 'Grace Hopper'
ano_nascimento = 1906
print(f'{nome} nasceu em {ano_nascimento}.')
````

### `print()` vs avaliação do notebook

No Colab/Jupyter, a **última expressão** de uma célula é exibida automaticamente.
Mas `print()` é necessário para:
- Exibir **múltiplos** valores em uma célula
- Exibir o **conteúdo** de strings (sem aspas, com `\n` processado)
- Exibir valores no **meio** do código (não apenas no final)


````python
# Sem print: apenas o último valor aparece
1 + 1  # esse desaparece...
2 + 2  # esse também...
3 + 3  # só esse aparece!
````

````python
# Com print: todos aparecem
print(1 + 1)
print(2 + 2)
print(3 + 3)
````

## 3.6 Entrada: `input()`

A função `input()` permite que o programa **leia dados do usuário**.
Ela exibe um *prompt* (mensagem) e espera que o usuário digite algo.

> ⚠️ `input()` **sempre retorna uma string**, mesmo que o usuário digite um número!


````python
# Lendo uma string
nome = input('Qual é o seu nome? ')
print(f'Olá, {nome}! Bem-vindo(a)!')
````

````python
# Lendo um número: é necessário converter!
idade_str = input('Qual a sua idade? ')
print(f'Tipo recebido: {type(idade_str)}')  # <class 'str'>

idade = int(idade_str)  # conversão: str → int
print(f'Daqui a 10 anos você terá {idade + 10} anos.')
````

````python
# Forma mais concisa: converter diretamente
peso = float(input('Qual o seu peso (kg)? '))
altura = float(input('Qual a sua altura (m)? '))

imc = peso / altura ** 2
print(f'Seu IMC é {imc:.1f}')
````

### Padrão completo: entrada → processamento → saída

A maioria dos programas segue este fluxo:

1. **Entrada:** ler dados (do teclado, arquivo, rede...)
2. **Processamento:** computar resultados (usando composição e abstração)
3. **Saída:** exibir resultados (`print()`, salvar em arquivo...)

Vamos ver dois exemplos completos:


### Exemplo completo 1: Conversor de temperatura


````python
# ================================================
# CONVERSOR DE TEMPERATURA — Celsius → Fahrenheit
# ================================================

# 1. ENTRADA
celsius = float(input('Temperatura em Celsius: '))

# 2. PROCESSAMENTO
fahrenheit = celsius * 9 / 5 + 32

# 3. SAÍDA
print(f'{celsius:.1f}°C = {fahrenheit:.1f}°F')
````

### Exemplo completo 2: Cálculo de troco


````python
# ==========================================
# CALCULADORA DE TROCO
# ==========================================

# 1. ENTRADA
preco = float(input('Preço do produto (R$): '))
pagamento = float(input('Valor pago (R$): '))

# 2. PROCESSAMENTO
troco = pagamento - preco

# Decompor em notas e moedas (trabalhando em centavos para evitar erros de float)
troco_centavos = round(troco * 100)

notas_100 = troco_centavos // 10000
troco_centavos %= 10000

notas_50 = troco_centavos // 5000
troco_centavos %= 5000

notas_20 = troco_centavos // 2000
troco_centavos %= 2000

notas_10 = troco_centavos // 1000
troco_centavos %= 1000

notas_5 = troco_centavos // 500
troco_centavos %= 500

notas_2 = troco_centavos // 200
troco_centavos %= 200

moedas_1 = troco_centavos // 100
troco_centavos %= 100

moedas_50 = troco_centavos // 50
troco_centavos %= 50

moedas_25 = troco_centavos // 25
troco_centavos %= 25

moedas_10 = troco_centavos // 10
troco_centavos %= 10

moedas_5 = troco_centavos // 5
troco_centavos %= 5

moedas_1c = troco_centavos

# 3. SAÍDA
print(f'\nTroco: R$ {troco:.2f}')
print(f'  Notas de R$100: {notas_100}')
print(f'  Notas de R$ 50: {notas_50}')
print(f'  Notas de R$ 20: {notas_20}')
print(f'  Notas de R$ 10: {notas_10}')
print(f'  Notas de R$  5: {notas_5}')
print(f'  Notas de R$  2: {notas_2}')
print(f'  Moedas de R$1:  {moedas_1}')
print(f'  Moedas de R$0.50: {moedas_50}')
print(f'  Moedas de R$0.25: {moedas_25}')
print(f'  Moedas de R$0.10: {moedas_10}')
print(f'  Moedas de R$0.05: {moedas_5}')
print(f'  Moedas de R$0.01: {moedas_1c}')
````

## Revisando os Três Pilares

Ao longo deste tutorial, vimos como Python implementa os três mecanismos fundamentais
do SICP:

### 1. Elementos Primitivos
Os "átomos" da linguagem — os valores mais simples que o interpretador reconhece:
- Tipos numéricos: `int`, `float`, `complex`
- Tipo lógico: `bool`
- Tipo textual: `str`
- Ausência: `None`

### 2. Meios de Composição
As formas de combinar primitivos em expressões mais complexas:
- Operadores aritméticos: `+`, `-`, `*`, `/`, `//`, `%`, `**`
- Operadores relacionais: `==`, `!=`, `<`, `>`, `<=`, `>=`
- Operadores booleanos: `and`, `or`, `not`
- Concatenação e repetição de strings: `+`, `*`
- f-strings: `f'...{expr}...'`
- Conversões de tipo: `int()`, `float()`, `str()`, `bool()`

### 3. Meios de Abstração
Os mecanismos para nomear e reutilizar valores:
- Variáveis e atribuição: `nome = valor`
- Atribuição múltipla e aumentada
- Entrada: `input()` — dá ao usuário o poder de fornecer valores
- Saída: `print()` — comunica resultados

Nos próximos tutoriais, veremos meios de composição e abstração ainda mais poderosos:
