# Booleanos
Python oferece o tipo `bool` para os valores de verdade, típicos da lógica matemática. O tipo tem apenas dois valores: `True`e `False`.
Python também oferece operadores booleanos que se correspondem aos conectivos lógicos da matemática.
+ Para a conjunção (e lógico) usa-se `and` como operador binário infixo.
+ Para a disjunção (ou lógico) usa-se `or` como operador binário infixo.
+ Para a negação (não lógico) usa-se  `not` como operador unário prefixo.

Os códigos a seguir apresentam o resultado destes operadores para todas as combinações possíveis de valores de verdade.


````python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
````

````python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
````

````python
print(not True)
print(not False)
````

## Operadores relacionais
Operadores relacionais retornam como resultado valores booleanos.
Dentre eles temos operadores de comparação `<`, `<=`, `>`, `>=`, `==` e `!=`.


````python
print(1 <= 3)
print(1 >= 3)
print(1 == 3)
print(1 != 3)
````

Operadores relacionais permitem também comparar Strings. Neste caso o critério de comparação é a ordem lexicográfica.

````python
print("amigo" <= "inimigo")
print("amigo" >= "inimigo")
````

Tuplas também podem ser comparadas, desde que seus componentes sejam comparáveis. A comparação é feita elemento por elemento (ordem lexicográfica):

+ O Python compara o primeiro item da primeira tupla com o primeiro item da segunda. Se forem diferentes, essa comparação já define o resultado final.
+ Se forem iguais, ele passa para o próximo componente, e assim por diante.
+ Se todos os elementos comparados forem iguais, mas uma das tuplas for mais curta, a tupla mais curta é considerada a "menor".

````python
print((1, 2, 3) < (1, 2, 4))
print((1, 2, 3) < (1, 3, 2))
print((1, 2, 3) < (0, 3, 2))
````

## Precedências
De maior para menor
+ Aritméticos vistos anteriormente
+ Relacionais (`<`, `<=`, `>`, `>=`, `==` e `!=`, `in`, `not in`).
+ `not`
+ `and`
+ `or`

Isto implica que, por exemplo
```python
x < y or not z < 0 and w == 10
```
deve ser entendido como
```python
(x < y) or ( (not (z < 0)) and (w == 10) )
```

## Operadores lógicos são preguiçosos: avaliação de curto-circuito.

Observe que para que `A and B` seja `False`, basta que A seja `False`.
Quando Python avalia que `A and B`, avalia primeiro `A` e se este for `False` termina com resultado final `False`. Observe que `B` não é avaliado neste caso. Este tipo de avaliação é denominada de **avaliação de curto-circuito**.

Similarmente, para que `A or B` seja `True`, basta que `A` seja `True` e, neste caso, `B` não precisa ser avaliado.

Veremos, mais na frente, que esta característica preguiçosa dos operadores lógicos têm algumas sutilezas e que, em alguns casos especiais, a comutatividade do `and` e do `or` não é válida.

## Simplificação de expressões booleanas

### Prefira afirmações a negações

Em geral, expressões que afirmam algo são mais diretas e fáceis de compreender que aquelas que afirmam a sua negação. Por exemplo, considere as duas seguintes afirmações equivalentes
  + João não é de maior idade
  + João é de menor idade

Qual é mais direta? Qual exige menos esforço mental para entender?

Cada operadores relacional tem outro que se corresponde com a negação da relação. Isto permite evitar aplicações do operador `not`.

Assim, por exemplo
| Ao invés de | Prefira |
| ------------| --------|
| ```not (idade > 18)``` | ```idade <= 18```|
| ```not x == y``` | ```x != y``` |
| ```not x != y``` | ```x == y```|

### Não compare uma expressão booleana com `True` ou com `False`

A expressão booleana `E == True` é equivalente com `E`. Obviamente, a opção `E` é mais simples.

A expressão booleana `E == False` é equivalente com `not E`. Prefira usar `not E`.

### Utilize propriedades algébricas da lógica matemática para simplificar expressões.

A seguir temos algumas propriedades algébricas frequentemente úteis. Nelas usamos `<==>` como símbolo para a equivalência lógica. Este operador não existe em Python.

**Negação dupla**
```python
 not (not A) <==> A
```

**Leis de De Morgan**
```python
not (A and B) <==> not A or not B
```

**Leis Distributivas/Fatoração**
```python
A and (B or C) <==> A and B  or  A and C
A or (B and C) <==> (A or B) and (A or C)
```

Quando a igualdade é lida de esquerda para a direita, ambas leis são vistas como *distributivas*. A primeira diz que o `and` distribui sobre o `or` enquanto que a segunda diz que o `or` distribui sobre o `and`.

Agora, quando as igualdades são lidas de direita para a esquerda, ambas as leis podem ser vistas como um tipo de  *fatoração algébrica* ou *colocar em evidência*.

**Leis Comutativas**
```python
A and B <==> B and A
A or B <==> B or A
```

**Elemento neutro**
```python
True and A <==> A and True <==> A
False or A <==> A or False <==> A
```

**Elemento absortivo ou anulador**
```python
False and A <==> A and False <==> False
True or A <==> A or True <==> True
```

### Encadeamento de opeadores relacionais

Na matemática é comum abreviar expressões relacionais usando *encadeamentos*. Por exemplo para exprimir os relacionamentos $x < y$ e $y < z$, na matemática abrevia-se assim $x < y < z$. Encadeamentos como estes são permitidos somente apenas se todos os operadores "apontam" na mesma direção, pois
como as relações de ordem são transitivas, a combinação também expressa uma relação entre o primeiro e último valor. Assim por exemplo, se temos $x < y \leq z$, além de termos que $x < y$ e $y \leq z$, por transitividade temos também que $x < z$.

Python permite encadeamentos de operadores relacionais. No entanto, Python é mais permissivo de tal forma que expressões como `x > y < z` também são válidas. Neste caso devemos interpretar que `y` é menor que `x` e `z` ao mesmo tempo. Em outras palavras, a expressão é equivalente com `x > y and y < z`.
